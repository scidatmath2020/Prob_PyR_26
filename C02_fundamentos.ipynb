# Fundamentos de Probabilidad

La vida cotidiana está llena de situaciones inciertas. Cuando un técnico de mantenimiento industrial escucha un ruido anómalo en una máquina, no sabe con certeza si fallará en las próximas horas. Cuando un analista de control de calidad toma una muestra de productos de una línea de ensamblaje, no sabe exactamente cuántos defectuosos habrá en toda la producción del día. Cuando un especialista en diagnóstico de fallos revisa los síntomas de un sistema electrónico, no tiene la seguridad absoluta de cuál componente se averió.

En todos estos casos, sin embargo, es posible hacer afirmaciones cuantitativas sobre el grado de creencia o la frecuencia esperada de los distintos sucesos. Esa es precisamente la función de la teoría de la probabilidad: proporcionar un marco matemático riguroso para manejar la incertidumbre.

A lo largo de este capítulo, construiremos los cimientos de ese marco. Comenzaremos con las definiciones fundamentales —experimento aleatorio, espacio muestral y evento— para luego establecer las reglas que rigen el cálculo de probabilidades.

## Experimento aleatorio y espacio muestral

Imaginemos que lanzamos un dado de seis caras, perfectamente equilibrado. Antes de que el dado caiga, no podemos predecir el resultado con certeza. Podemos, eso sí, enumerar todos los resultados posibles: 1, 2, 3, 4, 5 o 6. Esta situación —un procedimiento que produce resultados inciertos, pero cuyo conjunto de resultados posibles es conocido de antemano— recibe el nombre de experimento aleatorio.

Formalmente, definimos el espacio muestral, denotado por la letra griega omega mayúscula $\Omega$, como el conjunto de todos los resultados posibles de un experimento aleatorio. En el caso del dado: $$\Omega=\{1,2,3,4,5,6\}$$

Los espacios muestrales pueden tener distinta naturaleza. Si lanzamos una moneda tres veces, el espacio muestral contiene todas las secuencias posibles de caras (C) y cruces (X):
$$\Omega=\{CCC,CCX,CXC,XCC,CXX,XCX,XXC,XXX\}$$

Nótese que aquí $\Omega$ tiene ocho elementos. Si medimos el tiempo de vida de un componente electrónico, el espacio muestral es muy distinto:
$$\Omega=\{t\in\mathbb{R}\,:\,t\ge0\}=[0,\infty)$$


Es decir, todos los números reales no negativos. En este caso, $\Omega$ es un conjunto infinito no numerable. Como veremos más adelante, la naturaleza del espacio muestral (finito, infinito numerable, infinito no numerable) determinará las herramientas matemáticas que necesitemos emplear.

Un **evento** es simplemente un subconjunto del espacio muestral. Por ejemplo, en el lanzamiento de un dado, podemos definir el evento $A=$"obtener un número par". En términos de conjuntos, $$A=\{2,4,6\}\subseteq\Omega=\{1,2,3,4,5,6\}$$

Diremos que un evento ocurre si el resultado del experimento pertenece a dicho subconjunto. Si el dado muestra un 4, entonces ha ocurrido $A$; si muestra un 3, no ha ocurrido. El evento seguro es $\Omega$ mismo, el evento imposible es el conjunto vacío $\emptyset$, y el complemento de un evento $A$, denotado por $A^c$, es el conjunto de resultados que no están en $A$.

---

**Ejemplo del técnico: eventos.**

- $A$ = "obtener un número par" = $\{2, 4, 6\}$
- $B$ = "obtener un número mayor o igual a 5" = $\{5, 6\}$
- $C$ = "obtener un 6" = $\{6\}$
- El **complemento** de $A$, denotado $A^c$, es lo que no está en $A$: $A^c = \{1, 3, 5\}$
---

Podemos explorar estos conceptos directamente en R y Python. Aunque parezca sencillo, la capacidad de generar espacios muestrales y simular experimentos aleatorios es fundamental para todo lo que sigue.

```r
####################### EN R
# Definir el espacio muestral de un dado
espacio_muestral <- 1:6

# Definir un evento: números pares
evento_par <- espacio_muestral[espacio_muestral %% 2 == 0]

# Simular 10 lanzamientos de un dado
simulacion <- sample(espacio_muestral, size = 20, replace = TRUE)

# Verificar en cuáles ocurrió el evento "par"
ocurrencias <- simulacion %in% evento_par
ocurrencias
```

```python
####################### EN Python
import numpy as np

# Definir el espacio muestral de un dado
espacio_muestral = np.array([1, 2, 3, 4, 5, 6])

# Definir un evento: números pares
evento_par = espacio_muestral[espacio_muestral % 2 == 0]

# Simular 10 lanzamientos de un dado
simulacion = np.random.choice(espacio_muestral, size=20, replace=True)

# Verificar en cuáles ocurrió el evento "par"
ocurrencias = np.isin(simulacion, evento_par)
ocurrencias
```

## Tres visiones de la probabilidad

A lo largo de la historia, la probabilidad se ha entendido de tres maneras diferentes, cada una con su propio ámbito de aplicación. Conviene conocer las tres, pues todas aparecerán en distintos momentos del libro.

### La definición clásica o de Laplace

El matemático francés Pierre-Simon Laplace propuso en el siglo XVIII una definición sencilla: si un experimento tiene $N$ resultados posibles, todos igualmente probables, y un evento $A$ tiene $n$ resultados favorables, entonces la probabilidad de $A$ es $$P(A)=\frac{n}{N}$$

Por ejemplo, al lanzar un dado equilibrado, la probabilidad de obtener un número par es $3/6=1/2$. Al extraer una carta de una baraja española de 40 cartas, la probabilidad de obtener un oro es $10/40=1/4$.

La definición clásica es intuitiva y funciona perfectamente en juegos de azar y situaciones simétricas. Sin embargo, tiene una limitación fundamental: solo se aplica cuando podemos asegurar que todos los resultados son equiprobables. ¿Qué ocurre si lanzamos un dado que sospechamos está cargado? En ese caso, la definición de Laplace ya no sirve.

### La definición frecuentista

Ante la limitación anterior, surge la visión frecuentista: la probabilidad de un evento es el límite de su frecuencia relativa cuando el número de repeticiones del experimento tiende a infinito. En símbolos: $$P(A)=\lim_{n\to\infty}\frac{n_A}{n}$$ donde $n_A$ es el número de veces en que ocurre $A$ en $n$ repeticiones del experimento.

Esta definición es operativa: podemos estimar la probabilidad de un evento realizando muchas repeticiones. Si lanzamos un dado 10,000 veces y observamos que el número 6 aparece en 1,800 de ellas, estimaremos $P(6)\approx0.18$, lo que sugiere que el dado está cargado a favor del 6.


---
Podemos ilustrar la visión frecuentista con una simulación sencilla. A medida que aumentamos el número de lanzamientos de un dado, la frecuencia relativa de cada cara se acerca a 1/6.

```r
####################### EN R

# Simular lanzamientos de un dado y ver la convergencia
espacio_muestral = 1:6
n_lanzamientos = 10
lanzamientos = sample(espacio_muestral,size=n_lanzamientos,replace = TRUE)

sum(lanzamientos == 6)/n_lanzamientos
```

```python
####################### EN PYTHON
import numpy as np
import pandas as pd

espacio_muestral = np.arange(1,7)
n_lanzamientos = 10
lanzamientos = np.random.choice(espacio_muestral,size=n_lanzamientos,replace = True)

np.sum(lanzamientos == 6)/n_lanzamientos
```

### La definición axiomática de Kolmogórov

El matemático ruso Andrei Kolmogórov resolvió estas dificultades en 1933 al proponer una definición axiomática. En lugar de decir "qué es" la probabilidad, Kolmogórov estableció las propiedades que debe cumplir cualquier asignación de probabilidades que pretenda ser coherente. Son tres axiomas:

- Para cualquier evento $A$, $0\le P(A)\le 1$
- $P(\Omega)=1$
- Si $A_1,A_2,A_3,...$ son eventos mutuamente excluyentes (es decir, no pueden ocurrir dos a la vez), entonces la probabilidad de su unión es la suma de sus probabilidades individuales.

Esta formulación axiomática tiene la gran ventaja de ser general: cualquier función que asigne números reales a eventos cumpliendo estos tres axiomas es una probabilidad válida, sin importar si es la visión clásica, frecuentista o cualquier otra.



## Reglas básicas de probabilidad

A partir de los axiomas de Kolmogórov podemos deducir todas las reglas que utilizaremos para calcular probabilidades. Estas reglas son pocas, pero muy poderosas.

- **El complemento.** Si la probabilidad de todo el espacio muestral es 1, y $A$ es un evento entonces
$$P(A^c)=1-P(A)$$

Esta regla es sencilla pero útil: a veces es más fácil calcular la probabilidad de que un evento no ocurra y restar de 1.

**Ejemplo para programar.**  ¿Cuál es la probabilidad de que al lanzar dos dados, la suma sea al menos 3? Es más fácil calcular la probabilidad del complemento: que la suma sea 2. La suma 2 ocurre solo cuando ambos dados muestran 1. Hay 36 resultados posibles, todos equiprobables, por lo que $P(suma=2)=1/36$. Entonces $P(suma\ge3)=1−1/36=35/36$.

- **La unión de eventos.** ¿Cómo calculamos la probabilidad de que ocurra al menos uno de los eventos $A$ o $B$? Intuitivamente, sumamos las probabilidades individuales, pero esto contaría dos veces los resultados que pertenecen tanto a $A$ como a $B$. Por tanto:
$$P(A\cup B)=P(A)+P(B)−P(A\cap B)$$

Si $A$ y $B$ son mutuamente excluyentes (no pueden ocurrir juntos), entonces $P(A\cap B)=0$ y la fórmula se simplifica a $$P(A\cup B)=P(A)+P(B)$$

**Ejemplo:** En una fábrica, el 15% de los productos tiene un defecto estético, el 10% tiene un defecto funcional, y el 5% tiene ambos defectos. ¿Qué proporción de productos tiene al menos un defecto?

$$P(A\cup B)=0.15+0.10−0.05=0.20$$

Es decir, el 20% de los productos presenta algún tipo de defecto.

- **La diferencia de eventos.** La probabilidad de que ocurra $A$ pero no $B$ se obtiene restando a $A$ su intersección con 
$B$: $$P(A∖B)=P(A)-P(A\cap B)$$

**Ejemplo:** En una fábrica, el 15% de los productos tiene un defecto estético, el 10% tiene un defecto funcional, y el 5% tiene ambos defectos. ¿Qué proporción de productos que tiene defecto estético pero no funcional? 
$$0.15-0.05=0.10$$

**Ejemplo para programar.** Del ejemplo para programar anterior, con los eventos $A$ y $B$ calcular $P(A∖B)$

## La probabilidad como medida de conjuntos

Conviene detenerse un momento para apreciar la belleza del enfoque axiomático. La probabilidad no es más que una función que asigna números a conjuntos (los eventos) respetando las reglas que hemos visto. Esta visión nos permite trasladar toda la intuición que tenemos sobre áreas, longitudes o volúmenes al ámbito de la incertidumbre.

Pensemos en un espacio muestral $Ω$ como si fuera una figura geométrica de área total 1. Cada evento $A$ es una región dentro de esa figura, y su probabilidad $P(A)$ es el área de esa región. Desde esta perspectiva, la regla de la unión 
$$P(A\cup B)=P(A)+P(B)-P(A\cap B)$$ no es más que el principio de inclusión-exclusión para áreas: al sumar las áreas de $A$ y $B$, hemos contado dos veces la intersección, por lo que debemos restarla una vez.

Esta analogía geométrica es más que una simple ayuda didáctica. En problemas más avanzados, especialmente cuando trabajamos con variables aleatorias continuas, la probabilidad se definirá efectivamente como una integral —es decir, como el área bajo una curva— y todas las reglas que hemos establecido seguirán siendo válidas.

Podemos visualizar esta idea de manera sencilla simulando muchos puntos aleatorios dentro de un cuadrado y contando cuántos caen dentro de una región específica.

```r
# Simular puntos en un cuadrado [0,1] x [0,1]
n_puntos <- 10000
x <- runif(n_puntos)
y <- runif(n_puntos)

# Definir un evento: puntos dentro de un círculo de radio 0.5 centrado en (0.5, 0.5)
radio <- 0.5
dentro_circulo <- (x - 0.5)^2 + (y - 0.5)^2 <= radio^2

# La proporción de puntos dentro del círculo estima su área (πr² = π/4 ≈ 0.785)
prob_estimada <- mean(dentro_circulo)
print(paste("Área estimada del círculo:", prob_estimada))
print(paste("Área teórica:", pi * radio^2))

```

```python
import numpy as np

n_puntos = 10000
x = np.random.uniform(0, 1, n_puntos)
y = np.random.uniform(0, 1, n_puntos)

radio = 0.5
dentro_circulo = (x - 0.5)**2 + (y - 0.5)**2 <= radio**2

prob_estimada = np.mean(dentro_circulo)
print(f"Área estimada del círculo: {prob_estimada}")
print(f"Área teórica: {np.pi * radio**2}")

```

Este ejemplo es, en esencia, una simulación de Monte Carlo: usamos números aleatorios para estimar una cantidad determinista (el área de un círculo). Veremos muchas aplicaciones de esta idea a lo largo del curso.

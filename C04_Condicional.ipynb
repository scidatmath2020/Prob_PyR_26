# Probabilidad Condicional e Independencia

Anteriormente aprendimos a calcular probabilidades de eventos en situaciones estáticas. Pero el mundo real no funciona así. Constantemente recibimos nueva información que actualiza nuestras creencias.

Imagina que trabajas en una planta de manufactura. La probabilidad de que una máquina falle en un día cualquiera es baja, quizás 0.01. Pero de repente, el sensor de temperatura se dispara y el motor empieza a vibrar. Con esta nueva información, tu estimación de la probabilidad de fallo cambia drásticamente. Eso es la probabilidad condicional.

En este capítulo aprenderemos a responder preguntas como: *dado que ha ocurrido un evento, ¿cómo se modifica la probabilidad de otro?* Este concepto es la puerta de entrada al razonamiento bayesiano, una de las herramientas más poderosas en ciencia de datos, diagnóstico de fallos y control de calidad.

## ¿Qué significa "dado que"?

Supongamos que lanzamos un dado equilibrado. La probabilidad de obtener un 4 es $1/6$. Ahora añadamos información: alguien mira el dado y nos dice que el resultado es un número par. ¿Sigue siendo $1/6$ la probabilidad de que sea un 4?

No. Sabiendo que el resultado es par, el espacio muestral se reduce de $\{1,2,3,4,5,6\}$ a $\{2,4,6\}$. Dentro de este nuevo espacio, un solo resultado nos favorece: el 4. Por lo tanto, la probabilidad de que sea un 4 dado que es par es $1/3$.

Esta es la esencia de la probabilidad condicional: la información restringe el espacio muestral. No estamos calculando una probabilidad nueva, sino actualizando nuestra creencia a la luz de nueva evidencia.

Una forma útil de pensar en esto es: antes de recibir información, trabajamos con el espacio muestral completo. Después de recibir información, ese espacio se reduce, y las probabilidades se recalculan dentro de este nuevo espacio reducido.

## Definición formal

La probabilidad de que ocurra el evento $A$ dado que ocurrió el evento $B$ se denota $P(A|B)$ y se define como:

$$P(A|B) = \frac{P(A \cap B)}{P(B)}$$

siempre que $P(B) > 0$.

Interpretación: De la porción del espacio muestral donde ocurre $B$, ¿qué fracción corresponde también a $A$?

Es crucial entender que $P(A|B)$ y $P(B|A)$ son, en general, diferentes. La probabilidad de tener fiebre dado que tienes gripe no es lo mismo que la probabilidad de tener gripe dado que tienes fiebre. Esta distinción es fundamental y será la base del razonamiento bayesiano.

También es importante notar que $P(A|B)$ es una probabilidad propiamente dicha: satisface todos los axiomas de probabilidad cuando se restringe al espacio muestral $B$. Es decir, si consideramos $B$ como nuestro nuevo espacio muestral, $P(\cdot|B)$ es una medida de probabilidad válida.

## Ejemplos de probabilidad condicional

**Ejemplo 1: Dados**

Para el ejemplo anterior: $A = \{4\}$, $B = \{2,4,6\}$.

$$P(A|B) = \frac{P(A \cap B)}{P(B)} = \frac{P(\{4\})}{P(\{2,4,6\})} = \frac{1/6}{3/6} = \frac{1}{3}$$

**Ejemplo 2: Control de calidad en una fábrica**

En una fábrica, el 5% de los productos tiene defecto grave $G$, el 10% tiene defecto leve $L$, y el 2% tiene ambos. Si un producto tiene defecto leve, ¿cuál es la probabilidad de que también tenga defecto grave?

$$P(G|L) = \frac{P(G \cap L)}{P(L)} = \frac{0.02}{0.10} = 0.20$$

Interpretación: El 20% de los productos con defecto leve también tienen defecto grave. Esto es muy distinto a decir que el 2% de todos los productos tienen ambos defectos. La condición "dado que tiene defecto leve" ha reducido el espacio muestral del 100% de productos al 10% que tiene defecto leve, y dentro de ese subconjunto, el 20% tiene también defecto grave.

**Ejemplo 3: Preferencias de consumo**

Una empresa de telecomunicaciones sabe que el 70% de sus clientes tiene contratado internet, el 45% tiene contratada televisión, y el 35% tiene ambos servicios. Si seleccionamos un cliente al azar y resulta que tiene internet, ¿cuál es la probabilidad de que también tenga televisión?

Definimos:
- $I$: cliente tiene internet
- $T$: cliente tiene televisión

$$P(T|I) = \frac{P(T \cap I)}{P(I)} = \frac{0.35}{0.70} = 0.50$$

Interpretación: La mitad de los clientes que tienen internet también contratan televisión. Esto sugiere una oportunidad de venta cruzada.

**Ejemplo 4: Segmentación de clientes**

Una tienda en línea tiene tres categorías de clientes: nuevos (40%), recurrentes (35%) y VIP (25%). La probabilidad de que un cliente realice una compra en el próximo mes es 0.10 para nuevos, 0.30 para recurrentes y 0.60 para VIP. Si sabemos que un cliente realizó una compra, ¿cuál es la probabilidad de que sea VIP?

Primero calculamos la probabilidad total de compra (que es un paso necesario para aplicar la definición de condicional):

$$P(Compra) = 0.40(0.10) + 0.35(0.30) + 0.25(0.60) = 0.04 + 0.105 + 0.15 = 0.295$$

Luego:
$$P(VIP|Compra) = \frac{P(VIP \cap Compra)}{P(Compra)} = \frac{0.25(0.60)}{0.295} = \frac{0.15}{0.295} \approx 0.508$$

Interpretación: Aunque los VIP son solo el 25% de los clientes, representan más de la mitad de las compras.

## La regla del producto

De la definición de probabilidad condicional podemos despejar una expresión para la probabilidad de la intersección:

$$P(A \cap B) = P(A|B) \cdot P(B)$$

También podemos escribirla de la otra forma:

$$P(A \cap B) = P(B|A) \cdot P(A)$$

Esta regla es extremadamente útil cuando la probabilidad condicional es más fácil de obtener que la probabilidad conjunta. Es la base para construir diagramas de árbol y para modelar procesos secuenciales.

Observa que de estas dos expresiones igualadas obtenemos:

$$P(A|B) \cdot P(B) = P(B|A) \cdot P(A)$$

Que es una relación fundamental entre probabilidades condicionales, aunque no la desarrollaremos aquí.

## Ejemplos de la regla del producto

**Ejemplo 1: Extracción sin reposición**

Una caja contiene 5 bolas rojas y 3 azules. Se extraen dos bolas sin reposición. ¿Cuál es la probabilidad de que ambas sean rojas?

Definimos:
- $R_1$: primera bola roja
- $R_2$: segunda bola roja

$$P(R_1 \cap R_2) = P(R_1) \cdot P(R_2|R_1) = \frac{5}{8} \cdot \frac{4}{7} = \frac{20}{56} = \frac{5}{14} \approx 0.357$$

**Ejemplo 2: Componentes electrónicos**

Un sistema tiene dos componentes en serie (ambos deben funcionar para que el sistema funcione). El componente A tiene probabilidad 0.95 de funcionar. Si A funciona, el componente B tiene probabilidad 0.90 de funcionar. ¿Cuál es la probabilidad de que el sistema funcione?

$$P(A \cap B) = P(A) \cdot P(B|A) = 0.95 \times 0.90 = 0.855$$

**Ejemplo 3: Encuesta de opinión**

En una ciudad, el 60% de los adultos leen el periódico. Entre los que leen el periódico, el 70% está informado sobre política. Entre los que no leen, solo el 30% está informado. ¿Cuál es la probabilidad de que una persona al azar lea el periódico y esté informada?

$$P(Leer \cap Informado) = P(Leer) \cdot P(Informado|Leer) = 0.60 \times 0.70 = 0.42$$

**Ejemplo 4: Juego de cartas**

De una baraja española de 40 cartas (4 palos con 10 cartas cada uno), se extraen dos cartas sin reposición. ¿Cuál es la probabilidad de que ambas sean del mismo palo?

Definimos:
- Primera carta: cualquier palo (probabilidad 1)
- Segunda carta: debe ser del mismo palo que la primera, y quedan 9 cartas de ese palo entre 39 cartas restantes

$$P(\text{ mismo palo}) = 1 \cdot \frac{9}{39} = \frac{9}{39} = \frac{3}{13} \approx 0.231$$

**Ejemplo 5: Control de calidad en dos etapas**

Una pieza pasa por dos inspecciones. La probabilidad de pasar la primera inspección es 0.92. Si pasa la primera, la probabilidad de pasar la segunda es 0.95. ¿Cuál es la probabilidad de que una pieza pase ambas inspecciones?

$$P(I_1 \cap I_2) = 0.92 \times 0.95 = 0.874$$

## Generalización: la regla del producto para múltiples eventos

Para tres eventos:

$$P(A \cap B \cap C) = P(A) \cdot P(B|A) \cdot P(C|A \cap B)$$

Y así sucesivamente. Esta es la base de las probabilidades en procesos secuenciales y en sistemas con múltiples etapas.

Es importante notar que el orden en que condicionamos importa, pero el resultado final es el mismo independientemente del orden. Por ejemplo:

$$P(A \cap B \cap C) = P(C) \cdot P(B|C) \cdot P(A|B \cap C)$$

También es válido, siempre que las probabilidades condicionales estén bien definidas.

La regla del producto para múltiples eventos se puede interpretar como: la probabilidad de que ocurra una secuencia de eventos es igual a la probabilidad del primero, multiplicada por la probabilidad del segundo dado el primero, multiplicada por la probabilidad del tercero dados los dos primeros, y así sucesivamente.

## Ejemplos de regla del producto para múltiples eventos

**Ejemplo 1: Línea de producción con tres etapas**

Una línea de producción tiene tres etapas en serie. La probabilidad de que la etapa 1 falle es 0.10. Si la etapa 1 funciona, la etapa 2 falla con probabilidad 0.05. Si las dos primeras funcionan, la etapa 3 falla con probabilidad 0.02. ¿Cuál es la probabilidad de que el proceso completo funcione?

Definimos $E_i$ como el evento "la etapa $i$ funciona". Entonces:
- $P(E_1) = 0.90$
- $P(E_2|E_1) = 0.95$
- $P(E_3|E_1 \cap E_2) = 0.98$

$$P(E_1 \cap E_2 \cap E_3) = P(E_1) \cdot P(E_2|E_1) \cdot P(E_3|E_1 \cap E_2)$$
$$= 0.90 \times 0.95 \times 0.98 = 0.8379$$

**Ejemplo 2: Selección de personal**

Una empresa tiene un proceso de selección de tres fases. La probabilidad de pasar la primera fase es 0.60. Dado que se pasa la primera, la probabilidad de pasar la segunda es 0.50. Dado que se pasan las dos primeras, la probabilidad de pasar la tercera es 0.40. ¿Cuál es la probabilidad de que un candidato sea contratado?

$$P(F_1 \cap F_2 \cap F_3) = 0.60 \times 0.50 \times 0.40 = 0.12$$

Solo el 12% de los candidatos completan todo el proceso.

**Ejemplo 3: Transmisión de información**

Un mensaje se transmite a través de tres repetidores. La probabilidad de que el primer repetidor transmita correctamente es 0.95. Si el primero funciona, el segundo transmite correctamente con probabilidad 0.92. Si los dos primeros funcionan, el tercero transmite correctamente con probabilidad 0.97. ¿Cuál es la probabilidad de que el mensaje llegue sin errores?

$$P(T_1 \cap T_2 \cap T_3) = 0.95 \times 0.92 \times 0.97 = 0.8478$$

**Ejemplo 4: Extracción sin reposición de tres objetos**

Una urna contiene 7 bolas blancas y 3 negras. Se extraen tres bolas sin reposición. ¿Cuál es la probabilidad de que las tres sean blancas?

$$P(B_1 \cap B_2 \cap B_3) = \frac{7}{10} \times \frac{6}{9} \times \frac{5}{8} = \frac{210}{720} = \frac{7}{24} \approx 0.292$$

**Ejemplo 5: Sistema de seguridad**

Un sistema de seguridad tiene tres capas: una valla (probabilidad de ser saltada: 0.05), una puerta con cerradura (probabilidad de ser forzada si se saltó la valla: 0.10), y una alarma (probabilidad de no activarse si se superaron las dos anteriores: 0.02). ¿Cuál es la probabilidad de que un intruso penetre completamente?

$$P(\text{penetración}) = 0.05 \times 0.10 \times 0.02 = 0.0001$$

La seguridad en capas reduce drásticamente la probabilidad de penetración.

## Independencia de eventos

Dos eventos $A$ y $B$ son **independientes** si la ocurrencia de uno no afecta la probabilidad del otro. Es decir:

$$P(A|B) = P(A)$$

Equivalentemente, usando la regla del producto:

$$P(A \cap B) = P(A) \cdot P(B)$$

Esta segunda forma es más fácil de verificar en la práctica porque no requiere calcular probabilidades condicionales.

La independencia es un concepto fundamental que simplifica enormemente los cálculos. Sin embargo, es importante no asumir independencia sin verificarla, ya que muchos errores en aplicaciones prácticas provienen de esta suposición incorrecta.

Intuitivamente, dos eventos son independientes si el conocimiento de que uno ocurrió no cambia nuestra estimación de la probabilidad del otro. En términos de espacios muestrales, la ocurrencia de $B$ no reduce el espacio muestral de manera que afecte la proporción de $A$.

## Ejemplos de independencia y dependencia

**Ejemplo 1: Lanzamientos de moneda (independencia)**

Lanzamos una moneda dos veces. Sea $A$ = "cara en el primer lanzamiento" y $B$ = "cara en el segundo lanzamiento". ¿Son independientes?

$$P(A) = 1/2, \quad P(B) = 1/2, \quad P(A \cap B) = 1/4$$

Como $1/4 = (1/2) \times (1/2)$, los eventos son independientes. Esto tiene sentido: el resultado del primer lanzamiento no influye en el segundo.

**Ejemplo 2: Control de calidad con reposición (independencia)**

En un lote de 100 productos, 10 son defectuosos. Seleccionamos dos con reposición (devolviendo el primero antes de seleccionar el segundo). Sea $A$ = "el primero es defectuoso" y $B$ = "el segundo es defectuoso". ¿Son independientes?

$$P(A) = 10/100 = 0.1$$
$$P(B|A) = 10/100 = 0.1$$

Como $P(B|A) = P(B)$, los eventos son independientes. La reposición asegura que la primera selección no afecte la segunda.

**Ejemplo 3: Control de calidad sin reposición (dependencia)**

En un lote de 100 productos, 10 son defectuosos. Seleccionamos dos sin reposición. Sea $A$ = "el primero es defectuoso" y $B$ = "el segundo es defectuoso". ¿Son independientes?

$$P(A) = 10/100 = 0.1$$
$$P(B|A) = 9/99 \approx 0.0909$$

Como $P(B|A) \neq P(B)$, los eventos no son independientes. La extracción sin reposición crea dependencia: si el primero fue defectuoso, hay menos productos defectuosos disponibles para el segundo.

**Ejemplo 4: Clima y paraguas (dependencia)**

Sea $A$ = "llueve hoy" y $B$ = "una persona lleva paraguas". Claramente $P(A|B) \neq P(A)$, ya que la gente tiende a llevar paraguas cuando llueve. Hay dependencia entre ambos eventos. Llevar paraguas es más probable cuando llueve, y saber que alguien lleva paraguas aumenta nuestra creencia de que está lloviendo.

**Ejemplo 5: Resultados de exámenes (dependencia)**

En una universidad, el 70% de los estudiantes que estudian más de 10 horas aprueban el examen, mientras que solo el 30% de los que estudian menos de 10 horas aprueban. Si $A$ = "estudiar más de 10 horas" y $B$ = "aprobar el examen", entonces $P(B|A) = 0.70$ mientras que $P(B)$ es un promedio ponderado que será menor. Por lo tanto, $A$ y $B$ no son independientes. El estudio influye en el resultado.

**Ejemplo 6: Lanzamiento de dos dados (independencia sorprendente)**

Lanzamos dos dados. Sea $A$ = "el primer dado muestra un número par" y $B$ = "la suma de los dos dados es 7". ¿Son independientes?

$P(A) = 1/2$

$P(B) = 6/36 = 1/6$ (pares (1,6), (2,5), (3,4), (4,3), (5,2), (6,1))

$P(A \cap B) = 3/36 = 1/12$ (pares con primer dado par: (2,5), (4,3), (6,1))

Como $P(A) \times P(B) = (1/2) \times (1/6) = 1/12 = P(A \cap B)$, los eventos son independientes. ¡Sorprendente pero cierto! El hecho de que el primer dado sea par no cambia la probabilidad de que la suma sea 7.

**Ejemplo 7: Estado civil y género (independencia en poblaciones grandes)**

En una población grande, sea $A$ = "ser mujer" y $B$ = "estar casado". Si la proporción de casados es similar entre hombres y mujeres, entonces $A$ y $B$ pueden ser independientes. Por ejemplo, si el 50% de la población son mujeres y el 40% está casado, y además el 40% de las mujeres está casado, entonces $P(B|A) = 0.40 = P(B)$, por lo que son independientes.

## Independencia de tres o más eventos

Para que $A$, $B$ y $C$ sean independientes, no basta con que sean independientes por pares. Se requiere además:

$$P(A \cap B \cap C) = P(A) \cdot P(B) \cdot P(C)$$

Y de manera similar para cualquier subconjunto de eventos. Esta condición adicional es necesaria porque la independencia por pares no garantiza que no haya relaciones más complejas entre los eventos.

La definición completa de independencia para $n$ eventos requiere que la probabilidad de la intersección de cualquier subconjunto de ellos sea igual al producto de sus probabilidades individuales. Esto es una condición fuerte que no siempre se cumple.

## Ejemplos de independencia de múltiples eventos

**Ejemplo 1: Caso engañoso con monedas**

Lanzamos dos monedas. Definimos:
- $A$: la primera muestra cara
- $B$: la segunda muestra cara
- $C$: ambas muestran el mismo resultado

Los eventos son independientes por pares, pero no son completamente independientes:

$$P(A \cap B \cap C) = P(\text{ambas son cara}) = 1/4$$

Mientras que:
$$P(A) \cdot P(B) \cdot P(C) = (1/2) \times (1/2) \times (1/2) = 1/8$$

Como $1/4 \neq 1/8$, los tres eventos no son completamente independientes. La explicación es que $C$ está completamente determinado por $A$ y $B$: si $A$ y $B$ ocurren, entonces $C$ necesariamente ocurre. Esto crea una dependencia que no se captura con las independencias por pares.

**Ejemplo 2: Tres lanzamientos de moneda**

Lanzamos tres monedas. Sean:
- $A$: la primera es cara
- $B$: la segunda es cara
- $C$: la tercera es cara

Estos tres eventos son completamente independientes porque cada lanzamiento es independiente de los demás.

$$P(A \cap B \cap C) = 1/8 = (1/2) \times (1/2) \times (1/2)$$

Y también se cumple para cualquier subconjunto: $P(A \cap B) = 1/4$, $P(A \cap C) = 1/4$, $P(B \cap C) = 1/4$.

**Ejemplo 3: Sistema con componentes redundantes**

Un sistema tiene tres componentes idénticos en paralelo (basta con que uno funcione para que el sistema funcione). Cada componente funciona con probabilidad 0.90, independientemente de los demás. ¿Cuál es la probabilidad de que el sistema falle?

El sistema falla solo si los tres componentes fallan. Como son independientes:
$$P(\text{fallo}) = P(F_1 \cap F_2 \cap F_3) = 0.10 \times 0.10 \times 0.10 = 0.001$$

La redundancia aumenta drásticamente la confiabilidad del sistema.

**Ejemplo 4: Cuatro lanzamientos de moneda**

Lanzamos cuatro monedas. Sean $A_1, A_2, A_3, A_4$ los eventos de que cada moneda muestre cara. Estos cuatro eventos son completamente independientes. La probabilidad de que todas sean caras es $(1/2)^4 = 1/16$, y esto es igual al producto de las probabilidades individuales.

**Ejemplo 5: Fallos en componentes independientes**

Una máquina tiene tres componentes críticos que pueden fallar independientemente con probabilidades 0.01, 0.02 y 0.03 respectivamente. ¿Cuál es la probabilidad de que los tres fallen simultáneamente?

$$P(F_1 \cap F_2 \cap F_3) = 0.01 \times 0.02 \times 0.03 = 0.000006$$

Esta probabilidad extremadamente baja justifica el diseño con componentes redundantes.

## Relación entre independencia y eventos mutuamente excluyentes

Es importante distinguir entre **independencia** y **mutuamente excluyentes**:

- **Mutuamente excluyentes**: $P(A \cap B) = 0$, los eventos no pueden ocurrir juntos.
- **Independientes**: $P(A \cap B) = P(A) \cdot P(B)$, la ocurrencia de uno no afecta al otro.

Si dos eventos son mutuamente excluyentes y ambos tienen probabilidad positiva, entonces **no pueden ser independientes** porque $P(A \cap B) = 0$ pero $P(A) \cdot P(B) > 0$.

Por ejemplo, en un lanzamiento de dado, $A$ = "sale 1" y $B$ = "sale 2" son mutuamente excluyentes pero no independientes. Saber que salió 1 cambia drásticamente la probabilidad de que salga 2 (de $1/6$ a $0$).

Recíprocamente, si dos eventos son independientes, no pueden ser mutuamente excluyentes a menos que uno de ellos tenga probabilidad cero.

**11.** Un sistema de seguridad tiene tres capas: una valla (probabilidad de ser saltada: 0.08), una puerta con cerradura (probabilidad de ser forzada si se saltó la valla: 0.12), y una alarma (probabilidad de no activarse si se superaron las dos anteriores: 0.03). Calcula:
a) La probabilidad de que un intruso penetre completamente
b) La probabilidad de que supere la valla pero sea detenido en la puerta
c) La probabilidad de que supere valla y puerta pero active la alarma
d) Si sabemos que se activó la alarma, ¿cuál es la probabilidad de que el intruso haya superado la valla?

**12.** Una urna contiene 10 bolas numeradas del 1 al 10. Se extraen dos bolas sin reposición. Define:
$A$: la primera bola es par
$B$: la segunda bola es par
$C$: la suma de las dos bolas es par

a) Calcula $P(A)$, $P(B)$ y $P(C)$
b) ¿Son $A$ y $B$ independientes?
c) ¿Son $A$ y $C$ independientes?
d) ¿Son $B$ y $C$ independientes?
e) ¿Son $A$, $B$ y $C$ completamente independientes?

**13.** Una prueba para detectar una enfermedad tiene las siguientes características:
- Sensibilidad ($P(\text{positivo}|\text{enfermo})$) = 0.97
- Especificidad ($P(\text{negativo}|\text{sano})$) = 0.92
La enfermedad afecta al 1.5% de la población.
a) Si una persona da positivo, ¿cuál es la probabilidad de que esté enferma?
b) Si una persona da negativo, ¿cuál es la probabilidad de que esté sana?
c) ¿Qué conclusiones puedes extraer sobre la utilidad de la prueba?

**14.** Un proceso de manufactura produce piezas en lotes de 20. La probabilidad de que una pieza sea defectuosa es 0.04. Se seleccionan 3 piezas sin reposición.
a) ¿Cuál es la probabilidad de que las tres sean defectuosas?
b) ¿Cuál es la probabilidad de que ninguna sea defectuosa?
c) ¿Cuál es la probabilidad de que exactamente una sea defectuosa?
d) Si la primera pieza seleccionada fue defectuosa, ¿cuál es la probabilidad de que las otras dos también lo sean?

**15.** En un juego de cartas, se reparten 5 cartas de una baraja española de 40 cartas sin reposición. Define:
$A$: la primera carta es de oros
$B$: la segunda carta es de oros
$C$: la tercera carta es de oros
a) Calcula $P(A)$, $P(B|A)$ y $P(C|A \cap B)$
b) Usando la regla del producto, calcula la probabilidad de que las tres primeras cartas sean de de oros
c) ¿Son $A$, $B$ y $C$ independientes? Justifica.

# Probabilidad total y Teorema de Bayes.

## Teorema de Probabilidad Total

A veces no conocemos directamente $P(A)$, pero conocemos su comportamiento bajo diferentes escenarios. Si los escenarios forman una partición del espacio muestral, podemos calcular $P(A)$ promediando las probabilidades condicionales.

### Partición del espacio muestral

Una partición es un conjunto de eventos $B_1, B_2, \dots, B_k$ que cumplen:
- Son mutuamente excluyentes: $B_i \cap B_j = \emptyset$ para $i \neq j$
- Cubren todo el espacio: $B_1 \cup B_2 \cup \cdots \cup B_k = \Omega$

### Fórmula

Para cualquier evento $A$:

$$P(A) = \sum_{i=1}^{k} P(A|B_i) \cdot P(B_i)$$

---

**Ejemplo 1: Control de calidad con dos máquinas**

Una fábrica tiene dos máquinas que producen el mismo componente. La máquina 1 produce el 60% de la producción total, con un 5% de defectuosos. La máquina 2 produce el 40% restante, con un 3% de defectuosos. Si tomamos un componente al azar de toda la producción, ¿cuál es la probabilidad de que sea defectuoso?

Definimos:
- $B_1$: el componente proviene de máquina 1 ($P(B_1)=0.6$)
- $B_2$: el componente proviene de máquina 2 ($P(B_2)=0.4$)
- $A$: el componente es defectuoso

$$P(A) = P(A|B_1)P(B_1) + P(A|B_2)P(B_2)$$
$$P(A) = 0.05 \times 0.6 + 0.03 \times 0.4 = 0.03 + 0.012 = 0.042$$

El 4.2% de la producción total es defectuosa.

---

**Ejemplo 2: Urnas con canicas**

Tenemos tres urnas. La urna 1 contiene 3 canicas rojas y 7 azules. La urna 2 contiene 6 rojas y 4 azules. La urna 3 contiene 8 rojas y 2 azules. Se selecciona una urna al azar (cada una con probabilidad $1/3$) y luego se extrae una canica. ¿Cuál es la probabilidad de que sea roja?

- $B_i$: seleccionar la urna $i$ ($P(B_i)=1/3$)
- $A$: la canica es roja

$$P(A) = P(A|B_1)P(B_1) + P(A|B_2)P(B_2) + P(A|B_3)P(B_3)$$
$$P(A) = \frac{3}{10}\cdot\frac{1}{3} + \frac{6}{10}\cdot\frac{1}{3} + \frac{8}{10}\cdot\frac{1}{3} = \frac{1}{10} + \frac{2}{10} + \frac{8}{30} = 0.1 + 0.2 + 0.2667 = 0.5667$$

La probabilidad de obtener una canica roja es del 56.67%.

---

**Ejemplo 3: Tres proveedores de componentes**

Una empresa recibe componentes de tres proveedores. El proveedor A suministra el 50% de los componentes con un 2% de defectuosos. El proveedor B suministra el 30% con un 4% de defectuosos. El proveedor C suministra el 20% con un 6% de defectuosos. Si se selecciona un componente al azar, ¿cuál es la probabilidad de que sea defectuoso?

- $B_1$: proveedor A ($P=0.5$), $B_2$: proveedor B ($P=0.3$), $B_3$: proveedor C ($P=0.2$)
- $A$: componente defectuoso

$$P(A) = 0.02(0.5) + 0.04(0.3) + 0.06(0.2) = 0.01 + 0.012 + 0.012 = 0.034$$

El 3.4% de los componentes son defectuosos.

## Aplicación de Probabilidad Total: Diagnóstico Médico

Una prueba para una enfermedad tiene 99% de sensibilidad y 98% de especificidad. La prevalencia de la enfermedad en la población es 0.1%. ¿Cuál es la probabilidad de que una persona seleccionada al azar dé positivo?

**Solución:**

Definimos:
- $E$: la persona tiene la enfermedad ($P(E) = 0.001$)
- $E^c$: la persona no tiene la enfermedad ($P(E^c) = 0.999$)
- $+$: la prueba da positivo

Sensibilidad: $P(+|E) = 0.99$
Especificidad: $P(-|E^c) = 0.98$, por lo que $P(+|E^c) = 0.02$

Por probabilidad total:

$$P(+) = P(+|E)P(E) + P(+|E^c)P(E^c)$$
$$P(+) = 0.99(0.001) + 0.02(0.999) = 0.00099 + 0.01998 = 0.02097$$

La probabilidad de que una persona seleccionada al azar dé positivo es del **2.097%**.

---

**Ejemplo adicional 1: Prueba de COVID-19 con alta prevalencia**

Supongamos que una prueba rápida para COVID-19 tiene 90% de sensibilidad y 95% de especificidad. En una ciudad con prevalencia del 15%, ¿cuál es la probabilidad de que una persona seleccionada al azar dé positivo?

$$P(+) = 0.90(0.15) + 0.05(0.85) = 0.135 + 0.0425 = 0.1775$$

La probabilidad es del **17.75%**.

---

**Ejemplo adicional 2: Prueba de dopaje en atletas**

Una prueba antidopaje tiene 98% de sensibilidad y 99% de especificidad. Se estima que el 2% de los atletas profesionales usan sustancias prohibidas. Si se selecciona un atleta al azar para la prueba, ¿cuál es la probabilidad de que dé positivo?

$$P(+) = 0.98(0.02) + 0.01(0.98) = 0.0196 + 0.0098 = 0.0294$$

La probabilidad es del **2.94%**.

## Teorema de Bayes 

Podemos calcular la probabilidad de un suceso $A$, sabiendo además que ese $A$ cumple cierta característica que condiciona su probabilidad. 

El Teorema de la Probabilidad total hace inferencia sobre un suceso $A$, a partir de los resultados de los sucesos $E_i$ siempre que estos formen una partición del espacio muestral. Por su parte, Bayes calcula la probabilidad de cada $E_i$ condicionado a $B$.

### Teorema de Bayes (forma general)

Sean $E_1,E_2,...,E_k$ eventos disjuntos dos a dos cuya unión es todo $S$. Si ocurre un evento $A$, se cumple: 

$$P(E_i|A)=\frac{P(E_i)P(A|E_i)}{P(E_1)P(A|E_1)+P(E_2)P(A|E_2)+...+P(E_k)P(A|E_k)}$$

### Teorema de Bayes (forma para dos eventos)

Si $A$ y $B$ son eventos, entonces:

$$P(B|A)=\frac{P(B)}{P(A)}P(A|B)$$

---

**Ejemplo 1: SciData - Helado y pastel**

En SciData, la probabilidad de que a un alumno seleccionado al azar le guste el helado es del 60%, mientras que la probabilidad de que a un alumno le guste el pastel es del 36%. Además, se sabe que la probabilidad de que a un alumno le guste el pastel dado que le gusta el helado es del 40%. Calcular la probabilidad de que a un alumno le guste el helado, dado que le gusta el pastel.

**Solución:**

Definimos:
- $H$: le gusta el helado ($P(H) = 0.60$)
- $P$: le gusta el pastel ($P(P) = 0.36$)
- $P(P|H) = 0.40$

Aplicando Bayes:

$$P(H|P) = \frac{P(H)P(P|H)}{P(P)} = \frac{0.60 \times 0.40}{0.36} = \frac{0.24}{0.36} = 0.6667$$

La probabilidad es del **66.67%**.

---

**Ejemplo 2: Urnas invertido**

Usando el ejemplo de las urnas con canicas (urna 1: 3R,7A; urna 2: 6R,4A; urna 3: 8R,2A). Supongamos que se selecciona una urna al azar, se extrae una canica y resulta ser roja. ¿Cuál es la probabilidad de que provenga de la urna 3?

$$P(B_3|R) = \frac{P(B_3)P(R|B_3)}{P(R)} = \frac{(1/3)(8/10)}{17/30} = \frac{8/30}{17/30} = \frac{8}{17} \approx 0.4706$$

La probabilidad es del **47.06%**.

---

**Ejemplo 3: Proveedores invertido**

Del ejemplo de los tres proveedores (A: 50%, 2% defectuosos; B: 30%, 4%; C: 20%, 6%), si encontramos un componente defectuoso, ¿cuál es la probabilidad de que provenga del proveedor A?

$$P(B_1|D) = \frac{P(B_1)P(D|B_1)}{P(D)} = \frac{0.5 \times 0.02}{0.034} = \frac{0.01}{0.034} \approx 0.2941$$

La probabilidad es del **29.41%**.

## El ejemplo médico completo (Bayes)

Retomemos el ejemplo de la prueba médica. Una prueba tiene 99% de sensibilidad y 98% de especificidad. La prevalencia es 0.1%. Una persona da positivo. ¿Cuál es la probabilidad de que realmente tenga la enfermedad?

$$P(E|+) = \frac{P(+|E)P(E)}{P(+|E)P(E) + P(+|E^c)P(E^c)}$$

$$P(E|+) = \frac{0.99 \times 0.001}{0.99 \times 0.001 + 0.02 \times 0.999}$$

$$P(E|+) = \frac{0.00099}{0.00099 + 0.01998} = \frac{0.00099}{0.02097} \approx 0.0472$$

Sorprendentemente, incluso con una prueba muy precisa, una persona que da positivo tiene solo un **4.72%** de probabilidad de estar realmente enferma. Esto se debe a la baja prevalencia de la enfermedad.

---

**Ejemplo adicional 1: Prueba de COVID con prevalencia alta**

Usando la prueba de COVID con 90% sensibilidad y 95% especificidad, y prevalencia del 15%. Si una persona da positivo, ¿cuál es la probabilidad de que realmente esté infectada?

$$P(E|+) = \frac{0.90 \times 0.15}{0.90 \times 0.15 + 0.05 \times 0.85} = \frac{0.135}{0.135 + 0.0425} = \frac{0.135}{0.1775} \approx 0.7606$$

La probabilidad es del **76.06%**.

---

**Ejemplo adicional 2: Prueba de dopaje**

Para la prueba antidopaje (98% sensibilidad, 99% especificidad, 2% de usuarios). Si un atleta da positivo, ¿cuál es la probabilidad de que realmente use dopaje?

$$P(E|+) = \frac{0.98 \times 0.02}{0.98 \times 0.02 + 0.01 \times 0.98} = \frac{0.0196}{0.0196 + 0.0098} = \frac{0.0196}{0.0294} \approx 0.6667$$

La probabilidad es del **66.67%**.

## Actualización secuencial de creencias

Lo hermoso de Bayes es que podemos actualizar nuestra creencia repetidamente. La probabilidad a posteriori de un paso se convierte en la probabilidad a priori del siguiente.

**Ejemplo 1: Dos pruebas médicas (prueba del 99%/98% con prevalencia 0.1%)**

La misma persona del ejemplo anterior da positivo en una segunda prueba independiente (con las mismas características). ¿Cuál es ahora su probabilidad de estar enferma?

Ahora nuestra a priori es $P(E)=0.0472$ (el resultado del paso anterior). Aplicamos Bayes nuevamente:

$$P(E|+_2)=\frac{0.99\times 0.0472}{0.99\times0.0472+0.02\times(1-0.0472)} = \frac{0.046728}{0.046728 + 0.019056} = \frac{0.046728}{0.065784} \approx 0.7103$$

Con dos pruebas positivas, la probabilidad sube aproximadamente al **71.03%**. Esto muestra cómo la evidencia acumulada fortalece nuestra confianza.

---

**Ejemplo 2: Tres pruebas positivas de COVID (prevalencia 15%)**

Usando la prueba de COVID (90% sensibilidad, 95% especificidad) con prevalencia inicial del 15%. Después de una prueba positiva, la posterior es $P(E)=0.7606$.

Segunda prueba positiva:
$$P(E|+_2)=\frac{0.90\times0.7606}{0.90\times0.7606 + 0.05\times0.2394} = \frac{0.68454}{0.68454 + 0.01197} = \frac{0.68454}{0.69651} \approx 0.9828$$

Tercera prueba positiva:
$$P(E|+_3)=\frac{0.90\times0.9828}{0.90\times0.9828 + 0.05\times0.0172} = \frac{0.88452}{0.88452 + 0.00086} = \frac{0.88452}{0.88538} \approx 0.9990$$

Con tres pruebas positivas, la probabilidad es del **99.90%**.

---

**Ejemplo 3: Pruebas de dopaje secuenciales**

Para el dopaje (98% sensibilidad, 99% especificidad, 2% prevalencia). Después de una positiva: $P(E)=0.6667$.

Segunda prueba positiva:
$$P(E|+_2)=\frac{0.98\times0.6667}{0.98\times0.6667 + 0.01\times0.3333} = \frac{0.65337}{0.65337 + 0.00333} = \frac{0.65337}{0.65670} \approx 0.9949$$

Con dos pruebas positivas, la probabilidad es del **99.49%**.

**Solución:** 0.062

---

**2.** Tres máquinas producen tornillos. La máquina A produce el 25% con 2% defectuosos, la B produce el 35% con 3% defectuosos, y la C produce el 40% con 4% defectuosos. ¿Cuál es la probabilidad de que un tornillo seleccionado al azar sea defectuoso?

**Solución:** 0.0315

**Solución:** 0.100

**4.** Una empresa tiene tres sucursales. La sucursal 1 vende el 40% de los productos con 6% de devoluciones, la sucursal 2 vende el 35% con 4% de devoluciones, y la sucursal 3 vende el 25% con 3% de devoluciones. ¿Cuál es la probabilidad de que un producto seleccionado al azar sea devuelto?

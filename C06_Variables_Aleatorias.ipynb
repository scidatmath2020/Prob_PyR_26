# Variables Aleatorias

Hasta ahora, hemos trabajado con espacios muestrales donde los resultados podían ser: caras o cruces, números en un dado, personas enfermas o sanas, productos defectuosos o no. Las variables aleatorias una manera de **convertir los resultados de un experimento en números**.

Piensa en un experimento sencillo: lanzar una moneda tres veces. El espacio muestral es:

$$\Omega = \{CCC, CCS, CSC, SCC, CSS, SCS, SSC, SSS\}$$

Si queremos calcular probabilidades, podemos trabajar directamente con este espacio. Pero si lo que nos interesa es, por ejemplo, el **número de caras**, no necesitamos listar todos los resultados; solo necesitamos saber que el número de caras puede ser 0, 1, 2 o 3. Hemos pasado de un espacio muestral con 8 elementos a una variable que solo toma 4 valores. Esa es la utilidad de las variables aleatorias: **resumir la información relevante** de un experimento.

Una **variable aleatoria** es una función que asigna un número real a cada resultado de un experimento aleatorio:

$$X: \Omega \rightarrow \mathbb{R}$$

Por ejemplo, en el lanzamiento de tres monedas, podríamos definir:

$$X = \text{número de caras obtenidas}$$

Entonces:
- $X(CCC) = 3$
- $X(CCS) = 2$
- $X(CSS) = 1$
- $X(SSS) = 0$

La variable aleatoria es el puente entre la probabilidad y la estadística. Sin variables aleatorias:

- No podríamos calcular el **promedio** de una muestra, porque los datos no son números.
- No podríamos medir la **variabilidad** de un fenómeno.
- No podríamos construir modelos matemáticos para fenómenos aleatorios.

Con variables aleatorias, podemos empezar a responder preguntas como:

- En promedio, ¿cuántas caras obtengo al lanzar tres monedas?
- ¿Qué tan variable es el número de caras?
- ¿Cuál es la probabilidad de obtener al menos dos caras?

### Notación

Usaremos mayúsculas para la variable aleatoria: $X, Y, Z$

Usaremos minúsculas para los valores específicos que puede tomar: $x, y, z$

Así, cuando escribimos $P(X = 2)$, estamos preguntando: "¿cuál es la probabilidad de que la variable aleatoria $X$ tome exactamente el valor 2?"

**Ejemplo 1:** Seleccionamos una persona al azar y medimos su estatura.

Podemos definir:
$$X = \text{estatura en centímetros}$$

$X$ puede tomar cualquier valor entre aproximadamente 50 cm y 250 cm.

**Ejemplo 2:** Registramos el tiempo que tarda en llegar un autobús.

Podemos definir:
$$X = \text{tiempo de espera en minutos}$$

$X$ puede tomar cualquier valor mayor o igual a 0.

**Ejemplo 3:** Contamos el número de clientes que entran a una tienda en una hora.

Podemos definir:
$$X = \text{número de clientes}$$

$X$ puede tomar los valores 0, 1, 2, 3, ...

## Función de Distribución

Ahora que tenemos una variable aleatoria $X$, necesitamos una manera de **describir su comportamiento probabilístico**. Es decir, necesitamos saber qué tan probable es que $X$ tome ciertos valores.

La herramienta más poderosa para esto es la **función de distribución acumulada**, que se define simplemente como:

$$F(x) = P(X \leq x)$$

Esta función nos dice, para cualquier número $x$, cuál es la probabilidad de que la variable aleatoria tome un valor **menor o igual** a $x$.

La función de distribución **funciona para cualquier variable aleatoria**, sin importar si toma valores aislados o valores en un continuo. Es el concepto unificador que nos permite tratar todos los casos con la misma herramienta.

**Ejemplo.** Volvamos al lanzamiento de dos monedas. Sea $X$ el número de caras. Sabemos que:

$$P(X = 0) = 1/4, \quad P(X = 1) = 1/2, \quad P(X = 2) = 1/4$$

Calculemos la función de distribución paso a paso:

- Para cualquier $x < 0$: no hay valores menores o iguales a $x$, así que $F(x) = 0$.
- Para $0 \leq x < 1$: los valores posibles menores o iguales a $x$ son solo el 0, así que $F(x) = P(X=0) = 1/4$.
- Para $1 \leq x < 2$: los valores posibles son 0 y 1, así que $F(x) = P(X=0) + P(X=1) = 1/4 + 1/2 = 3/4$.
- Para $x \geq 2$: todos los valores posibles son menores o iguales a $x$, así que $F(x) = 1$.

Podemos escribir esto de manera compacta:

$$F(x) = \begin{cases}
0, & x < 0 \\
1/4, & 0 \leq x < 1 \\
3/4, & 1 \leq x < 2 \\
1, & x \geq 2
\end{cases}$$

La función de distribución nos permite calcular cualquier probabilidad relacionada con $X$:

- $P(X \leq 1) = F(1) = 3/4$
- $P(X > 1) = 1 - F(1) = 1 - 3/4 = 1/4$
- $P(0 < X \leq 1) = F(1) - F(0) = 3/4 - 1/4 = 1/2$

### Propiedades que siempre se cumplen

La función de distribución siempre tiene estas propiedades:

1. **Es no decreciente:** si $a < b$, entonces $F(a) \leq F(b)$. Esto tiene sentido: si $a$ es más pequeño que $b$, la probabilidad de ser menor o igual a $a$ no puede ser mayor que la de ser menor o igual a $b$.

2. **Tiende a 0 en $-\infty$ y a 1 en $+\infty$:**
   $$\lim_{x \to -\infty} F(x) = 0 \quad \text{y} \quad \lim_{x \to \infty} F(x) = 1$$
   Esto también tiene sentido: si $x$ es muy pequeño, es casi imposible que $X$ sea menor o igual a $x$; si $x$ es muy grande, es casi seguro que $X$ sea menor o igual a $x$.

3. **Siempre está entre 0 y 1:** $0 \leq F(x) \leq 1$ para todo $x$.

4. **Siempre es continua por la derecha**

**Ejemplo.** Imagina que tienes un reloj analógico (de manecillas) que está roto: la manecilla de los segundos gira continuamente, pero tú no sabes en qué posición está cuando lo miras. La posición de la manecilla puede estar en cualquier punto del círculo, y todos los puntos son igualmente probables.

Pensemos en el tiempo transcurrido desde las 12:00 (medido en segundos). Como el reloj da una vuelta completa cada 60 segundos, sea $X$ "número de segundos que han pasado desde las 12:00 cuando miras el reloj". 

Así:
- $X$ puede tomar cualquier valor entre 0 y 60 segundos.
- No hay ninguna posición más probable que otra.

**¿Cuál es la probabilidad de que la manecilla esté entre 0 y 30 segundos?** Hay 30 segundos de longitud favorable sobre 60 totales así que Probabilidad = 30/60 = 0.5

**¿Probabilidad de que esté entre 10 y 25 segundos?** Longitud = 15 segundos así que Probabilidad = 15/60 = 0.25

**¿Probabilidad de que esté entre 0 y x segundos, donde x está entre 0 y 60?** Probabilidad = $x/60$

Así, en general $F(x)=P(X\le x)=\left\{\begin{array}{l}0\mbox{ si }x<0\\\frac{x}{60}\mbox{ si }0\le x<60\\1\mbox{ si }60\le x\end{array}\right.$

## Dos tipos de variables aleatorias

Si observamos con atención los dos ejemplos anteriores, notamos una diferencia fundamental:

- En el lanzamiento de monedas, $X$ solo puede tomar valores aislados: 0, 1, 2. La función de distribución tiene **saltos** en esos valores.
- En el tiempo de espera, $X$ puede tomar **cualquier valor** entre 0 y 10. La función de distribución es **continua**, sin saltos.

Esta diferencia es tan importante que clasificamos las variables aleatorias en dos grandes familias.

---

### Variables aleatorias discretas

Una variable aleatoria es **discreta** cuando solo puede tomar un número finito o infinito numerable de valores. Estos valores están "separados" entre sí.

**Características:**

- Los valores posibles se pueden listar: $x_1, x_2, x_3, \ldots$
- La función de distribución es **escalonada** (tiene saltos).
- Tiene sentido hablar de $P(X = x)$ para un valor específico, y esta probabilidad puede ser positiva.

**Ejemplos cotidianos:**

- El número de caras al lanzar monedas.
- El número de clientes que entran a una tienda en una hora.
- El número de accidentes en una carretera en un mes.
- El número de hijos en una familia.
- La calificación obtenida en un examen (si solo puede ser 0, 1, 2, ..., 10).

### Variables aleatorias continuas

Una variable aleatoria es **continua** cuando puede tomar cualquier valor dentro de un intervalo de la recta real. Los valores posibles forman un continuo.

**Características:**

- Los valores posibles no se pueden listar (son infinitos e incontables).
- La función de distribución es **continua** (no tiene saltos).
- $P(X = x) = 0$ para cualquier valor específico $x$. Solo tiene sentido hablar de probabilidades en intervalos.

**Ejemplos cotidianos:**

- La estatura de una persona.
- El tiempo de espera en una fila.
- La temperatura ambiente.
- El peso de un producto.
- La duración de una batería.

## Función de Probabilidad: variables discretas

Cuando trabajamos con una variable discreta, la función de distribución nos da la probabilidad acumulada, pero a veces queremos saber la probabilidad de un valor específico. Para eso necesitamos la **función de probabilidad**.

### Definición

La función de probabilidad $f(x)$ se define como:

$$f(x) = P(X = x)$$

Es decir, $f(x)$ nos dice la probabilidad de que la variable tome exactamente el valor $x$.

### Propiedades

La función de probabilidad siempre cumple dos propiedades:

1. $0 \leq f(x) \leq 1$ para todo $x$ (son probabilidades, después de todo).
2. $\sum_{x} f(x) = 1$ (la suma de todas las probabilidades debe ser 1).

### Relación con la función de distribución

La función de distribución y la función de probabilidad están íntimamente relacionadas:

- **De la probabilidad a la distribución:** $F(x) = \sum_{t \leq x} f(t)$
- **De la distribución a la probabilidad:** $f(x) = F(x) - F(x^-)$

Donde $F(x^-)$ es el límite de $F$ por la izquierda en $x$ (el valor justo antes del salto).

**Ejemplo** Se lanzan dos monedas. Para $X =$ número de caras:

$$f(0) = 1/4, \quad f(1) = 1/2, \quad f(2) = 1/4$$

Verificamos que $\sum f(x) = 1/4 + 1/2 + 1/4 = 1$ 

La función de distribución se puede reconstruir a partir de la función de probabilidad:

$$F(0) = f(0) = 1/4$$
$$F(1) = f(0) + f(1) = 1/4 + 1/2 = 3/4$$
$$F(2) = f(0) + f(1) + p(2) = 1$$

### Ejemplo: lanzamiento de tres monedas

Para $X =$ número de caras en tres lanzamientos:

| $x$ | $f(x)$ |
|-----|--------|
| 0 | 1/8 |
| 1 | 3/8 |
| 2 | 3/8 |
| 3 | 1/8 |

Verificamos: $1/8 + 3/8 + 3/8 + 1/8 = 1$

La función de distribución sería:

$$F(x) = \begin{cases}
0, & x < 0 \\
1/8, & 0 \leq x < 1 \\
4/8, & 1 \leq x < 2 \\
7/8, & 2 \leq x < 3 \\
1, & x \geq 3
\end{cases}$$

**Ejemplo.**

Imagina que registramos el número de llamadas que recibe un call center por hora durante muchos días, y obtenemos la siguiente distribución:

$$f(0) = 0.1, \quad f(1) = 0.2, \quad f(2) = 0.3, \quad f(3) = 0.25, \quad f(4) = 0.15$$

Podemos calcular, por ejemplo:
- $P(X \leq 2) = f(0) + f(1) + f(2) = 0.1 + 0.2 + 0.3 = 0.6$
- $P(X > 2) = f(3) + f(4) = 0.25 + 0.15 = 0.4$

Y la función de distribución sería:
$$F(0) = 0.1, \quad F(1) = 0.3, \quad F(2) = 0.6, \quad F(3) = 0.85, \quad F(4) = 1$$

## Función de Densidad: variables continuas

Para variables continuas, la función de distribución es continua, pero no tiene "saltos" que nos permitan aislar probabilidades puntuales. De hecho, para una variable continua:

$$P(X = x) = 0 \quad \text{para cualquier valor específico } x$$

Esto puede sonar extraño al principio: ¿cómo puede ser que la probabilidad de un valor específico sea cero? Piensa en la estatura de una persona: la probabilidad de que alguien mida exactamente 175.000000... cm (con infinitos decimales) es prácticamente cero. Lo que tiene sentido es preguntar por la probabilidad de que la estatura esté entre 174 y 176 cm.

Para variables continuas, necesitamos una herramienta diferente: la **función de densidad**.

### Definición

La función de densidad $f(x)$ se define como la derivada de la función de distribución:

$$f(x) = \frac{d}{dx} F(x)$$

O equivalentemente, la función de distribución es la integral de la densidad:

$$F(x) = \int_{-\infty}^{x} f(t) dt$$

### Propiedades

1. $f(x) \geq 0$ para todo $x$ (la densidad nunca es negativa).
2. $\int_{-\infty}^{\infty} f(x) dx = 1$ (el área total bajo la curva es 1).

La probabilidad de que $X$ esté entre $a$ y $b$ es el área bajo la curva en ese intervalo:

$$P(a \leq X \leq b) = \int_a^b f(x) dx = F(b) - F(a)$$

### Ejemplo: distribución uniforme

La distribución más sencilla es la **uniforme**: todos los valores en un intervalo son igualmente probables.

Para $X \sim \text{Uniforme}(0,10)$:

$$f(x) = \begin{cases}
1/10, & 0 \leq x \leq 10 \\
0, & \text{en otro caso}
\end{cases}$$

Verificamos: $\int_0^{10} (1/10) dx = 1$ 

La función de distribución es:

$$F(x) = \begin{cases}
0, & x < 0 \\
x/10, & 0 \leq x \leq 10 \\
1, & x > 10
\end{cases}$$

Podemos calcular, por ejemplo:
$$P(2 \leq X \leq 5) = \int_2^5 \frac{1}{10} dx = \frac{3}{10} = 0.3$$

### Ejemplo: distribución exponencial

La distribución exponencial se usa para modelar tiempos de espera o tiempos de vida.

$$f(x) = \begin{cases}
\lambda e^{-\lambda x}, & x \geq 0 \\
0, & x < 0
\end{cases}$$

Para $\lambda = 0.5$ (tiempo promedio de vida de 2 unidades):

$$F(x) = \begin{cases}
0, & x < 0 \\
1 - e^{-0.5x}, & x \geq 0
\end{cases}$$

Podemos calcular:
$$P(1 \leq X \leq 3) = F(3) - F(1) = (1 - e^{-1.5}) - (1 - e^{-0.5}) = e^{-0.5} - e^{-1.5} \approx 0.3834$$

### Comparación entre discreta y continua

| Aspecto | Discreta | Continua |
|---------|----------|----------|
| Función | $p(x) = P(X=x)$ | $f(x) = F'(x)$ |
| Rango | $0 \leq p(x) \leq 1$ | $f(x) \geq 0$ (puede ser >1) |
| Normalización | $\sum p(x) = 1$ | $\int f(x) dx = 1$ |
| Probabilidad puntual | $P(X=x) = p(x)$ | $P(X=x) = 0$ |
| Probabilidad en intervalo | $\sum_{x=a}^b p(x)$ | $\int_a^b f(x) dx$ |

## Esperanza Matemática

Ahora que sabemos cómo describir una variable aleatoria, queremos **resumir** su comportamiento con unos pocos números clave. El más importante es la **esperanza matemática** (también llamada **valor esperado** o **media**).

### ¿Qué es la esperanza?

La esperanza es el **promedio ponderado** de todos los valores posibles de una variable aleatoria, donde los pesos son las probabilidades. Es el centro de masa de la distribución.

Imagina que repites un experimento muchas veces y calculas el promedio de los resultados obtenidos. La esperanza es el valor al que se acerca ese promedio cuando el número de repeticiones tiende a infinito.

### Definición formal

**Para variables discretas:**

$$E[X] = \sum_{x} x \cdot f(x)$$

Multiplicamos cada valor posible por su probabilidad y sumamos.

**Para variables continuas:**

$$E[X] = \int_{-\infty}^{\infty} x \cdot f(x) dx$$

Multiplicamos cada valor posible por su densidad y sumamos (integrando).

### Ejemplo: lanzamiento de dos monedas

Para $X =$ número de caras:

$$E[X] = 0\cdot\frac{1}{4} + 1\cdot\frac{1}{2} + 2\cdot\frac{1}{4} = 0 + 0.5 + 0.5 = 1$$

Interpretación: si lanzamos dos monedas muchas veces, en promedio obtendremos 1 cara.

### Ejemplo: juego de dados (una situación con ganancias y pérdidas)

Supongamos que jugamos el siguiente juego: lanzamos un dado. Si sale 6, ganamos \$10. Si sale 5, ganamos \$5. Si sale cualquier otro número, perdemos \$2.

¿Cuál es nuestra ganancia esperada?

Definimos $X =$ ganancia en el juego. La función de probabilidad es:

$$p(10) = 1/6, \quad p(5) = 1/6, \quad p(-2) = 4/6$$

Entonces:

$$E[X] = 10\cdot\frac{1}{6} + 5\cdot\frac{1}{6} + (-2)\cdot\frac{4}{6} = \frac{10 + 5 - 8}{6} = \frac{7}{6} \approx 1.17$$

Interpretación: en promedio, ganamos aproximadamente \$1.17 por juego. Esto significa que si jugamos muchas veces, es probable que terminemos ganando dinero.

### Ejemplo: tiempo de espera uniforme

Para $X \sim \text{Uniforme}(0,10)$:

$$E[X] = \int_0^{10} x\cdot\frac{1}{10} dx = \frac{1}{10}\cdot\frac{x^2}{2}\Big|_0^{10} = \frac{1}{10}\cdot\frac{100}{2} = 5$$

Interpretación: el tiempo promedio de espera es 5 minutos.

### Ejemplo: distribución exponencial

Para $X \sim \text{Exponencial}(\lambda)$:

$$E[X] = \int_0^{\infty} x\lambda e^{-\lambda x} dx = \frac{1}{\lambda}$$

Si $\lambda = 2$, entonces $E[X] = 0.5$. Interpretación: el tiempo de vida promedio es 0.5 unidades.

### Una advertencia importante

**La esperanza no siempre es un valor posible de la variable.** En el ejemplo del dado, la esperanza es \$1.17, pero eso no es un resultado posible del juego (solo podemos ganar 10, ganar 5 o perder 2). La esperanza es un valor teórico que resume el comportamiento promedio.

## Esperanza de una función de variable aleatoria

A menudo necesitamos calcular la esperanza de una función de $X$, como $X^2$, $\sqrt{X}$ o $e^X$. La fórmula es similar:

**Para variables discretas:**
$$E[g(X)] = \sum_{x} g(x) \cdot p(x)$$

**Para variables continuas:**
$$E[g(X)] = \int_{-\infty}^{\infty} g(x) \cdot f(x) dx$$

### Ejemplo: $E[X^2]$ para dos monedas

$$E[X^2] = 0^2\cdot\frac{1}{4} + 1^2\cdot\frac{1}{2} + 2^2\cdot\frac{1}{4} = 0 + 0.5 + 1 = 1.5$$

### Propiedades de la esperanza (muy útiles)

1. **Linealidad:** $E[aX + b] = aE[X] + b$

   Esto significa que si multiplicamos la variable por una constante y sumamos otra constante, la esperanza se transforma de la misma manera.

2. **Suma:** $E[X + Y] = E[X] + E[Y]$

   Esto siempre se cumple, incluso si $X$ y $Y$ no son independientes.

3. **Producto:** Si $X$ y $Y$ son independientes, $E[XY] = E[X]E[Y]$

4. **Constante:** $E[c] = c$

### Ejemplo de linealidad

Si sabemos que $E[X] = 10$, podemos calcular sin necesidad de la distribución:

$$E[3X + 5] = 3E[X] + 5 = 3(10) + 5 = 35$$

## Varianza y Desviación Estándar.

La esperanza nos dice dónde está el centro de la distribución, pero no nos dice **qué tan dispersos** están los valores alrededor de ese centro. Dos variables pueden tener la misma media pero comportamientos muy diferentes.

Imagina dos procesos de fabricación:

- El proceso A produce piezas con peso promedio de 100 gramos, con variaciones muy pequeñas (entre 99 y 101 gramos).
- El proceso B también produce piezas con peso promedio de 100 gramos, pero con variaciones grandes (entre 80 y 120 gramos).

Ambos tienen la misma media, pero el proceso B es mucho menos confiable. Necesitamos una medida de esta dispersión.

### La varianza

La **varianza** mide la dispersión promedio al cuadrado de los valores alrededor de la media:

$$\text{Var}(X) = E[(X - E[X])^2]$$

Es decir, calculamos la diferencia entre cada valor y la media, elevamos al cuadrado (para que las diferencias negativas no se cancelen con las positivas), y calculamos el promedio de esos cuadrados.

### Fórmula alternativa (más práctica)

$$\text{Var}(X) = E[X^2] - (E[X])^2$$

Esta fórmula es más fácil de usar en cálculos, porque solo necesitamos calcular $E[X]$ y $E[X^2]$.

### Definiciones según el tipo de variable

**Para variables discretas:**
$$\text{Var}(X) = \sum_{x} (x - \mu)^2 p(x) = \sum_{x} x^2 p(x) - \mu^2$$

**Para variables continuas:**
$$\text{Var}(X) = \int_{-\infty}^{\infty} (x - \mu)^2 f(x) dx = \int_{-\infty}^{\infty} x^2 f(x) dx - \mu^2$$

Donde $\mu = E[X]$.

### Desviación estándar

La varianza tiene un problema: está en unidades al cuadrado. Si $X$ mide centímetros, la varianza mide centímetros cuadrados. Para tener una medida en las mismas unidades que la variable original, definimos la **desviación estándar**:

$$\sigma = \sqrt{\text{Var}(X)}$$

### Propiedades de la varianza

1. $\text{Var}(X) \geq 0$ (siempre es no negativa)
2. $\text{Var}(aX + b) = a^2 \text{Var}(X)$
3. Si $X$ y $Y$ son independientes: $\text{Var}(X + Y) = \text{Var}(X) + \text{Var}(Y)$
4. $\text{Var}(c) = 0$ (una constante no varía)

### Ejemplo 1: lanzamiento de dos monedas

Ya calculamos $\mu = E[X] = 1$ y $E[X^2] = 1.5$.

$$\text{Var}(X) = 1.5 - 1^2 = 0.5$$
$$\sigma = \sqrt{0.5} \approx 0.7071$$

Interpretación: el número de caras tiene una desviación estándar de aproximadamente 0.707 caras.

### Ejemplo 2: tiempo de espera uniforme

Para $X \sim \text{Uniforme}(0,10)$:

$$\mu = 5$$
$$E[X^2] = \int_0^{10} x^2\cdot\frac{1}{10} dx = \frac{1000}{30} = 33.333$$
$$\text{Var}(X) = 33.333 - 25 = 8.333$$
$$\sigma = \sqrt{8.333} \approx 2.887$$

Interpretación: el tiempo de espera tiene una desviación estándar de aproximadamente 2.89 minutos.

### Ejemplo 3: inversión en acciones (aplicación financiera)

Un inversionista está evaluando una acción con los siguientes retornos anuales posibles:

| Probabilidad | Retorno |
|--------------|---------|
| 20% | 15% |
| 40% | 8% |
| 30% | 2% |
| 10% | -5% |

Primero calculamos el retorno esperado:

$$\mu = 0.15(0.2) + 0.08(0.4) + 0.02(0.3) + (-0.05)(0.1) = 0.03 + 0.032 + 0.006 - 0.005 = 0.063 = 6.3\%$$

Ahora calculamos $E[X^2]$:

$$E[X^2] = 0.15^2(0.2) + 0.08^2(0.4) + 0.02^2(0.3) + (-0.05)^2(0.1)$$
$$E[X^2] = 0.0225(0.2) + 0.0064(0.4) + 0.0004(0.3) + 0.0025(0.1)$$
$$E[X^2] = 0.0045 + 0.00256 + 0.00012 + 0.00025 = 0.00743$$

Finalmente:
$$\text{Var}(X) = 0.00743 - 0.063^2 = 0.00743 - 0.003969 = 0.003461$$
$$\sigma = \sqrt{0.003461} \approx 0.0588 = 5.88\%$$

Interpretación: la acción tiene un retorno esperado de 6.3% con una desviación estándar de 5.88%. Esto permite al inversionista evaluar el riesgo: a mayor desviación estándar, mayor riesgo.

### Uniendo todo: la importancia de la varianza

La varianza (o su raíz cuadrada, la desviación estándar) es fundamental en probabilidad y estadística porque:

1. Nos permite cuantificar el **riesgo** en finanzas.
2. Nos ayuda a entender la **precisión** de mediciones.
3. Es la base de muchos teoremas importantes (como el Teorema Central del Límite).
4. Nos permite comparar la variabilidad de diferentes fenómenos.

En el contexto de las distribuciones discretas (especialmente Bernoulli, Binomial y Geométrica), la palabra *éxito* **NO significa necesariamente algo bueno, positivo o deseable:**

*Éxito =* el evento que estamos contando o que nos interesa medir. Por ejemplo:
- En control de calidad: *éxito* puede ser encontrar un producto defectuoso.
- En marketing: *éxito* = el cliente compra el producto.
- En medicina: *éxito* = el paciente responde al tratamiento.
- En medicina: *éxito* = el paciente está enfermo.
- En un examen: *éxito* = aprobar el examen.

# Distribución de Bernoulli

Imaginemos dos situaciones:

**Primer ejemplo:** Estás inspeccionando una máquina que produce piezas electrónicas. Sabes por experiencia que cada pieza tiene una probabilidad de 0.08 (8%) de salir defectuosa. Si tomas una sola pieza al azar, ¿qué puede pasar? O sale bien o sale defectuosa. Eso es exactamente una Bernoulli.

**Segundo ejemplo:**
Un votante es seleccionado al azar y la probabilidad de que vote por el candidato A es del 65%. Al preguntarle, solo hay dos posibles respuestas: vota por A (éxito) o no vota por A (fracaso). Otra Bernoulli pura.

## Definición
+
Decimos que una variable aleatoria $X$ sigue una distribución Bernoulli con parámetro $p$ si $X$ solo puede tomar dos valores (1 para éxito y 0 para fracaso) y además $$P(X=1)=p\mbox{ y }P(X=0)=1-p$$

## Significado de los parámetros

- $p$: Probabilidad de que ocurra el éxito en ese único ensayo ($0\le p\le 1$).
- El valor $X=1$ representa éxito y el $X=0$ representa fracaso.

## Propiedades principales

- **Esperanza (valor esperado): $E[X] = p$**. Nos dice cuál es la proporción promedio de éxitos que esperamos.

- **Varianza: $Var(X) = p(1-p)$.** Mide la dispersión. Es máxima cuando $p = 0.5$ (máxima incertidumbre) y mínima (cero) cuando estamos seguros del resultado.

## Ejercicios.

1. En un proceso industrial, la probabilidad de que una pieza salga defectuosa es p = 0.08. Calcula la probabilidad de que la primera pieza inspeccionada esté defectuosa (éxito) y la probabilidad de que esté en buen estado. 

2. Un votante seleccionado al azar tiene probabilidad p = 0.65 de votar por el candidato A. Calcula E[X] y Var(X). Interpreta: ¿cuántos votos esperaría obtener el candidato A si se encuestan 200 personas similares?

3. Simula en Python o R 5000 variables Bernoulli con p = 0.4. Calcula la proporción de éxitos obtenida y compárala con la esperanza teórica. Explica si el resultado es razonable.

4. En un hospital, la probabilidad de que un paciente presente una reacción adversa a un medicamento es p = 0.12. Calcula la probabilidad de que el primer paciente no presente reacción. Interpreta cómo afecta esto al monitoreo inicial.

5. Un jugador de básquet tiene probabilidad p = 0.78 de anotar un tiro libre. Calcula E[X] y Var(X). Si lanza 50 tiros libres, ¿cuántos aciertos esperaría? Interpreta la varianza en términos de consistencia del jugador.

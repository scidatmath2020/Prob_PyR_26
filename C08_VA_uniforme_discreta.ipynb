# Distribución Uniforme Discreta

Imaginemos dos situaciones:

**Primer ejemplo:** Estás en un casino y vas a lanzar un dado equilibrado de 6 caras. Cada cara (1, 2, 3, 4, 5, 6) tiene exactamente la misma probabilidad de salir: 1/6. No hay un resultado privilegiado, todos son igualmente probables. Eso es una uniforme discreta.

**Segundo ejemplo:** Una máquina expendedora de números tiene 10 botones numerados del 0 al 9. Cada botón tiene la misma probabilidad de ser presionado al azar. Si seleccionas un número al azar, todos tienen la misma oportunidad de ser elegidos. Otra uniforme discreta pura.

**Tercer ejemplo:** En una rifa con 100 boletos numerados del 1 al 100, cada boleto tiene exactamente la misma probabilidad de ser el ganador. La distribución del número ganador es uniforme discreta.

## Definición: $X\sim U\{a,a+1,a+2,...,b\}$

Decimos que una variable aleatoria $X$ sigue una **distribución Uniforme Discreta** con parámetros $a$ y $b$ si $X$ puede tomar cualquier valor entero en el intervalo $[a, b]$ (incluyendo ambos extremos) y todos tienen la misma probabilidad. Además:

$$P(X = x) = \frac{1}{b - a + 1} \quad \text{para } x = a, a+1, a+2, \ldots, b$$

Donde:
- $a$ es el valor mínimo que puede tomar $X$
- $b$ es el valor máximo que puede tomar $X$
- $n = b - a + 1$ es el número total de valores posibles

**Nota importante:** La distribución solo está definida para valores enteros entre $a$ y $b$.

## Significado de los parámetros

- **$a$ (mínimo)**: El valor más pequeño que puede tomar la variable. Representa el límite inferior del rango.
- **$b$ (máximo)**: El valor más grande que puede tomar la variable. Representa el límite superior del rango.
- **$n = b - a + 1$**: Número total de resultados posibles en el espacio muestral.

**Característica clave:** Todos los valores entre $a$ y $b$ son **equiprobables** (tienen la misma probabilidad). No hay valores más probables que otros.

## Propiedades principales

- **Esperanza (valor esperado)**: 
  $$E[X] = \frac{a + b}{2}$$
  
  *Significado:* Es el punto medio del intervalo. Si lanzamos un dado muchas veces, el promedio de los resultados tiende a 3.5.

- **Varianza**: 
  $$Var(X) = \frac{(b - a + 1)^2 - 1}{12} = \frac{n^2 - 1}{12}$$
  
  *Significado:* Mide la dispersión. La varianza aumenta cuanto más grande es el rango de valores posibles. Para un dado ($n=6$), la varianza es $(36-1)/12 = 35/12 \approx 2.917$.

- **Desviación estándar**:
  $$\sigma = \sqrt{\frac{(b - a + 1)^2 - 1}{12}}$$

- **Función de masa de probabilidad (fmp)**:
  $$f(x) = P(X = x) = \begin{cases} \frac{1}{n} & \text{si } x \in \{a, a+1, \ldots, b\} \\ 0 & \text{en otro caso} \end{cases}$$

- **Función de distribución acumulada (FDA)**:
  $$F(x) = P(X \leq x) = \begin{cases} 0 & \text{si } x < a \\ \frac{\lfloor x \rfloor - a + 1}{n} & \text{si } a \leq x < b \\ 1 & \text{si } x \geq b \end{cases}$$

## Tabla de Funciones para la Distribución Uniforme Discreta

*Tirar `n` veces un dado de ``b-a+1`` caras numeradas de `a` a `b`*

| Explicación | R (stats) | Python (numpy) |
|-------------|-----------|---------------------|
| Generador de números aleatorios | `sample(a:b, n, replace=TRUE)` | `np.random.choice(np.arange(a, b + 1), size=n, replace=True)` |

**Importante.** La $n$ de esta tabla se refiere al número de simulaciones que se quieren hacer, y no al rango de valores que la v.a. puede tomar, que es $b-a+1$.

## Ejercicios

### Ejercicio 1 (Básico)
Un dado equilibrado de 6 caras se lanza una vez. 
- a) Define la variable aleatoria $X$ y sus parámetros.
- b) Calcula la probabilidad de obtener un número par.
- c) Calcula la probabilidad de obtener un número mayor que 4.

### Ejercicio 2 (Esperanza y varianza)
Una rifa tiene 50 boletos numerados del 1 al 50. 
- a) ¿Cuál es la distribución del número ganador?
- b) Calcula la esperanza y varianza del número ganador.
- c) Interpreta el valor esperado en el contexto de la rifa.

### Ejercicio 3 (Simulación en Python)
Simula 10000 lanzamientos de un dado de 12 caras (numerado del 1 al 12). 
- a) Calcula la proporción de veces que sale cada número.
- b) Compara estas proporciones con la probabilidad teórica ($1/12$).
- c) Calcula la media de los resultados y compárala con la esperanza teórica.
- d) Grafica un histograma de frecuencias relativas.

### Ejercicio 4 (Aplicación práctica)
Un sistema de numeración aleatoria genera números enteros del 0 al 99 con igual probabilidad.
- a) ¿Cuál es la probabilidad de generar un número menor que 30?
- b) ¿Cuál es la probabilidad de generar un número entre 40 y 60 (inclusive)?
- c) Si se generan 500 números, ¿cuántos se espera que estén entre 0 y 19?

### Ejercicio 5 (Comparación con Bernoulli)
Compara la Uniforme Discreta con la Bernoulli:
- a) ¿En qué se diferencian fundamentalmente?
- b) ¿Puede considerarse la Bernoulli como un caso particular de la Uniforme Discreta? ¿Por qué?
- c) ¿Cuándo usarías una y cuándo la otra?

### Ejercicio 6 (Juego de azar)
En un juego de casino, se lanza un dado de 20 caras (numerado del 1 al 20). El jugador gana si sale un número primo.
- a) ¿Cuál es la probabilidad de ganar?
- b) Calcula la esperanza y varianza del número obtenido.
- c) Si el premio es de $5 por cada punto obtenido, ¿cuál es el premio esperado?

### Ejercicio 7 (Simulación avanzada)
Genera 5000 muestras de una Uniforme Discreta $U\{5,6,7,...,15\}$ en R o Python.
- a) Calcula la media muestral y la varianza muestral.
- b) Compara con los valores teóricos.
- c) Calcula el porcentaje de valores que caen en el intervalo $E[X] \pm \sigma$.
- d) ¿Coincide aproximadamente con lo que esperarías?

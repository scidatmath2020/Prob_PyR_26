# Distribución binomial

Imaginemos las siguientes situaciones:

**Primer ejemplo:** Lanzamos una moneda equilibrada 10 veces. Contamos el número de veces que sale “Águila” (éxito). Cada lanzamiento es independiente y la probabilidad de éxito es siempre 1/2. El número de águilas sigue una distribución binomial.

**Segundo ejemplo:** En una fábrica, el 5% de los productos son defectuosos. Se toman 20 productos al azar y se cuenta el número de defectuosos. Cada producto es un ensayo independiente con probabilidad de “éxito” (defectuoso) = 0.05.

**Tercer ejemplo:** Un estudiante responde 15 preguntas de verdadero/falso al azar. La probabilidad de acertar cada pregunta es 0.5. El número de respuestas correctas sigue una binomial.

## $X\sim Bin(n,p)$

### Problemas que resuelve

1. Si se realizan $n$ Bernoullis independientes, cada una con probabilidad de éxito $p$, en cuántas de ella hay éxito.

2. Si en un conjunto de ítems hay $p\%$ que tienen una característica $A$ y se seleccionan al azar, con reemplazo, $n$ de ellos, cuántos de ellos tienen la característica $A$.

### Definición 

Decimos que una variable aleatoria $X$ sigue una distribución Binomial con parámetros $n$ y $p$ si, para $x=0,1,2,...,n-1,n$ se tiene $$f(x)=P(X=x)=\binom{n}{x}p^x(1-p)^{n-x}$$

donde:
- $n$ es el número de ensayos independientes (entero positivo).
- $p$ es la probabilidad de éxito en cada ensayo ($0\le p\le 1$).
- $X$ representa el número de éxitos en los $n$ ensayos.

## Parámetros y Significado

- $n$ (ensayos): Número total de pruebas independientes.
- $p$ (probabilidad de éxito): Probabilidad de que ocurra el evento de interés en un solo ensayo.
- $q = 1 − p$: Probabilidad de fracaso.

**Característica clave:** Los ensayos son independientes y cada uno tiene dos resultados posibles (éxito o fracaso) con probabilidades constantes.

## Propiedades principales

- Función de distribución: $$F(x)=P(X\le x)=\sum_{k=0}^x\binom{n}{k}p^k(1-p)^{n-k}$$
- Esperanza (Valor esperado): $E[X] = np$ (número promedio de éxitos esperado).
- Varianza: $\text{Var}(X) = np(1-p)$
- Desviación estándar: $\sigma = \sqrt{np(1-p)}$

### Tabla de Funciones

| Explicación                    | R (stats)                          | Python (scipy)                |
|--------------------------------|------------------------------------|-------------------------------|
| Probabilidad puntual P(X = k)  | `dbinom(k, size=n, prob=p)`        | `binom.pmf(k, n, p)`          |
| Probabilidad acumulada P(X<=k) | `pbinom(k, size=n, prob=p)`        | `binom.cdf(k, n, p)`          |
| Generador de números aleatorios| `rbinom(n_sim, size=n, prob=p)`    | `binom.rvs(n, p, size=n_sim)` |

donde 

- `n` es el tamaño de la muestra
- `k` valor a calcular
- `p` probabilidad de 1 éxito
- `n_sim` número de simulaciones

**Importante.** En **Python** es necesario cargar bibliotecas: `from scipy.stats import binom`

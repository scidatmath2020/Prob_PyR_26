# Distribución Geométrica

Imaginemos las siguientes situaciones:

**Primer ejemplo:** Lanzamos una moneda equilibrada repetidamente hasta que aparece "Águila" por primera vez. Contamos el número de lanzamientos necesarios (incluyendo el éxito). Cada lanzamiento es independiente y la probabilidad de éxito es siempre 1/2. El número de lanzamientos hasta el primer éxito sigue una distribución geométrica.

**Segundo ejemplo:** Un vendedor llama a clientes hasta lograr su primera venta. La probabilidad de vender en cada llamada es 0.1. El número de llamadas necesarias hasta la primera venta sigue una distribución geométrica.

**Tercer ejemplo:** En una máquina tragamonedas, la probabilidad de ganar en cada intento es 0.02. Se juega hasta obtener el primer premio. El número de intentos necesarios sigue una distribución geométrica.

## $X\sim Geom(p)$

### Problemas que resuelve

1. Si se realizan ensayos Bernoulli independientes, cada uno con probabilidad de éxito $p$, ¿en qué ensayo ocurre el primer éxito? 

2. ¿Cuántos intentos se necesitan hasta obtener el primer evento de interés? 

**Ejemplo muy importante.**

Luis pregunta a chicas de forma independiente hasta que una acepta ser su novia. Cada pregunta es un ensayo de Bernoulli:
- *Éxito:* la chica dice que sí con probabilidad $p$.
- *Fracaso:* la chica dice que no.

Suponemos que las respuestas son independientes y que $p$ es constante (idealización fuerte, pero necesaria para el modelo).

## Definición 

**Versión número de ensayos hasta el primer éxito**

En este caso, $X=$ *número total de chicas a las que Luis les pregunta hasta que una dice que sí* (incluyendo a la que acepta).

Decimos que una variable aleatoria $X$ sigue una distribución Geométrica con parámetro $p$ si para $x=1,2,3,4,5,...$ se tiene $$f(x)=P(X=x)=(1-p)^{x-1}p$$

donde:

- $p$ es la probabilidad de éxito en cada ensayo ($0\le p\le 1$)
- $X$ representa el número de ensayos necesarios para obtener el primer éxito

**Importante.** En este caso, $X=1$ significa que la primer chica le dijo que sí; $X=2$ significa que la primer chica dijo que no pero la segunda dijo que sí.

**Versión número de fracasos antes del primer éxito.**

Algunos textos definen la geométrica como el número de fracasos antes del primer éxito. Con esta versión, $X=$ *número de rechazos que recibe Luis antes de que la primera acepte*: $$P(X=x)=(1-p)^xp\mbox{ con }x=0,1,2,3,...$$

**Importante.** En este caso, $X=0$ significa que la primer chica le dijo que sí (no hay fracasos); $X=1$ que hubo un fracaso: la primera dijo que no pero la segunda dijo que sí.

## Parámetros y Significado

- $p$ (probabilidad de éxito): Probabilidad de que ocurra el evento de interés en un solo ensayo.
- $q=1−p$: Probabilidad de fracaso en cada ensayo.
- **Propiedad clave (falta de memoria):** $P(X > m+n|X>m)=P(X>n)$.

Esto significa que si ya hemos realizado $m$ ensayos sin éxito, la probabilidad de necesitar $n$ ensayos adicionales es la misma que si empezáramos desde cero. ¡El proceso "no recuerda" los fracasos pasados!

## Propiedades principales
Para la versión $X =$ *número de ensayos hasta el primer éxito* ($x=1,2,3,...$)

- **Función de distribución:** $F(x)=P(X\le x)=1-(1-p)^x$
- **Esperanza:** $E[X]=\frac{1}{p}$ 
- **Varianza:** $Var(X)=\frac{1-p}{p^2}$
- **Desviación estándar:** $sd(X)=\frac{\sqrt{1-p}}{p^2}$

Para la versión $X=$ *número de fracasos antes del primer éxito* ($x=0,1,2,3,...$)

- **Función de distribución:** $F(x)=P(X\le x)=1-(1-p)^{x+1}$
- $E[X]=\frac{1-p}{p}$
- $Var(X)=\frac{1-p}{p^2}$

## Tabla de Funciones para versión Total de ensayos hasta primer éxito (x=1,2,3...)

| Explicación                    | R (stats)                          | Python (scipy)                |
|--------------------------------|------------------------------------|-------------------------------|
| Probabilidad puntual P(X = k)  | `dgeom(k-1, prob=p)`               | `geom.pmf(k, p)`            |
| Probabilidad acumulada P(X≤k)  | `pgeom(k-1, prob=p)`               | `geom.cdf(k, p)`            |
| Generador de números aleatorios| `rgeom(n_sim, prob=p)`             | `geom.rvs(p, size=n_sim)`     |

**Donde:**
- `k` = número de ENSAYOS hasta el primer éxito (incluyendo el éxito)
- `p` = probabilidad de éxito en cada ensayo
- `n_sim` = número de simulaciones a generar

**¡IMPORTANTE!** 
1. En R las funciones trabajan con la versión de **número de FRACASOS** antes del primer éxito ($x=0,1,2,...$). Por eso usamos `k-1`.
2. En Python las funciones trabajan con la versión de **número de ensayos hasta primer éxito** antes del primer éxito ($x=1,2,...$). Por eso usamos `k`.


## Ejercicios

**Ejercicio 1 (Básico)**
Un estudiante responde preguntas de opción múltiple hasta acertar una. Cada pregunta tiene 4 opciones y responde al azar.

- Define la variable aleatoria X (número de intentos) y su parámetro.
- ¿Cuál es la probabilidad de que necesite exactamente 5 intentos?
- ¿Cuál es la probabilidad de que necesite más de 3 intentos?
- ¿Cuál es el número esperado de intentos?

**Ejercicio 2 (Esperanza y varianza)**
En un juego de azar, la probabilidad de ganar es 0.15 en cada partida. Se juega hasta ganar por primera vez.

- Define la variable aleatoria X y su parámetro.
- Calcula la esperanza y la varianza del número de partidas necesarias.
- Interpreta el valor esperado en el contexto del juego.

**Ejercicio 3 (Propiedad de falta de memoria)**
Un jugador ha lanzado un dado 10 veces sin obtener un 6. Si la probabilidad de obtener 6 es 1/6:

- ¿Cuál es la probabilidad de obtener el primer 6 en el lanzamiento 15?
- ¿Cuál es la probabilidad de obtener el primer 6 en el lanzamiento 5, dado que no ha salido en los 4 primeros lanzamientos?
- ¿Qué propiedad se cumple? Explica por qué es contraintuitiva.

**Ejercicio 4 (Simulación en Python)**
Simula 10,000 repeticiones del número de lanzamientos de una moneda hasta obtener la primera cara (p=0.5).

- Calcula la proporción de simulaciones donde se necesitaron exactamente 3 lanzamientos.
- Compara esta proporción con la probabilidad teórica.
- Calcula la media y la varianza de la simulación y compáralas con los valores teóricos.

**Ejercicio 5 (Aplicación práctica)**
En un control de calidad, se inspeccionan productos hasta encontrar el primer defectuoso. La probabilidad de que un producto sea defectuoso es 0.02.

- ¿Cuál es la probabilidad de que el primer defectuoso aparezca en el producto número 10?
- ¿Cuál es la probabilidad de que se necesiten más de 15 inspecciones?
- ¿Cuántos productos se espera inspeccionar en promedio?
- ¿Cuál es la probabilidad de que el primer defectuoso aparezca antes del producto 5?

**Ejercicio 6 (Comparación con Binomial)**
Compara la distribución Geométrica con la Binomial:

- ¿En qué se diferencian fundamentalmente en cuanto al número de ensayos?
- ¿En qué tipo de situaciones usarías cada una?
- ¿Qué relación existe entre ambas distribuciones?

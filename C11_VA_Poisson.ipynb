# Distribución de Poisson

Imaginemos las siguientes situaciones:

**Primer ejemplo:** En una central telefónica, se reciben en promedio 5 llamadas por minuto. 
El número de llamadas que llegan en un minuto específico sigue una distribución de Poisson.

**Segundo ejemplo:** El número de accidentes de tráfico que ocurren en una intersección 
en una semana sigue una distribución de Poisson, con un promedio de 2 accidentes por semana.

**Tercer ejemplo:** El número de clientes que llegan a un banco durante una hora sigue 
una distribución de Poisson con media 10 clientes por hora.

**Cuarto ejemplo:** El número de errores de mecanografía por página en un libro sigue 
una distribución de Poisson con media 0.5 errores por página.

**Característica clave:** La Poisson modela el número de ocurrencias de un evento en 
un intervalo de tiempo, espacio o área, cuando los eventos ocurren de forma aleatoria 
e independiente a una tasa promedio constante.


## $X \sim Pois(\lambda)$

### Problemas que resuelve

1. ¿Cuántos eventos ocurren en un intervalo fijo de tiempo/espacio cuando:
   - Los eventos ocurren de forma independiente
   - La tasa promedio de ocurrencia es constante ($\lambda$)
   - Dos eventos no pueden ocurrir en el mismo instante

### Definición

Decimos que una variable aleatoria $X$ sigue una distribución de Poisson con 
parámetro $\lambda$ si, para $x = 0, 1, 2, 3, ...$ se tiene

$$f(x) = P(X = x) = \frac{e^{-\lambda} \lambda^x}{x!}$$

donde:
- $\lambda$ es la tasa promedio de ocurrencia (media) $(\lambda > 0)$
- $X$ representa el número de eventos en el intervalo de interés
- $x$ es el número de ocurrencias (0, 1, 2, ...)

**Nota:** $e \\approx 2.71828$ es la base del logaritmo natural.


## Parámetros y Significado

- $\lambda$ (lambda): Tasa promedio de ocurrencia del evento en el intervalo de interés.
  - Representa el número esperado de eventos en el intervalo
  - Es tanto la media como la varianza de la distribución
  - $\lambda > 0$

**Características clave:**
1. Los eventos ocurren de forma independiente unos de otros
2. La probabilidad de que ocurra un evento en un intervalo muy pequeño es proporcional a la longitud del intervalo
3. La probabilidad de que ocurran dos o más eventos en un intervalo muy pequeño es despreciable
4. Los eventos ocurren a una tasa promedio constante

**Propiedad importante.** La Poisson NO tiene un número fijo de ensayos (como la Binomial) 
ni espera hasta el primer éxito (como la Geométrica). Simplemente cuenta eventos en un intervalo fijo.


## Propiedades principales

- Función de distribución: $$F(x) = P(X \le x) = \sum_{k=0}^x \frac{e^{-\lambda} \lambda^k}{k!}$$

- Esperanza (Valor esperado): $$E[X] = \lambda$$

- Varianza: $$\text{Var}(X) = \lambda$$

- Desviación estándar: $$\sigma = \sqrt{\lambda}$$

## Tabla de Funciones

| Explicación                    | R (stats)                          | Python (scipy)                |
|--------------------------------|------------------------------------|-------------------------------|
| Probabilidad puntual P(X = k)  | `dpois(k, lambda=lambda)`          | `poisson.pmf(k, lambda)`      |
| Probabilidad acumulada P(X≤k)  | `ppois(k, lambda=lambda)`          | `poisson.cdf(k, lambda)`      |
| Generador de números aleatorios| `rpois(n_sim, lambda=lambda)`      | `poisson.rvs(lambda, size=n_sim)` |

Donde:
- `k` = valor a evaluar (número de ocurrencias)
- `lambda` = tasa promedio (parámetro de la Poisson)
- `n_sim` = número de simulaciones a generar

**NOTA:**
1. Tanto en R como en Python, el parámetro `lambda` tiene el MISMO significado (tasa promedio de ocurrencia). Solo cambia la sintaxis.
2. En *Python* se requiero `from scipy.stats import poisson`

## Ejercicios

**Ejercicio 1 (Básico)**
En una biblioteca, el número de personas que entran por hora sigue una distribución 
de Poisson con media 8 personas por hora.

- Define la variable aleatoria X y su parámetro.
- ¿Cuál es la probabilidad de que entren exactamente 5 personas en una hora?
- ¿Cuál es la probabilidad de que entren a lo sumo 3 personas en una hora?
- ¿Cuál es la probabilidad de que entren al menos 10 personas en una hora?

**Ejercicio 2 (Esperanza y varianza)**
El número de errores por página en un libro sigue una Poisson con media 0.3 errores por página.

- Define la variable aleatoria X y su parámetro.
- Calcula la esperanza y la varianza.
- ¿Cuál es la probabilidad de que una página tenga exactamente 1 error?
- ¿Cuál es la probabilidad de que una página no tenga errores?

**Ejercicio 3 (Simulación en Python)**
Simula 10,000 valores de una Poisson con $\lambda = 4$.

- Calcula la proporción de simulaciones con exactamente 4 ocurrencias.
- Compara con la probabilidad teórica.
- Calcula la media y varianza de la simulación y compáralas con los valores teóricos.

**Ejercicio 4 (Aplicación práctica)**
En un centro de llamadas, se reciben en promedio 12 llamadas cada 10 minutos.

- ¿Cuál es la probabilidad de recibir exactamente 10 llamadas en 10 minutos?
- ¿Cuál es la probabilidad de recibir menos de 5 llamadas en 10 minutos?
- ¿Cuántas llamadas se esperan en 30 minutos?
- ¿Cuál es la probabilidad de recibir exactamente 15 llamadas en 30 minutos?

**Ejercicio 5 (Comparación con Binomial y Geométrica)**
Compara la distribución de Poisson con la Binomial y la Geométrica:

- ¿En qué se diferencia fundamentalmente en cuanto al número de ensayos?
- ¿Cuándo usarías Poisson en lugar de Binomial?
- ¿Qué relación existe entre Poisson y Binomial?

**Ejercicio 6 (Simulación avanzada)**
Genera 5000 muestras de una Poisson con $\lambda = 2.5$.

- Calcula la media muestral y la varianza muestral.
- Compara con los valores teóricos.

**Ejercicio 7 (reescalado).** Un banco recibe en promedio 6 clientes cada 15 minutos. 

- ¿Cuál es la probabilidad de que lleguen exactamente 4 clientes en 15 minutos?
- ¿Cuál es la probabilidad de que lleguen a lo sumo 2 clientes en 15 minutos?
- ¿Cuántos clientes se esperan en 1 hora (60 minutos)?
- ¿Cuál es la probabilidad de que lleguen mas de 35 clientes en dos horas?

**Ejercicio 8 (Reescalado).** En una autopista ocurren en promedio 2.5 accidentes por día (8 horas de operación).

- ¿Cuál es la probabilidad de exactamente 3 accidentes en un día?
- ¿Cuál es la probabilidad de menos de 1 accidente en un día? (es decir, cero accidentes)
- ¿Cuántos accidentes se esperan en un mes de 31 días?
- ¿Cuál es la probabilidad de al menos 12 accidentes en tres días?

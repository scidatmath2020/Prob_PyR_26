# Distribución Uniforme Continua

Imaginemos las siguientes situaciones:

**Primer ejemplo:** Un autobús pasa cada 10 minutos. Llegamos a la parada en un momento 
aleatorio sin saber el horario. El tiempo de espera hasta que llegue el próximo autobús 
sigue una distribución uniforme continua entre 0 y 10 minutos.

**Segundo ejemplo:** Se elige un número real al azar entre 0 y 1. La probabilidad de que 
caiga en cualquier subintervalo de longitud fija es la misma.

**Tercer ejemplo:** El error de redondeo al medir con una regla milimétrica sigue 
una distribución uniforme entre -0.5 y 0.5 mm.

**Cuarto ejemplo:** Se selecciona un punto al azar dentro de un círculo de radio R. 
La coordenada x del punto sigue una distribución uniforme en [-R, R].

**Característica clave:** Todos los valores en el intervalo [a, b] tienen la misma 
"densidad" de probabilidad (son igualmente probables).


## $X \sim U(a, b)$  o  $X \sim \text{Uniforme}(a, b)$

### Problemas que resuelve

1. Modelar situaciones donde todos los valores en un intervalo continuo son igualmente probables.

2. Generar números aleatorios (base para simular otras distribuciones).

3. Modelar el error de redondeo, tiempos de espera sin información previa, 
   o selección aleatoria en un intervalo.

### Definición

Decimos que una variable aleatoria continua $X$ sigue una distribución Uniforme en el intervalo $[a, b]$ si su función de densidad de probabilidad (fdp) es:

$$f(x) = \begin{cases} 
\frac{1}{b-a} & \text{si } a \le x \le b \\\\
0 & \text{en otro caso}
\end{cases}$$

donde:
- $a$ es el límite inferior del intervalo ($a < b$)
- $b$ es el límite superior del intervalo ($a < b$)
- $X$ puede tomar cualquier valor real en $[a, b]$

**Nota importante:** Como es una variable continua, la probabilidad de un valor puntual es cero: $P(X = x) = 0$ para cualquier $x$ específico. Solo tienen sentido las probabilidades en intervalos.


## Parámetros y Significado

- $a$ (límite inferior): Valor mínimo que puede tomar la variable.
- $b$ (límite superior): Valor máximo que puede tomar la variable.
- **Rango:** $b - a$ es la longitud total del intervalo.

**Características clave:**
1. **Densidad constante:** $f(x) = 1/(b-a)$ en todo el intervalo.
2. **Área total bajo la curva:** $\int_a^b \frac{1}{b-a} dx = 1$
3. **Probabilidad en subintervalos:** 
   $$P(c \le X \le d) = \frac{d-c}{b-a}$$ para $a \le c \le d \le b$

## Propiedades principales

- Función de densidad: $$f(x) = \frac{1}{b-a}, \quad a \le x \le b$$

![imagenes](im005.png)

- Función de distribución (FDA): $$F(x) = P(X \le x) = \begin{cases}
0 & \text{si } x < a \\\\
\frac{x-a}{b-a} & \text{si } a \le x \le b \\\\
1 & \text{si } x > b
\end{cases}$$

![imagenes](im006.png)

- Esperanza (Valor esperado): $$E[X] = \frac{a+b}{2}$$

- Varianza: $$\text{Var}(X) = \frac{(b-a)^2}{12}$$

- Desviación estándar: $$\sigma = \frac{b-a}{\sqrt{12}}$$

## Tabla de Funciones

| Explicación                    | R (stats)                          | Python (scipy)                |
|--------------------------------|------------------------------------|-------------------------------|
| Densidad f(x)                  | `dunif(x, min=a, max=b)`           | `uniform.pdf(x, a, b-a)`      |
| Distribución acumulada F(x)    | `punif(x, min=a, max=b)`           | `uniform.cdf(x, a, b-a)`      |
| Cuantil (percentil inverso)    | `qunif(p, min=a, max=b)`           | `uniform.ppf(p, a, b-a)`      |
| Generador de números aleatorios| `runif(n_sim, min=a, max=b)`       | `uniform.rvs(a, b-a, size=n_sim)` |

**Donde:**
- `x` = valor a evaluar
- `a` = límite inferior del intervalo
- `b` = límite superior del intervalo
- `p` = probabilidad (entre 0 y 1) para el cuantil
- `n_sim` = número de simulaciones a generar

**¡IMPORTANTE!** En Python se requiere ``scipy.stats.uniform``

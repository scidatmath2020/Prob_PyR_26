# Distribución Exponencial

Imaginemos las siguientes situaciones:

**Primer ejemplo:** El tiempo que transcurre entre la llegada de dos clientes consecutivos 
a un banco sigue una distribución exponencial. Si en promedio llegan 10 clientes por hora, 
el tiempo entre llegadas es Exponencial(λ=10).

**Segundo ejemplo:** La duración de vida de un componente electrónico sigue una 
distribución exponencial con tasa de fallo constante. Si el promedio de vida es 1000 horas, 
entonces λ = 1/1000.

**Tercer ejemplo:** El tiempo de espera hasta que ocurra el primer éxito en un proceso 
de Poisson (como llamadas telefónicas, llegadas de clientes, etc.) sigue una 
distribución exponencial.

**Cuarto ejemplo:** El tiempo que un servidor web tarda en responder a una solicitud 
sigue una distribución exponencial con media 0.5 segundos.

## $X \sim \text{Exp}(\lambda)$

### Problemas que resuelve

1. Modelar el tiempo de espera hasta que ocurre un evento en un proceso de Poisson.

### Definición

Decimos que una variable aleatoria continua $X$ sigue una distribución Exponencial 
con parámetro $\lambda > 0$ si su función de densidad de probabilidad (fdp) es:

$$f(x) = \begin{cases} 
\lambda e^{-\lambda x} & \text{si } x \ge 0 \\
0 & \text{si } x < 0
\end{cases}$$

donde:
- $\lambda$ es la **tasa de ocurrencia** (tasa de fallo o tasa de llegada)
- $\lambda > 0$ es el número esperado de eventos por unidad de tiempo
- $X$ representa el tiempo de espera hasta el próximo evento

**Nota:** Algunos textos parametrizan la exponencial con $\beta = 1/\lambda$ (la media):
$$f(x) = \frac{1}{\beta} e^{-x/\beta}, \quad x \ge 0$$
donde $\beta = E[X] = 1/\lambda$ es el tiempo promedio entre eventos.
	

## Parámetros y Significado

- $\lambda$ (tasa): Número promedio de eventos por unidad de tiempo.
  - También llamada **tasa de fallo** o **tasa de llegada**
  - $\lambda > 0$
  - Unidades: eventos por unidad de tiempo

- **Alternativa:** $\beta = 1/\lambda$ (tiempo promedio entre eventos)
  - $\beta$ es la media de la distribución
  - Unidades: tiempo por evento

**Características clave:**
1. **Soporte:** $x \ge 0$ (tiempos no negativos)
2. **Monótona decreciente:** La densidad es máxima en $x=0$ y decrece exponencialmente
3. **Falta de memoria:** $P(X > s+t \mid X > s) = P(X > t)$
4. **Relación con Poisson:** Si los eventos siguen un proceso Poisson con tasa $\lambda$, 
   entonces el tiempo entre eventos consecutivos es $\text{Exp}(\lambda)$

**Interpretación de $\lambda$:**
- $\lambda$ grande → eventos frecuentes → tiempos entre eventos cortos
- $\lambda$ pequeña → eventos raros → tiempos entre eventos largos


## Propiedades principales

- Función de densidad: $$f(x) = \lambda e^{-\lambda x}, \quad x \ge 0$$

- Función de distribución (FDA): $$F(x) = P(X \le x) = 1 - e^{-\lambda x}, \quad x \ge 0$$

- Esperanza (Valor esperado): $$E[X] = \frac{1}{\lambda}$$

- Varianza: $$\text{Var}(X) = \frac{1}{\lambda^2}$$

- Desviación estándar: $$\sigma = \frac{1}{\lambda}$$

## Tabla de Funciones

| Explicación                    | R (stats)                          | Python (scipy)                |
|--------------------------------|------------------------------------|-------------------------------|
| Densidad f(x)                  | `dexp(x, rate=lambda)`             | `expon.pdf(x, scale=1/lambda)`|
| Distribución acumulada F(x)    | `pexp(x, rate=lambda)`             | `expon.cdf(x, scale=1/lambda)`|
| Cuantil (percentil inverso)    | `qexp(p, rate=lambda)`             | `expon.ppf(p, scale=1/lambda)`|
| Generador de números aleatorios| `rexp(n_sim, rate=lambda)`         | `expon.rvs(scale=1/lambda, size=n_sim)` |

**Donde:**
- `x` = valor a evaluar (tiempo)
- `lambda` = tasa de ocurrencia (parámetro de la exponencial)
- `p` = probabilidad (entre 0 y 1) para el cuantil
- `n_sim` = número de simulaciones a generar

**¡IMPORTANTE!** En Python (scipy.stats.expon), la parametrización es por `scale`:
- `scale = 1/lambda` (tiempo promedio entre eventos)
- NO se usa `lambda` directamente

En R, se usa `rate = lambda` directamente.

## Ejercicios

**Ejercicio 1 (Básico)**
El tiempo de vida de una lámpara sigue una distribución exponencial con una vida 
media de 1000 horas.

- Define la variable aleatoria X y su parámetro (tasa λ).
- ¿Cuál es la probabilidad de que la lámpara dure más de 1200 horas?
- ¿Cuál es la probabilidad de que la lámpara falle antes de las 800 horas?
- ¿Cuál es la vida mediana de la lámpara?

**Ejercicio 2 (Esperanza y varianza)**
En una central telefónica, las llamadas llegan a razón de 3 por minuto. 
El tiempo entre llamadas sigue una distribución exponencial.

- Define la variable aleatoria X y su parámetro (tasa λ).
- Calcula la esperanza y la varianza del tiempo entre llamadas.
- ¿Cuál es la probabilidad de que pasen más de 30 segundos sin llamadas?
- ¿Cuál es el tiempo máximo que se espera entre llamadas el 90% de las veces?

**Ejercicio 3 (Propiedad de falta de memoria)**
Un sistema ha estado funcionando sin fallos durante 500 horas. Su tiempo de vida 
sigue una exponencial con media 1000 horas.

- ¿Cuál es la probabilidad de que falle en las próximas 200 horas?
- Compara con la probabilidad de que un sistema nuevo falle en sus primeras 200 horas.
- ¿Qué propiedad se cumple? Explica su significado en este contexto.

**Ejercicio 4 (Simulación en Python)**
Simula 10,000 valores de una distribución exponencial con λ = 0.02.

- Calcula la proporción de simulaciones con valores mayores a 50.
- Compara con la probabilidad teórica.
- Calcula la media y varianza de la simulación y compáralas con los valores teóricos.
- Grafica un histograma de frecuencias relativas y superpone la densidad teórica.

**Ejercicio 5 (Aplicación práctica - Teoría de colas)**
En un restaurante de comida rápida, el tiempo de servicio en el mostrador sigue 
una distribución exponencial con media 3 minutos por cliente.

- Define la variable aleatoria X y su parámetro λ.
- ¿Cuál es la probabilidad de que un cliente sea atendido en menos de 2 minutos?
- ¿Cuál es la probabilidad de que un cliente tarde más de 5 minutos en ser atendido?
- Si ya han pasado 2 minutos desde que comenzó a atender a un cliente, 
  ¿cuál es la probabilidad de que termine en el próximo minuto?
- Calcula el tiempo máximo de servicio que cubre el 80% de los clientes.

**Ejercicio 6 (Comparación con Poisson y Geométrica)**
Compara la distribución Exponencial con la Poisson y la Geométrica:

- ¿Qué relación existe entre la Exponencial y la Poisson?
- ¿Qué relación existe entre la Exponencial y la Geométrica?
- ¿En qué se diferencia la propiedad de falta de memoria en la Exponencial vs la Geométrica?

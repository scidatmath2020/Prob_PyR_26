# Distribución Normal o Gaussiana

## $X \sim N(\mu, \sigma^2)$  o  $X \sim \mathcal{N}(\mu, \sigma^2)$

### Problemas que resuelve

1. Modelar fenómenos donde los valores se concentran alrededor de un valor central 
   con desviaciones simétricas.

2. Modelar errores aleatorios en mediciones y procesos de fabricación.

3. Modelar características físicas y biológicas (estaturas, pesos, coeficientes 
   intelectuales, etc.).

4. Base para muchos métodos estadísticos (pruebas de hipótesis, intervalos de 
   confianza, regresión, etc.).

### Definición

Decimos que una variable aleatoria continua $X$ sigue una distribución Normal 
(o Gaussiana) con parámetros $\mu$ y $\sigma^2$ si su función de densidad de 
probabilidad (fdp) es:

$$f(x) = \frac{1}{\sigma \sqrt{2\pi}} \exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right), \quad -\infty < x < \infty$$

donde:
- $\mu$ es la media (centro de la distribución)
- $\sigma$ es la desviación típica (dispersión)
- $\sigma^2$ es la varianza
- $\mu \in \mathbb{R}$, $\sigma > 0$

**Nota:** La densidad tiene soporte en toda la recta real ($-\infty, \infty$).


## Parámetros y Significado

- $\mu$ (media): 
  - Centro de la distribución (punto de máxima densidad)
  - Mediana y moda coinciden con $\mu$
  - Desplaza la campana horizontalmente
  - $\mu \in \mathbb{R}$

- $\sigma$ (desviación típica):
  - Mide la dispersión alrededor de la media
  - A mayor $\sigma$, más aplanada y ancha la campana
  - A menor $\sigma$, más apuntada y estrecha la campana
  - $\sigma > 0$

- $\sigma^2$ (varianza):
  - Cuadrado de la desviación típica
  - $\sigma^2 > 0$

**Forma de la fdp:** Campana de Gauss simétrica alrededor de $\mu$.

**Características clave:**
1. **Simetría:** $f(\mu + x) = f(\mu - x)$
2. **Máximo en $x = \mu$:** $f(\mu) = \frac{1}{\sigma \sqrt{2\pi}}$
3. **Puntos de inflexión:** En $x = \mu \pm \sigma$
4. **Asintótica:** $f(x) \to 0$ cuando $x \to \pm\infty$
5. **Regla empírica (aproximada):**
   - 68% de los datos en $\mu \pm \sigma$
   - 95% de los datos en $\mu \pm 2\sigma$
   - 99.7% de los datos en $\mu \pm 3\sigma$

**Distribución Normal Estándar:** $Z \sim N(0, 1)$
- $\mu = 0$, $\sigma = 1$
- Densidad: $\phi(z) = \frac{1}{\sqrt{2\pi}} e^{-z^2/2}$
- Función de distribución: $\Phi(z)$
- Tablas y software usualmente trabajan con $Z$

**Estandarización (tipificación):**
$$Z = \frac{X - \mu}{\sigma} \sim N(0, 1)$$

**Estandarización inversa:**
$$X = \mu + \sigma Z \sim N(\mu, \sigma^2)$$


## Propiedades principales

- Función de densidad: $$f(x) = \frac{1}{\sigma \sqrt{2\pi}} \exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right), \quad -\infty < x < \infty$$

- Función de distribución (FDA): $$F(x) = \Phi\left(\frac{x-\mu}{\sigma}\right)$$
  donde $\Phi$ es la FDA de la Normal estándar.

- Esperanza (Valor esperado): $$E[X] = \mu$$

- Varianza: $$\text{Var}(X) = \sigma^2$$

- Desviación estándar: $$\sigma$$

**Percentiles notables:**
- Q1 = $\mu - 0.6745\sigma$
- Q2 = $\mu$ (mediana)
- Q3 = $\mu + 0.6745\sigma$
- IQR = $1.349\sigma$
- Percentil 95 = $\mu + 1.645\sigma$
- Percentil 97.5 = $\mu + 1.96\sigma$
- Percentil 99 = $\mu + 2.326\sigma$


## Tabla de Funciones

| Explicación                    | R (stats)                          | Python (scipy)                |
|--------------------------------|------------------------------------|-------------------------------|
| Densidad f(x)                  | `dnorm(x, mean=mu, sd=sigma)`      | `norm.pdf(x, mu, sigma)`      |
| Distribución acumulada F(x)    | `pnorm(x, mean=mu, sd=sigma)`      | `norm.cdf(x, mu, sigma)`      |
| Cuantil (percentil inverso)    | `qnorm(p, mean=mu, sd=sigma)`      | `norm.ppf(p, mu, sigma)`      |
| Generador de números aleatorios| `rnorm(n_sim, mean=mu, sd=sigma)`  | `norm.rvs(mu, sigma, size=n_sim)` |

**Donde:**
- `x` = valor a evaluar
- `mu` = media de la distribución
- `sigma` = desviación típica
- `p` = probabilidad (entre 0 y 1) para el cuantil
- `n_sim` = número de simulaciones a generar

**¡IMPORTANTE!** Tanto en R como en Python, los parámetros son:
- `mean = mu` (media)
- `sd = sigma` (desviación típica)
- **NO** se usa la varianza directamente, sino la desviación típica

# Teorema del Límite Central (TLC)

Imaginemos las siguientes situaciones:

**Primer ejemplo:** Lanzamos un dado 30 veces y calculamos el promedio de los resultados.
Si repetimos este experimento muchas veces, los promedios se distribuirán aproximadamente 
como una Normal, aunque la distribución de cada lanzamiento sea Uniforme discreta.

**Segundo ejemplo:** Tomamos muestras de 50 personas y calculamos la estatura promedio 
de cada muestra. Aunque las estaturas individuales no sean perfectamente Normales, 
la distribución de las medias muestrales se aproximará a una Normal.

**Característica clave:** Sin importar la distribución de la población original, 
la distribución de la media muestral (para muestras suficientemente grandes) 
se aproxima a una Normal.

## Teorema del Límite Central

### ¿Qué problema resuelve?

1. Permite hacer inferencias sobre la media poblacional sin conocer la distribución 
   exacta de la población.

2. Justifica el uso de la distribución Normal para aproximar distribuciones de 
   sumas y promedios de variables aleatorias.

3. Base para intervalos de confianza y pruebas de hipótesis sobre la media.


## Enunciado del Teorema

Sea $X_1, X_2, ..., X_n$ una muestra aleatoria de tamaño $n$ de cualquier población 
con media $\mu$ y varianza $\sigma^2$ finitas ($0 < \sigma^2 < \infty$).

Definimos la media muestral:
$$\bar{X}_n = \frac{1}{n} \sum_{i=1}^n X_i$$

Entonces, cuando $n \to \infty$:

$$\sqrt{n}(\bar{X}_n - \mu) \xrightarrow{d} N(0, \sigma^2)$$

Equivalentemente:
$$\frac{\bar{X}_n - \mu}{\sigma/\sqrt{n}} \xrightarrow{d} N(0, 1)$$

Es decir, para $n$ suficientemente grande:

$$\bar{X}_n \approx N\left(\mu, \frac{\sigma^2}{n}\right)$$

### Notación

- $\xrightarrow{d}$ denota convergencia en distribución
- La distribución de $\bar{X}_n$ se aproxima a una Normal
- El error de aproximación decrece como $1/\sqrt{n}$

**MUY IMPORTANTE** Si tomo $n$ medidas de una característica $X$ ($n$ grande), la cual tiene media $\mu$ y desviación $\sigma$, donde las medidas son independientes, entonces, sin importar cómo sea la variable aleatoria $X$, se cumple $$P(\overline{X}_n\le x)\approx \Phi\left(\frac{x-\mu}{\sigma/\sqrt{n}}\right)$$

## Versiones del TLC

- Si $S_n=X_1+X_2+...+X_n$ entonces $S_n\approx N(n\mu,n\sigma^2)$ (la varianza es $n\sigma^2$)

- Si $\overline{X}_n=\frac{S_n}{n}$ entonces $\overline{X}_n\approx N(\mu,\sigma^2/n)$ (la varianza es $\sigma^2/n$)

En particular, si $X_i\sim Ber(p)$ entonces $S_n$ es el total de éxitos y $\overline{X}_n$ es la proporcición.

## Condiciones y Parámetros

### Condiciones necesarias

1. **Muestras independientes:** $X_1, X_2, ..., X_n$ deben ser independientes.

2. **Misma distribución:** Las variables deben tener la misma distribución 
   (o al menos la misma media y varianza).

3. **Varianza finita:** $\sigma^2 < \infty$ (la varianza debe existir y ser finita).

4. **Tamaño muestral suficiente:** $n$ debe ser "suficientemente grande".
   - Generalmente, $n \ge 30$ es un buen punto de partida
   - Para distribuciones muy asimétricas, puede necesitarse $n$ mayor
   - Para distribuciones simétricas, $n \ge 20$ puede ser suficiente

### Lo que NO dice el TLC

1. No dice que la distribución de los datos individuales sea Normal.
2. No dice nada sobre la varianza de la población (necesita estimarse).
3. No aplica si la varianza no existe (distribuciones de colas muy pesadas).
4. No garantiza buena aproximación para muestras pequeñas.

## Propiedades principales

### 1. Media y varianza de la media muestral

Para cualquier $n$ (sin necesidad de TLC):
$$E[\bar{X}_n] = \mu$$
$$\text{Var}(\bar{X}_n) = \frac{\sigma^2}{n}$$
$$\sigma_{\bar{X}_n} = \frac{\sigma}{\sqrt{n}}$$

### 2. Error estándar de la media (SEM)

El error estándar de la media es:
$$\text{SEM} = \frac{\sigma}{\sqrt{n}}$$
- Disminuye con $\sqrt{n}$ (para reducir a la mitad, necesito $4\times$ el tamaño muestral)
- Mide la variabilidad de la media muestral

### Tabla de Funciones para Aplicaciones del TLC

| Aplicación                    | R (stats)                          | Python (scipy)                |
|--------------------------------|------------------------------------|-------------------------------|
| $P(\bar{X} ≤ x)$ con μ, σ, n   | `pnorm(x, mean=mu, sd=sigma/sqrt(n))` | `norm.cdf(x, mu, sigma/sqrt(n))` |
| $P(\bar{X} > x)$ con μ, σ, n   | `1 - pnorm(x, mean=mu, sd=sigma/sqrt(n))` | `1 - norm.cdf(x, mu, sigma/sqrt(n))` |
| $P(a ≤ \bar{X} ≤ b)$ con μ, σ, n | `pnorm(b, mu, sigma/sqrt(n)) - pnorm(a, mu, sigma/sqrt(n))` | `norm.cdf(b, mu, sigma/sqrt(n)) - norm.cdf(a, mu, sigma/sqrt(n))` |
| Percentil de $\bar{X}$ para α  | `qnorm(alpha, mean=mu, sd=sigma/sqrt(n))` | `norm.ppf(alpha, mu, sigma/sqrt(n))` |

**Donde:**
- `mu` = media poblacional
- `sigma` = desviación típica poblacional
- `n` = tamaño de la muestra
- `x`, `a`, `b` = valores a evaluar
- `alpha` = probabilidad para el cuantil

**NOTA:** El TLC se aplica usando la distribución Normal con:
$$\text{mean} = \mu$$
$$\text{sd} = \frac{\sigma}{\sqrt{n}}$$

## Ejercicios

**Ejercicio 1 (Básico)**
Las estaturas de una población tienen media 170 cm y desviación típica 10 cm. 
Se toma una muestra de 36 personas.

- ¿Cuál es la distribución aproximada de la media muestral?
- ¿Cuál es la probabilidad de que la media muestral sea menor a 168 cm?
- ¿Cuál es la probabilidad de que la media muestral esté entre 168 y 172 cm?
- ¿Cuál es la media muestral que deja el 90% de las medias por debajo?

**Ejercicio 2 (Esperanza y error estándar)**
El tiempo de espera en un banco tiene media 5 minutos y desviación típica 2 minutos. 
Se toma una muestra de 49 clientes.

- Calcula la media y el error estándar de la media muestral.
- ¿Cuál es la probabilidad de que el tiempo promedio de espera sea mayor a 5.5 minutos?
- ¿Qué tamaño muestral se necesita para que el error estándar sea 0.2 minutos?

**Ejercicio 3 (Simulación del TLC en Python)**
Simula el TLC para una distribución exponencial con λ=1 (media=1, varianza=1).

- Genera 5000 muestras de tamaño n=5, n=30 y n=100.
- Para cada n, calcula la media de cada muestra.
- Grafica los histogramas de las medias muestrales.
- Superpone la curva Normal teórica N(μ, σ²/n).
- Comenta cómo mejora la aproximación al aumentar n.

**Ejercicio 4 (Aproximación de la Binomial)**
En una fábrica, el 40% de los productos son defectuosos. Se toman 200 productos.

- Usando el TLC, aproxima la probabilidad de que haya entre 70 y 90 defectuosos.
- Compara con el valor exacto usando la distribución Binomial.
- ¿Se cumplen las condiciones para usar la aproximación Normal?

**Ejercicio 5 (Aplicación práctica)**
Una máquina llena botellas con un promedio de 500 ml y desviación típica 5 ml. 
Se toma una muestra de 16 botellas.

- ¿Cuál es la probabilidad de que la media muestral sea menor a 498 ml?
- Si la máquina se ajusta para que la media sea 502 ml, ¿cómo cambia la probabilidad anterior?
- ¿Qué tamaño muestral se necesita para que $P(\bar{X} < 498) \le 0.01$ cuando μ=500?

**Ejercicio 6 (Simulación avanzada)**
Genera 5000 muestras de tamaño n=50 de una distribución Uniforme(0, 1).

- Para cada muestra, calcula la media.
- Calcula la media y varianza de las medias muestrales.
- Compara con los valores teóricos (μ=0.5, σ²=1/12).
- Calcula el porcentaje de medias en el intervalo μ ± 2σ/√n.
- Compara con la regla empírica de la Normal.
- Grafica el histograma con la curva Normal superpuesta.

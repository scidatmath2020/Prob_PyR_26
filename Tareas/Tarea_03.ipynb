# Tarea 03 Cálculo de Probabilidades con R y Python

## Espacio Muestral, Eventos y Eventos Complementarios

Para cada uno de los 5 experimentos:

1. Define el **espacio muestral** $\Omega$ y lista todos sus elementos.
2. Para cada evento:
   - Escríbelo en notación matemática.
   - Escribe su complemento
   - Calcula cuántos elementos tiene el evento y cuántos elementos tiene su complemento

## Experimento 1: Color de automóviles

Se elige al azar un automóvil de un estacionamiento grande y se registra su color. Los colores posibles son: rojo, azul, blanco, negro y plata.

**Evento A:** El auto es de color rojo o azul.  
**Evento B:** El auto NO es de color blanco.

## Experimento 2: Preferencia de bebidas

Se selecciona una persona al azar y se le pregunta: “¿Prefieres café, té o ninguna de las dos?”

**Evento A:** La persona prefiere café.  
**Evento B:** La persona prefiere alguna de las dos bebidas (café o té).

## Experimento 3: Calificación en examen

Un estudiante realiza un examen de 10 preguntas de opción múltiple. La calificación final va de 0 a 10.

**Evento A:** El estudiante aprueba el examen (calificación 6 o mayor).  
**Evento B:** El estudiante obtiene una calificación excelente (9 o 10).

## Experimento 4: Fruta seleccionada

En una canasta hay 4 naranjas, 5 manzanas y 3 plátanos (total 12 frutas). Se saca una fruta al azar.

**Evento A:** Se saca una manzana.  
**Evento B:** Se saca una fruta cítrica (naranja).

## Experimento 5: Estado de un vuelo

Se registra el estado de un vuelo comercial que sale desde el Aeropuerto Internacional de la Ciudad de México. Los posibles resultados son: sale a tiempo, sale con retraso o es cancelado.

**Evento A:** El vuelo sale a tiempo.  
**Evento B:** El vuelo NO es cancelado.

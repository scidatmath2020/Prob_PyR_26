# Problema de Monty Hall

Hay tres puertas: detrás de una hay un coche (premio) y detrás de las otras dos hay cabras (nada). Las ubicaciones son igualmente probables. Tú eliges una puerta. Monty Hall (que sabe dónde está el coche) abre otra puerta que tiene una cabra.

Te pregunta: ¿te quedas con la puerta que elegiste o cambias de puerta?

**¿Cuál es la probabilidad de ganar el coche si cambias de puerta?**

Sean
- $C_i$ el evento "el coche está detrás de la puerta $i$". Entonces $P(C_i)=\frac{1}{3}$.
- $J_i$ el evento "el jugador elige la puerta $i$". Entonces $P(J_i)=\frac{1}{3}$.

Queremos calcular la probabilidad de ganar cambiando después de que Monty abre una puerta. Como Monty siempre abre una puerta con cabra diferente a la elegida por el jugador, definimos el evento de interés como:

**$G$:** evento "el jugador gana el coche cambiando de puerta después de que Monty abre una puerta con cabra". Queremos calcular $P(G)$.

Pero, por Probabilidad Total y regla del producto: 
$$P(G)=\sum_{i,j=1}^3P(G|C_i\cap J_j)P(C_i\cap J_j)=\sum_{i=1}^3\sum_{j=1}^3P(G|C_i\cap J_j)P(C_i)P(J_j)=\frac{1}{9}\sum_{i=1}^3\sum_{j=1}^3P(G|C_i\cap J_j)$$

Ahora, para $P(G|C_i\cap J_j)$, el condicionantes es *el jugador eligió la puerta $j$ y el coche está en la puerta $i$*, de modo que $Monty$ abrirá la puerta $k$, con $k\neq j$, que tiene una cabra. Y el condicionado implica que el jugador cambia a la única puerta restante.

Así, 
- si $i=j$ entonces $P(G|C_i\cap J_j)=0$, pues el jugador estaría cambiando a una puerta que no es la $i$.
- Y si $i\neq j$,  Monty está obligado a abrir la única puerta $k\neq i,j$, que contiene una cabra. Por lo tanto, la única puerta disponible para cambiar es la $i$, que contiene el coche, así que $P(G |C_i\cap J_j) = 1$.

Por lo tanto $$P(G)=\frac{6}{9}=\frac{2}{3}$$

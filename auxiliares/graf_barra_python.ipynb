```python
import numpy as np
import pandas as pd
from plotnine import *
from scipy.stats import *
```

```python
np.random.seed(2026)
simulaciones = binom.rvs(30, 0.8, size=1000)
```

```python
# mi_primer_tabla = data.frame("mis_binomiales"=simulaciones)

mi_primer_tabla = pd.DataFrame({"mis_binomiales":simulaciones})
```

```python
(
ggplot(data=mi_primer_tabla) +
  geom_bar(mapping=aes(x="mis_binomiales"))
)
```

```python
(
ggplot(data = mi_primer_tabla) +
  geom_bar(mapping = aes(x = "factor(mis_binomiales)"),color="red",fill="yellow") +
  scale_x_discrete(
    breaks = sorted(mi_primer_tabla['mis_binomiales'].unique()),
    labels = sorted(mi_primer_tabla['mis_binomiales'].unique())
  ) +
  labs(x = "Valores de mis binomiales",
       y = "Frecuencias")
)
```

```python
#set.seed(2026)
#mis_geometricas = rgeom(1000,0.125) + 1
#mi_primer_tabla["mis_geometricas"] = mis_geometricas

np.random.seed(2026)
mis_geometricas = geom.rvs(0.125, size=1000)
mi_primer_tabla["mis_geometricas"] = mis_geometricas

(
ggplot(data = mi_primer_tabla) +
  geom_bar(mapping=aes(x="mis_geometricas"))
)
```

```r
library(tidyverse)
```

```r
set.seed(2026)
simulaciones = rbinom(1000,30,0.8)
```

```r
mi_primer_tabla = data.frame("mis_binomiales"=simulaciones)
```

```r
ggplot(data = mi_primer_tabla) +
  geom_bar(mapping = aes(x = as.factor(mis_binomiales)),color="red",fill="yellow") +
  scale_x_discrete(
    breaks = sort(unique(mi_primer_tabla$mis_binomiales)),
    labels = sort(unique(mi_primer_tabla$mis_binomiales))
  ) +
  labs(x = "Valores de mis binomiales",
       y = "Frecuencias")
```

```r
set.seed(2026)
mis_geometricas = rgeom(1000,0.125) + 1
mi_primer_tabla["mis_geometricas"] = mis_geometricas
```

```r
ggplot(data = mi_primer_tabla) +
  geom_bar(mapping=aes(x=as.factor(mis_geometricas),y = after_stat(count / sum(count)))) +
  scale_x_discrete(
    breaks = sort(unique(mi_primer_tabla$mis_geometricas)),
    labels = sort(unique(mi_primer_tabla$mis_geometricas))
  )
```

```r
set.seed(2026)
n = 100
dado1 = sample(1:6, n, replace=TRUE)
dado2 = sample(1:6, n, replace=TRUE)

suma = dado1 + dado2

mi_tercera_tabla = data.frame("dado1"=dado1,"dado2"=dado2,"suma"=dado1+dado2)

ggplot(data= mi_tercera_tabla) +
  geom_bar(mapping=aes(x=as.factor(dado1)),fill="red",alpha=0.5,color="black") +
  geom_bar(mapping=aes(x=as.factor(suma)),fill="yellow",alpha=0.5,color="black")
```

```python
# Ejemplo práctico para probabilidad
#Se obtuvieron estos resultados 1, 3, 5, 2, 6, 1, 4, 3, 6, 2, 5, 1
resultados = np.array([1, 3, 5, 2, 6, 1, 4, 3, 6, 2, 5, 1])
resultados
```

```python
#Primer lanzamiento
resultados[0]
```

```python
# Últimos 3 lanzamientos
n = len(resultados)
resultados[n-3:]
```

```python
#Proporción de unos

100 * np.sum(resultados == 1) / len(resultados)
```

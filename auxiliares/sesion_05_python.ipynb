```python
import numpy as np
import pandas as pd
```

```python
# np.arange(a,b) genera los números a,a+1,a+2,...,b-1
#### Lanzar un dado
espacio_muestral = np.arange(1,7)
evento_A = espacio_muestral[espacio_muestral % 2 == 0]
evento_B = espacio_muestral[espacio_muestral >= 5]
evento_C = np.array([6])
evento_Ac = espacio_muestral[espacio_muestral % 2 != 0]

# en R los vectores se construyen con c(elemento1, elemento2, ....). Si solo hay un elemento es c(elemento)
# En Python se construyen con np.array([elemento1, elemento2, ....). Si solo hay un elemento es np.array([elemento])
```

```python
#### tirar un dado 10 veces.

lanzamientos = np.random.choice(espacio_muestral,
       size = 10,
       replace = True, p = np.array([0.5,0.1,0.1,0.2,0.05,0.05])
       )
pd.Series(lanzamientos).value_counts()
```

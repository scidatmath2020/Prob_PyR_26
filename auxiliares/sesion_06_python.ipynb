```python
import numpy as np
import pandas as pd
```

```python
# np.arange(a,b) genera los números a,a+1,a+2,...,b-1
#### Lanzar un dado
espacio_muestral = np.arange(1,7)
evento_A = espacio_muestral[espacio_muestral % 2 == 0]
evento_B = espacio_muestral[espacio_muestral >= 5]
evento_C = np.array([6])
evento_Ac = espacio_muestral[espacio_muestral % 2 != 0]
```

```python
num_lanzamientos = 100000000

lanzamientos = np.random.choice(espacio_muestral,
       size = num_lanzamientos,
       replace = True
      )
lanzamientos
```

```python
### Si yo lanzo el dado num_lanzamiento veces, ¿Cuántas veces se cumple el evento
### evento_Ac?

total_evento_C = np.sum(np.isin(lanzamientos,evento_C))   # en R usamos lanzamientos %in% evento_Ac
total_evento_C
```

```python
### ¿En qué proporción?

total_evento_C / num_lanzamientos
```

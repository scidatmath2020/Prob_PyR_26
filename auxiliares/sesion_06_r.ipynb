```r
#### Lanzar un dado
espacio_muestral = 1:6
evento_A = espacio_muestral[espacio_muestral %% 2 == 0]
evento_B = espacio_muestral[espacio_muestral >= 5]
evento_C = c(6)
evento_Ac = espacio_muestral[espacio_muestral %% 2 != 0]
```

```r
#### tirar un dado 10 veces.
num_lanzamientos = 1000
lanzamientos = sample(espacio_muestral,
       size = num_lanzamientos,
       replace = TRUE,
       )
#lanzamientos
```

```r
### Si yo lanzo el dado num_lanzamiento veces, ¿Cuántas veces se cumple el evento
### evento_Ac?

#lanzamientos %in% evento_Ac   ##### qué elementos de los lanzamientos cayeron en evento_Ac
total_evento_Ac = sum(lanzamientos %in% evento_Ac)
total_evento_Ac
```

```r
### ¿En qué proporción?
total_evento_Ac / num_lanzamientos
```

```r
#### tirar un dado muchas veces pero con caras más pesadas que otras
num_lanzamientos = 100000000
lanzamientos = sample(espacio_muestral,
       size = num_lanzamientos,
       replace = TRUE,
       prob = c(0.8,0.05,0.05,0.05,0.025,0.025)
       )
```

```r
sum(lanzamientos %in% evento_Ac) / num_lanzamientos
```

```r
#### tirar un dado muchas veces pero con caras más pesadas que otras donde los valores de prob no suman 1
num_lanzamientos = 1000000
lanzamientos = sample(espacio_muestral,
       size = num_lanzamientos,
       replace = TRUE,
       prob = c(1,5,7,2,100,5)
       )

sum(lanzamientos %in% evento_Ac) / num_lanzamientos
```

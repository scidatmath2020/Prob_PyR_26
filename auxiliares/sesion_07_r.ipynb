```r
#### Lanzar un dado
espacio_muestral = 1:6
evento_A = espacio_muestral[espacio_muestral %% 2 == 0]
evento_B = espacio_muestral[espacio_muestral %% 3 == 0]
evento_C = c(5)
```

```r
#### tirar un dado 10 veces.
num_lanzamientos = 1000000
lanzamientos = sample(espacio_muestral,
       size = num_lanzamientos,
       replace = TRUE,
       )
```

```r
### P(A)+P(C)

sum(lanzamientos %in% evento_A)/num_lanzamientos + sum(lanzamientos %in% evento_C)/num_lanzamientos

```

```r
espacio_muestral = 1:52
### evento A: obtener un 5
evento_A = c(5,18,31,44) # espacio_muestral %% 13 == 5

### evento B: obtener un 7
evento_B = c(7, 20, 33, 46)

### union A con B
evento_union_AB = c(5,18,31,44,7,20,33,46)
```

```r
# seleccionar la carta

num_selecciones = 100000
selecciones = sample(espacio_muestral,size = num_selecciones,replace=TRUE)
```

```r
## P(A U B)
sum(selecciones %in% evento_union_AB) / num_selecciones
```

```r
## P(A) + P(B)
sum(selecciones %in% evento_A) / num_selecciones + sum(selecciones %in% evento_B) / num_selecciones
```

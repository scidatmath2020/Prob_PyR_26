```r
## Ejercicio 2 (Esperanza y varianza) El peso de los paquetes de café en una fábrica sigue una
## distribución Normal con media 500 g y desviación típica 8 g.

# Define la variable aleatoria X y sus parámetros.
media = 500
desv = 8
varianza = desv**2
```

```r
#Calcula la esperanza y la varianza.
# Respuesta. La esperanza es 500 y la varianza es
varianza
```

```r
# ¿Cuál es la probabilidad de que un paquete pese menos de 490 g?
# P(X<490)

pnorm(490,media,desv)
```

```r
# ¿Entre qué pesos se encuentra el 95% central de los paquetes?
x = qnorm(0.975,media,desv)
c(500-(x-500),x)
```

```r
## Pero aplicando la regla 3 sigma: el 95% central está en media +- 2*desv

media - 2*desv
media + 2*desv
```

```r
# Ejercicio 4 (Estandarización) Las calificaciones de un examen siguen una distribución Normal
# con media 65 y desviación típica 12.

#Estandariza la variable X a Z ~ N(0,1).
# Tenemos Z = (X-mu)/sigma = (X-65)/12
```

```r
# Calcula P(X > 80) usando la estandarización.

# P(X>80)=P(X-65>80-65)=P((X-65)/12 > 15/12)=P(Z>15/12)=1-P(Z<15/12)=1-P(Z<1.25)
1-pnorm(15/12,0,1)
```

```r
1-pnorm(80,65,12)
```

```r
#Calcula P(55 < X < 75) usando la estandarización.
# P(55 < X < 75) = P((55-65)/12 < (X-65)/12 < (75-65)/12) = P(-10/12<Z<10/12)

pnorm(10/12,0,1)-pnorm(-10/12,0,1)
```

```r
pnorm(75,65,12)-pnorm(55,65,12)
```

```r
# Si un estudiante obtuvo 72, ¿cuál es su puntuación Z y qué percentil ocupa?
Z_72 = (72-65)/12
Z_72
```

```r
qnorm(Z_72,0,1)
```

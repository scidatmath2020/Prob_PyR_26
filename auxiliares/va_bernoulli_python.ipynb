```python
import numpy as np
from scipy.stats import binom
```

```python
# El número de experimentos de tipo Bernoulli
n = 1325

# Probabilidad de éxito
p = 0.67

np.random.seed(2026)
simulaciones = binom.rvs(1,p, size=n)
```

```python
simulaciones
```

```python
simulaciones.mean()
```

```python
simulaciones.var()
```

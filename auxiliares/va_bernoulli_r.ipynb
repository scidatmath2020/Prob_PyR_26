```r
# El número de experimentos de tipo Bernoulli
n = 1325

# Probabilidad de éxito
p = 0.67

set.seed(2026)
simulaciones = rbinom(n, size=1, prob=p)
```

```r
simulaciones
```

```r
# Media de la simulación
mean(simulaciones)
```

```r
# Varianza de la simulación
var(simulaciones)
```

```r
# Varianza teorica
p*(1-p)
```

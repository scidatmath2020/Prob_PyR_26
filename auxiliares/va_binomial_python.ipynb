```python
from scipy.stats import binom
import numpy as np
```

```python
# Ejercicio 4 (Aplicación práctica) En una línea de producción,
# la probabilidad de que un artículo sea defectuoso es 0.03.
#Se inspeccionan 50 artículos.
```

```python
# ¿Cuál es la probabilidad de que haya exactamente 2 artículos defectuosos?
# P(X=2)=f(2)

binom.pmf(k=2, n=50, p=0.03)
```

```python
# ¿Cuál es la probabilidad de que haya menos de 3 artículos defectuosos?
# P(X <= 2)

binom.cdf(k=2, n=50, p=0.03)
```

```python
# Supongamos que me piden probabilidad de que tenga al menos 7 defectuosos
# P(X=7)+P(X=8)+...+P(X=50) = P(X >= 7) = 1-P(X<7) = 1-P(X<=6)

1-binom.cdf(k=6, n=50, p=0.03)
```

```python
# Supongamos que me piden probabilidad de que tenga mas de 15 defectuosos
# P(X>15)=1-P(X<=15)

1-binom.cdf(k=15,n=50,p=0.03)
```

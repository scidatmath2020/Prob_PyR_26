```r
# Ejercicio 4 (Aplicación práctica) En una línea de producción,
# la probabilidad de que un artículo sea defectuoso es 0.03.
#Se inspeccionan 50 artículos.
```

```r
# ¿Cuál es la probabilidad de que haya exactamente 2 artículos defectuosos?
# P(X=2)=f(2)

dbinom(2, size=50, prob=0.03)
```

```r
# ¿Cuál es la probabilidad de que haya menos de 3 artículos defectuosos?
# P(X <= 2)

pbinom(2,size=50,prob=0.03) # con esta calcula P(X=0)+P(X=1)+P(X=2)
```

```r
# Supongamos que me piden probabilidad de que tenga al menos 7 defectuosos
# P(X=7)+P(X=8)+...+P(X=50) = P(X >= 7) = 1-P(X<7) = 1-P(X<=6)
1-pbinom(6,size=50,prob=0.03)
```

```r
# Supongamos que me piden probabilidad de que tenga mas de 15 defectuosos
# P(X>15)=1-P(X<=15)

1-pbinom(15,size=50,prob=0.03)
```

```r
# Simula 10,000 repeticiones de lanzar una moneda equilibrada 20 veces
# y cuenta el número de águilas obtenidas.
```

```r
set.seed(2026)
simulacion = rbinom(10000,size=20,prob=0.5)
```

```r
simulacion
```

```r
# Calcula la proporción de simulaciones en las que se obtuvieron
# exactamente 10 águilas.
sum(simulacion == 10)/10000
```

```r
# Compara esta proporción con la probabilidad teórica.
# Me están pidiendo $P(X=10)$

dbinom(10, size=20, prob=0.5)
```

```r
# Calcula la media y la varianza de la simulación y compáralas con los valores teóricos.
mean(simulacion) #simulacion.mean()
var(simulacion) #simulacion.var()
```

```r
# media y varianza teoricas
# E[X]=np, Var(X)=npq

20*0.5
20*0.5*(1-0.5)
```

```python
#Ejericio 1. El tiempo de vida de una lámpara sigue una distribución exponencial con una vida media de 1000 horas.
#Define la variable aleatoria X y su parámetro (tasa λ).

# Me dicen beta = 1000; por lo tanto lambda = 1/1000
beta = 1000
```

```python
#¿Cuál es la probabilidad de que la lámpara dure más de 1200 horas?
# P(X>1200) = 1-P(X<=1200)

1 - expon.cdf(1200, scale=beta)
```

```python
#¿Cuál es la probabilidad de que la lámpara falle antes de las 800 horas?
# P(X<800)

expon.cdf(800, scale=beta)
```

```python
# Ejercicio 3 (Propiedad de falta de memoria) Un sistema ha estado funcionando sin fallos durante 500 horas.
# Su tiempo de vida sigue una exponencial con media 1000 horas (lambda = 1/1000)
# X: tiempo sin fallar

# ¿Cuál es la probabilidad de que falle en las próximas 200 horas?
# P(X<500+200|X>500) = P(X<200)

beta = 1000
expon.cdf(200, scale=beta)
```

```python
# Ejercicio 4. Simula 10,000 valores de una distribución exponencial con λ = 0.02.
# Por ejemplo, estamos diciendo que cada minuto pasan mas o menos 0.02 camiones
# Es decir, 2 camiones por cada 100 minutos, o bien 1 camión cada 50 minutos
#set.seed(2026)
#simulaciones = rexp(10000,0.02)
#simulaciones

np.random.seed(2026)
beta = 1/0.02

simulaciones = expon.rvs(scale=beta, size=10000)
simulaciones
```

```python
# Calcula la proporción de simulaciones con valores mayores a 50.
sum(simulaciones > 50)/10000
```

```python
#Compara con la probabilidad teórica.
# P(X>50) = 1-P(X<50)

1-expon.cdf(50, scale=beta)
```

```python
# Calcula la media y varianza de la simulación y compáralas con los valores teóricos.
simulaciones.mean() # media
```

```python
# varianza
simulaciones.var()
```

```python
# Grafica un histograma de frecuencias relativas y superpone la densidad teórica.

mis_datos = pd.DataFrame({"exponenciales" : simulaciones})

(
ggplot(data=mis_datos) +
  geom_histogram(mapping=aes(x="exponenciales",y="..density.."),bins=50,color="black",fill="yellow") +
  geom_density(mapping=aes(x="exponenciales"))
)
```

```python
from scipy.stats import geom
import numpy as np
```

```python
 # Un jugador ha lanzado un dado 10 veces sin obtener un 6.
 # Si la probabilidad de obtener 6 es 1/6:
 # X es el lanzamiento donde sale 6 la primera vez
p = 1/6
 # ¿Cuál es la probabilidad de obtener el primer 6 en el lanzamiento 15?
 # P(X=15|X>10) = P(X=5)
k = 5
geom.pmf(k, p)
```

```python
# ¿Cuál es la probabilidad de obtener el primer 6 en el lanzamiento 5,
# dado que no ha salido en los 4 primeros lanzamientos?
# P(X=5|X>4) = P(X=1)
k = 1
geom.pmf(k, p)
```

```python
#Ejercicio 5 (Aplicación práctica) En un control de calidad,
# se inspeccionan productos hasta encontrar el primer defectuoso.
# La probabilidad de que un producto sea defectuoso es 0.02.
p = 0.02

# ¿Cuál es la probabilidad de que el primer defectuoso aparezca en el producto número 10?
# P(X=10)
k = 10
geom.pmf(k, p)
```

```python
# ¿Cuál es la probabilidad de que se necesiten más de 15 inspecciones?
# P(X>15) = 1-P(X<=15)
k = 15
1-geom.cdf(k, p)
```

```python
# ¿Cuántos productos se espera inspeccionar en promedio?
1/p
```

```python
# ¿Cuál es la probabilidad de que el primer defectuoso aparezca antes del producto 5?
# P(X<5) = P(X<=4)
k = 4
geom.cdf(k, p)
```

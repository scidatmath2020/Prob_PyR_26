```r
 # Un jugador ha lanzado un dado 10 veces sin obtener un 6.
 # Si la probabilidad de obtener 6 es 1/6:
 # X es el lanzamiento donde sale 6 la primera vez
p = 1/6
 # ¿Cuál es la probabilidad de obtener el primer 6 en el lanzamiento 15?
 # P(X=15|X>10) = P(X=5)
k = 5
dgeom(k-1, p)
```

```r
# ¿Cuál es la probabilidad de obtener el primer 6 en el lanzamiento 5,
# dado que no ha salido en los 4 primeros lanzamientos?
# P(X=5|X>4) = P(X=1)
k = 1
dgeom(k-1, p)
```

```r
n_sim = 10000
p = 1/3
set.seed(2016)
simulaciones = rgeom(n_sim, prob=p) + 1
```

```r
# Calcula la proporción de simulaciones donde se necesitaron exactamente 3 lanzamientos.
sum(simulaciones == 3)/n_sim
```

```r
# Compara esta proporción con la probabilidad teórica.
# P(X=3)
k = 3
dgeom(k-1, prob=p)
```

```r
# Calcula la media y la varianza de la simulación y compáralas con los valores teóricos.
mean(simulaciones)
1/p
var(simulaciones)
(1-p)/p^2
```

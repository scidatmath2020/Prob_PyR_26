```python
from scipy.stats import poisson
import numpy as np
```

```python
# En un centro de llamadas,
#se reciben en promedio 12 llamadas cada 10 minutos.
lamb = 12 # 12 cada 10 MINUTOS
#¿Cuál es la probabilidad de recibir exactamente 10 llamadas en 10 minutos?
# P(X=10)
k = 10
poisson.pmf(k, lamb)
```

```python
# ¿Cuál es la probabilidad de recibir menos de 5 llamadas en 10 minutos?
# P(X<5)=P(X<=4)
k = 4
poisson.cdf(k,lamb)
```

```python
# ¿Cuántas llamadas se esperan en 30 minutos?
lamb_30 = lamb * 3  # ventana es de 30 minutos
lamb_30
```

```python
# ¿Cuál es la probabilidad de recibir exactamente 15 llamadas en 30 minutos?
# P(X=15) con lamb_30
k = 15
poisson.pmf(k,lamb_30)
```

```python
# ¿Cuál es la probabilidad de recibir exactamente 7 llamadas en un minuto?
lamb_1 = lamb / 10
# P(X=7) con lamb_1
k = 7
poisson.pmf(k,lamb_1)
```

```python
# ¿Cuál es la probabilidad de recibir al menos 7 llamadas en un minuto?
# P(X>=7) = 1-P(X<7)=1-P(X<=6)
k = 6
1 - poisson.cdf(k,lamb_1)
```

```python
# ¿Cuál es la probabilidad que de 1pm a 1:17pm se reciban mas de 15 llamadas?
# Ventana: 17 minutos
# 12 por 10 minutos. En 17 minutos: lamb_17 = 17*12/10
lamb_17 = 17*12/10
# P(X>15) = 1-P(X<=15)
k = 15
1 - poisson.cdf(k,lamb_17)
```

```python
### Simulamos el número de accidentes en una autopista durante un día

# Simula 10,000 valores de una Poisson con lambda = 4
n_sim = 10000
lamb = 4
np.random.seed(2026)
simulaciones = poisson.rvs(lamb, size=n_sim)
```

```python
#Calcula la proporción de simulaciones con exactamente 6 ocurrencias.
np.sum(simulaciones == 6) / n_sim
```

```python
#Compara con la probabilidad teórica.
# P(X=6) con lambda = 4
k = 6
poisson.pmf(k,lamb)
```

```python
# Calcula la media y varianza de la simulación y compáralas con los valores teóricos.
simulaciones.mean()
```

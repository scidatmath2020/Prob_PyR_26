```r
# En una biblioteca, el número de personas que entran por hora tiene una media 8 personas por hora.

# Define la variable aleatoria X y su parámetro.
# X total de personas por hora
lamb = 8  # 8 personas POR HORA
```

```r
# ¿Cuál es la probabilidad de que entren exactamente 5 personas en una hora?
# P(X=5)
k = 5
dpois(k, lambda=lamb)  #poisson.pmf(k,lamb)
```

```r
# ¿Cuál es la probabilidad de que entren a lo sumo 3 personas en una hora?
# P(X<=3)
k = 3
ppois(k,lambda=lamb)
```

```r
# ¿Cuál es la probabilidad de que entren al menos 10 personas en una hora?
# P(X>=10) = 1-P(X<10) = 1-P(X<=9)
k = 9
1-ppois(k,lambda=lamb)
```

```r
### Simulamos el número de accidentes en una autopista durante un día

# Simula 10,000 valores de una Poisson con lambda = 4
n_sim = 10000
lamb = 4
set.seed(2026)
simulaciones = rpois(n_sim,lamb)
```

```r
#Calcula la proporción de simulaciones con exactamente 4 ocurrencias.
sum(simulaciones == 4)/n_sim
```

```r
#Compara con la probabilidad teórica.
# P(X=4) con lambda = 4
k = 4
dpois(k,lamb)
```

```r
# Calcula la media y varianza de la simulación y compáralas con los valores teóricos.
mean(simulaciones)
var(simulaciones)
```

```python
import numpy as np
import pandas as pd
from plotnine import *
from scipy.stats import *
```

```python
#Ejercicio 1 (Básico) El tiempo de espera para ser atendido en un banco sigue una distribución uniforme entre 2 y 8 minutos.
a = 2
b = 8
```

```python
#¿Cuál es la probabilidad de esperar menos de 5 minutos?
# P(X<5)

uniform.cdf(5, a,b-a)
```

```python
#¿Cuál es la probabilidad de esperar entre 4 y 6 minutos?
# P(4<X<6) = (6-4)/(8-2) = 2/6 = 1/3 = 0.333
# F(6)-F(4)

uniform.cdf(6, a,b-a) - uniform.cdf(4, a,b-a)
```

```python
# Calcula el tiempo de espera esperado.
(a+b)/2
```

```python
#Ejercicio 3 Simula 10,000 valores de una distribución uniforme en [2,8]

#set.seed(2026)
#simulaciones = runif(10000,2,8)
a = 2
b = 8

np.random.seed(2026)
simulaciones = uniform.rvs(a, b-a, size=10000)
```

```python
#mi_tabla = data.frame("uniformes"=simulaciones)
#K_sturges =  floor(log2(10000))

mi_tabla = pd.DataFrame({"uniformes":simulaciones})

K_sturges =  np.floor(np.log2(10000))
(
ggplot(data=mi_tabla) +
  geom_histogram(mapping=aes(x="uniformes"),color="black",fill="yellow",bins=K_sturges)
)
```

```python
(
ggplot(data=mi_tabla) +
  geom_histogram(mapping=aes(x="uniformes",y="..density.."),color="black",fill="yellow",bins=K_sturges) +
  geom_density(mapping=aes(x="uniformes"),color="red",size=1)
)
```

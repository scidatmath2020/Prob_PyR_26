```r
#Ejercicio 1 (Básico) El tiempo de espera para ser atendido en un banco sigue una distribución uniforme entre 2 y 8 minutos.
a = 2
b = 8
```

```r
#¿Cuál es la probabilidad de esperar menos de 5 minutos?
# P(X<5)
k = 5
punif(k, min=a, max=b)
```

```r
#¿Cuál es la probabilidad de esperar entre 4 y 6 minutos?
# P(4<X<6) = (6-4)/(8-2) = 2/6 = 1/3 = 0.333
# F(6)-F(4)

punif(6, min=a, max=b) - punif(4, min=a, max=b)
```

```r
# Calcula el tiempo de espera esperado.
(a+b)/2
```

```r
#Ejercicio 3 Simula 10,000 valores de una distribución uniforme en [2,8]

set.seed(2026)
simulaciones = runif(10000,2,8)
```

```r
#Calcula la proporción de simulaciones con valores menores a 5.
sum(simulaciones < 5)
```

```r
# Compara con la probabilidad teórica.
# P(X<5)

punif(5,2,8)
```

```r
# Calcula la media y varianza de la simulación y compáralas con los valores teóricos.
mean(simulaciones)
```

```r
# varianza teorica: U(a=2,b=8)
(8-2)^2/12
```

```r
var(simulaciones)
```

```r
# Grafica un histograma de frecuencias relativas.
library(tidyverse)
```

```r
mi_tabla = data.frame("uniformes"=simulaciones)

K_sturges =  floor(log2(10000))

ggplot(data=mi_tabla) +
  geom_histogram(mapping=aes(x=uniformes),color="black",bins=K_sturges)
```

```r
ggplot(data=mi_tabla) +
  geom_histogram(mapping=aes(x=uniformes,y = after_stat(density)),color="black",binwidth = 0.5,boundary=2,fill="yellow") +
  geom_density(mapping=aes(x=uniformes),linewidth=2,color="red")
```

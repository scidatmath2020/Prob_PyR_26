- Número de formas de **ordenar** $n$ objetos diferentes **en una fila**: $n!=1\times2\times3\times...\times n$ si $n\ge 1$ y $0!=1$

**Ejemplo.** ¿De cuántas maneras pueden sentarse 5 personas en una fila de 5 sillas? 

**Solución.** En el primer lugar puedo sentar a cualquiera de las 5. En el siguiente lugar ya solo puedo sentar a las restantes 4 personas. Para el siguiente lugar puedo elegir ya solo entre 3 personas. Para el lugar siguiente, me quedan 2 personas. Y finalmente para el última lugar ya solo tengo una opción. 

La respuesta es $5\times 4\times3\times2\times1=5!$


**Ejemplo.** ¿De cuántas formas se pueden asignar los premios de oro, plata y bronce en una carrera de 8 corredores?

Sin razonamiento, noto que se ajusta precisamente a un problema de tipo permutación de $k$ elementos seleccionados. Por lo tanto $n=8$ y $k=3$ y la respuesta es $$\frac{n!}{(n-k)!}=\frac{8!}{(8-3)!}=\frac{8!}{5!}=\frac{1\times2\times3\times4\times5\times6\times7\times8}{1\times2\times3\times4\times5}=6\times7\times8=336$$

Ahora bien, el total de resultados diferentes de carreras son $8!=40320$

¿Cuántas de ellas las ganó Juan o Pedro?

- $A:$ "Juan ganó"
- $B:$ "Pedro ganó"

Son eventos excluyentes: $A\cap B=\emptyset$. Como la intersección es vacía, entonces el número de formas de realizar $A\cup B$ es igual al total de formas en que gana Juan + el total de formas en que gana Pedro.

Para el evento $A$ hay $7!$ maneras. Para el evento $B$ hay $7!$. Por lo tanto en $7!+7!=2\times 7!=10080$ carreras gana Juan o Pedro 




Por lo tanto la probabilidad de que gane Juan o Pedro, sin mas información, es $\frac{10080}{40320}=0.25$

**Ejemplo.** ¿Cuál es el número de formas de ordenar libro en una repisa si solo caben 5 y en total hay 10 libros?

**Respuesta.** Para el primer lugar tengo 10 opciones; luego 9 opciones; luego 8; luego 7; luego 6. La respuesta es $6\times7\times8\times9\times10=30240$




**Ejemplo.** ¿Cuántos comités de 3 personas pueden formarse a partir de un grupo de 10? El orden dentro del comité no importa. En este caso $n=10$ y $k=3$. Por lo tanto la respuesta es $$\binom{10}{3}=C(10,3)=\mbox{"10 en 3"}=\frac{10!}{3!(10-3)!}=\frac{10!}{3!7!}$$

**Observación.** Por definición $$\binom{n}{k}=\frac{n!}{k!(n-k)!}=\binom{n}{n-k}$$

Además $$\binom{n}{0}=1=\binom{n}{n}$$

**Ejemplo.** ¿De cuántas formas se pueden asignar los premios de oro, plata y bronce en una carrera de 8 corredores?

Yo tengo 8 corredores y quiero seleccionar 3. Una vez que los seleccione, voy a querer ordenar a esos tres para darles su medalla.

La selección de los tres es $\textcolor{red}{\binom{8}{3}}$

Una vez seleccionados, los ordeno: $\textcolor{blue}{3!}$

Por el Principio Multiplicativo, el total es $$\textcolor{red}{\binom{8}{3}}\times\textcolor{blue}{3!}=\textcolor{red}{\frac{8!}{3!5!}}\times\textcolor{blue}{3!}=\frac{8!}{5!}=6\times7\times 8$$

**Observación.** Si no queremos ordenar todos los $n$ objetos, sino solo seleccionar $k$ de ellos y ordenarlos, el número de permutaciones es

- *1er paso:* Seleccionar los $k$ dentro de esos $n$: $\textcolor{red}{\binom{n}{k}}$
- *2do paso:* Una vez que tengo los $k$ seleccionados, los ordeno: $\textcolor{blue}{k!}$

Por el Principio Fundamental del Conteo, el total es
$$\textcolor{red}{\binom{n}{k}}\times\textcolor{blue}{k!}=\textcolor{red}{\frac{n!}{k!(n-k)!}}\times\textcolor{blue}{k!}=\frac{n!}{(n-k)!}=P(n,k),$$

que es exactamente la fórmula de permutación de $k$ elementos tomados de un grupo de $n$

## Póker

- **Escalera real.** ¿Cuántas manos son escalera real? Son 4 en total. Por lo tanto la probabilidad de obtener una escalera real es $\frac{4}{\binom{52}{5}}$


- **Escalera de color.** Hay 36 escaleras de color. Entonces la probabilidad es $\frac{36}{\binom{52}{5}}$

- **Póker.** hay $13\times 48=624$ manos de póker y la probabilidad queda $\frac{13\times 48}{\binom{52}{5}}$

- **Full.** Un ejemplo es "3T, 3C, 3D, 5D,5E".
   - Para calcular la tercia: 13 opciones para el valor y $\binom{4}{3}$ para los palos.
   - Para el par: 12 opciones para el valor y $\binom{4}{2}$

La probabilidad es $$\frac{13\times\binom{4}{3}\times12\times\binom{4}{2}}{\binom{52}{5}}$$

- **Color.** 5 cartas del mismo palo pero que no sean escalera.
    - elijo el palo: 4
    - elijo las 5 cartas: $\binom{13}{5}$
    - resto 10 escaleras de ese palo.

Por lo tanto hay $4\left(\binom{13}{5}-10\right)$

- **Escalera.** 5 cartas consecutivas sin importar el palo (pero que no sean del mismo palo)
  - Selecciono el valor inicial: 10 opciones. Para los palos: $4*4*4*4*4=4^5$. Entonces hay $10\times 4^5$ escaleras EN TOTAL (es decir tanto las que me interesan como las especiales).
 
    Por lo tanto hay $10\times 4^5-36-4=10200$ y la probabilidad es $\frac{10200}{\binom{52}{5}}$

**Permutaciones con repetición**

¿Cuántas palabras distintas pueden formarse reordenando las letras de la palabra "MISSISSIPPI"?

- MISSISSIPPI
- MSISISSIPPI

- $\mbox{MI\textcolor{red}{S}\textcolor{blue}{S}}ISSIPPI$
- $\mbox{MI\textcolor{blue}{S}\textcolor{red}{S}}ISSIPPI$

Lo que me dice la fórmula $$\frac{n!}{n_1!n_2!n_3!...n_k!}$$ donde $n_1,n_2,...,n_k$ es que

$n_1=\mbox{¿Cuántas veces está la M?}=1$ 

$n_2=\mbox{¿Cuántas veces está la I?}=4$

$n_3=\mbox{¿Cuántas veces está la S?}=4$

$n_4=\mbox{¿Cuántas veces está la P?}=2$

Aplicando la fórmula: $$\frac{(1+4+4+2)!}{1!4!4!2!}=\frac{11!}{4!^22!}=34650$$

Pero mejor razonemos:
- Tengo 11 lugares: $-----------$ y quiero llenar esos con lugares con las 11 letras.



¿Cuántas M tengo? 1. Así que selecciono el lugar para colocar esa letra: $\binom{11}{1}$

Por lo tanto ya solo me sobran 10 lugares. Voy a colocar las 4 I's. Selecciono lo cuatro lugares: $\binom{10}{4}$.

Por lo tanto ya solo me sobran 6 lugares. Voy a colocar las 4 S's. Selecciono lo cuatro lugares: $\binom{6}{4}$.

Por lo tanto ya solo me sobran 2 lugares. Voy a colocar las 2 P's. Selecciono los dos lugares: $\binom{2}{2}$.

La respuesta es $\binom{11}{1}\binom{10}{4}\binom{6}{4}\binom{2}{2}$

\begin{eqnarray*}
&&\binom{11}{1}\binom{10}{4}\binom{6}{4}\binom{2}{2}\\
&=&\frac{11!}{1!10!}\times\frac{10!}{4!6!}\times\frac{6!}{4!2!}\times\frac{2!}{2!0!}\\
&=&\frac{11!}{1!4!4!2!}
\end{eqnarray*}

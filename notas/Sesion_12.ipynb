**Ejemplo.** En una fábrica, el 5% de los productos tiene defecto grave $G$, el 10% tiene defecto leve $L$ y el 2% tiene ambos. Si un producto tiene defecto leve, ¿cuál es la probabilidad de que también tenga defecto grave?

Me están preguntando $P(\mbox{explote}|\mbox{está mal pintado})=\frac{P(\mbox{explote y este mal pintado})}{P(\mbox{este mal pintado})}=\frac{0.02}{0.1}=0.2$

**Ejemplo 1: Extracción sin reposición**

Una caja contiene 5 bolas rojas y 3 azules. Se extraen dos bolas sin reposición. ¿Cuál es la probabilidad de que ambas sean rojas?

- $A$: la primera es roja
- $B$: la segunda es roja

$P(A\cap B)=\textcolor{red}{P(B|A)}\textcolor{blue}{P(A)}=\textcolor{red}{\frac{4}{7}}\times\textcolor{blue}{\frac{5}{8}}=\frac{20}{56}$




**Ejemplo 2: Componentes electrónicos**

Un sistema tiene dos componentes en serie (ambos deben funcionar para que el sistema funcione). El componente A tiene probabilidad 0.95 de funcionar. Si A funciona, el componente B tiene probabilidad 0.90 de funcionar. ¿Cuál es la probabilidad de que el sistema funcione?

Me están preguntando $$P(\mbox{A funcione y la $B$ funcione})=\textcolor{red}{P(\mbox{B funcione}|\mbox{A funciona})}\textcolor{blue}{P(\mbox{A funciona})}=\textcolor{red}{0.90}\times\textcolor{blue}{0.95}$$

**Ejemplo 3: Encuesta de opinión**

En una ciudad, el 60% de los adultos leen el periódico. Entre los que leen el periódico, el 70% está informado sobre política. Entre los que no leen, solo el 30% está informado. ¿Cuál es la probabilidad de que una persona al azar lea el periódico y esté informada?

**Solución.** 

- $L$: leer el periódico
- $I$: estar informado

Me están pidiendo $P(L\cap I)=P(L|I)P(I)=P(I|L)P(L)=0.7\times 0.6=0.42$

**Ejemplo 1: Línea de producción con tres etapas**

Una línea de producción tiene tres etapas en serie. La probabilidad de que la etapa 1 falle es 0.10. Si la etapa 1 funciona, la etapa 2 falla con probabilidad 0.05. Si las dos primeras funcionan, la etapa 3 falla con probabilidad 0.02. ¿Cuál es la probabilidad de que el proceso completo funcione?

**Solución.** Para que el sistema funcione completo, se requiere que las tres etapas funcionen.

- $A$: la etapa 1 funciona; entonces sabemos que $P(A)=1-P(A^c)=1-0.10=0.9$
- $B$: la etapa 2 funciona; entonces sabemos que $P(B^c|A)=0.05$ pero $P(B|A)=1-P(B^c|A)=1-0.05=0.95$ 
- $C$: la etapa 3 funciona; entonces sabemos que $P(C^c|A\cap B)=0.02$ pero $P(C|A\cap B)=1-P(C^c|A\cap B)=1-0.02=0.98$


Por lo tanto me piden $$P(A\cap B\cap C)=P(A)P(B|A)P(C|A\cap B)=0.9\times0.95\times0.98=0.8379$$

**Ejemplo 4: Extracción sin reposición de tres objetos**

Una urna contiene 7 bolas blancas y 3 negras. Se extraen tres bolas sin reposición. ¿Cuál es la probabilidad de que las tres sean blancas?

**Solución.**

- $A$: la primera es blanca
- $B$: la segunda es blanca
- $C$: la tercera es blanca

y me preguntan $P(A\cap B\cap C)$

En *abstracto*, tenemos $$P(algo\cap otro\cap alguien)=P(algo)P(otro|algo)P(alguien|algo\cap otro)$$

$P(A\cap B\cap C)=\textcolor{blue}{P(A)}\textcolor{red}{P(B|A)}\textcolor{purple}{P(C|A\cap B)}$

$P(A\cap B\cap C)=P(C)P(A|C)P(B|A\cap C)$

Utilizando la primera, tenemos $$P(A\cap B\cap C)=\textcolor{blue}{\frac{7}{10}}\times\textcolor{red}{\frac{6}{9}}\times\textcolor{purple}{\frac{5}{8}}=0.2916666$$

**Independencia estocástica.** La regla del producto me dice que $P(A\cap B)=P(A|B)P(B)$

Si tomo como definición de independencia que $P(A\cap B)=P(A)P(B)$, entonces $P(A)P(B)=P(A|B)P(B)$

Si tomo como definición de independencia que $P(A|B)=P(A)$, entonces $P(A\cap B)=P(A)P(B)$

**¿Qué pasa si $A\subseteq B$?**

Si fueran independientes, entonces, por una parte: $P(A\cap B)=P(A)P(B)$. Pero por otra parte $P(A\cap B)=P(A)$. Por lo tanto $P(A)=P(A)P(B)$ y entonces $P(A)=0$ o $P(B)=1$

**¿Qué pasa si $A\cap B=\emptyset$?**

$0=P(A\cap B)=P(A)P(B)$ por lo tanto $P(A)=0$ o $P(B)=0$

Equivalentemente, si $P(A)>0$ y $P(B)>0$ y son mutuamente excluyentes, entonces **no pueden ser independientes**

**Lanzar una moneda dos veces**

Lanzamos una moneda dos veces. Sea 
- $A$ = "cara en el primer lanzamiento" y 
- $B$ = "cara en el segundo lanzamiento".

¿Son independientes?


**Solución.** El espacio muestral son las parejas $\Omega=\{(Ca,Ca),(Ca,Cr),(Cr,Ca),(Cr,Cr)\}$

Entonces $P(A\cap B)=\frac{1}{4}$

Y, por otra parte, $P(A)=\frac{1}{2}$ y $P(B)=\frac{1}{2}$. Como $P(A)P(B)=\frac{1}{2}\times\frac{1}{2}=\frac{1}{4}$, entonces $P(A\cap B)=P(A)P(B)$ así que hay independencia

**Ejemplo 7: Estado civil y género (independencia en poblaciones grandes)**

$P(B|A)=0.4$

**Independencia de varios eventos**

En general, $A_1$, $A_2$, ..., $A_k$ son independientes si $$P(A_1\cap A_2\cap A_3\cap...\cap A_k)=P(A_1)P(A_2)P(A_3)...P(A_k)=\prod_{i=1}^k P(A_i)$$




**Ejemplo 1: Caso engañoso con monedas**

Lanzamos dos monedas. Definimos:

- $A$: la primera muestra cara
- $B$: la segunda muestra cara
- $C$: ambas muestran el mismo resultado

Los eventos son independientes por pares:

- $A$ y $B$ son independientes
- $A$ y $C$ son independientes: $P(C|A)=\frac{1}{2}$ y $P(C)=\frac{2}{4}=\frac{1}{2}$; entonces son independientes.
- $B$ y $C$ son independientes: $P(C|B)=\mbox{ cuántos cumplen C sabiendo que se cumple B}=\frac{1}{2}$


Pero $P(A\cap B\cap C)=\frac{1}{4}$

Sin embargo, $P(A)=\frac{1}{2}$, $P(B)=\frac{1}{2}$ y $P(C)=\frac{1}{2}$ así que $P(A)P(B)P(C)=\frac{1}{8}$

Por lo tanto **los eventos $A$, $B$ y $C$ no son independientes**

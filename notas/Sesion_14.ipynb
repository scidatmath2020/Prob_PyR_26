## Probabilidad Total

Tengo la partición $B_1,B_2,...,B_k$. Entonces $\Omega = B_1\cup B_2\cup...\cup B_k$
\begin{eqnarray*}
&&P(A)\\
&=&P(A\cap \Omega)\\
&=&P(A\cap (\textcolor{red}{B_1\cup B_2\cup...\cup B_k}))\\
&=&P\left((A\cap B_1)\bigcup(A\cap B_2)\bigcup...\bigcup (A\cap B_k)\right)\\
&=&P(A\cap B_1)+P(A\cap B_2)+...+P(A\cap B_k)\\
&=&P(A|B_1)P(B_1)+P(A|B_2)P(B_2)+...+P(A|B_k)P(B_k)
\end{eqnarray*}



**Ejemplo 1: Control de calidad con dos máquinas**

Una fábrica tiene dos máquinas que producen el mismo componente. La máquina 1 produce el 60% de la producción total, con un 5% de defectuosos. La máquina 2 produce el 40% restante, con un 3% de defectuosos. Si tomamos un componente al azar de toda la producción, ¿cuál es la probabilidad de que sea defectuoso?

- $D$: "el producto es defectuoso." Entonces me están pididendo $P(D)$

- Sea $B_1$: "el producto es de la máquina 1"
- Sea $B_2$: "el producto es de la máquina 2"

$$P(D)=\textcolor{red}{P(D|B_1)}\textcolor{blue}{P(B_1)}+\textcolor{purple}{P(D|B_2)}\textcolor{darkgreen}{P(B_2)}$$

- $\textcolor{blue}{P(B_1)}=0.6$
- $\textcolor{darkgreen}{P(B_2)}=0.4$
- $\textcolor{red}{P(D|B_1)}=0.05$
- $\textcolor{purple}{P(D|B_2)}=0.03$

**Ejemplo 2: Urnas con canicas**

Tenemos tres urnas. La urna 1 contiene 3 canicas rojas y 7 azules. La urna 2 contiene 6 rojas y 4 azules. La urna 3 contiene 8 rojas y 2 azules. Se selecciona una urna al azar (cada una con probabilidad $1/3$) y luego se extrae una canica. ¿Cuál es la probabilidad de que sea roja?

- Sea $B_i$ el evento "se selecciona la urna $i$". Nos dicen que $P(B_i)=\frac{1}{3}$
- Sea $R$ el evento "la pelota es roja". Nos piden $P(R)$

$$P(R)=P(R|B_1)P(B_1)+P(R|B_2)P(B_2)+P(R|B_3)P(B_3)$$

$$=\frac{1}{3}\left(\frac{3}{10}+\frac{6}{10}+\frac{8}{10}\right)=\frac{17}{30}=0.5666$$

**Aplicación de Probabilidad Total: Diagnóstico Médico**

Una prueba para una enfermedad tiene 99% de sensibilidad y 98% de especificidad. La prevalencia de la enfermedad en la población es 0.1%. ¿Cuál es la probabilidad de que una persona seleccionada al azar dé positivo?

- **Sensibilidad:** probabilidad de que la prueba sea *positivo* cuando la persona *sí está enferma*
- **Especificidad:** probabilidad de que la prueba sea *negativa* cuando la persona *no está enferma*.
- **Prevalencia:** porcentaje de la población que tiene la enfermedad

Sea **+** el evento "la prueba de la persona dá positivo". Entonces me piden $P(+)$ 

$$P(+)=P(+|\mbox{enfermo})P(\mbox{enfermo})+P(+|\mbox{no enfermo})P(\mbox{no enfermo})$$

- $P(+|\mbox{enfermo})=0.99$
- $P(\mbox{enfermo})=0.001$
- $P(\mbox{no enfermo})=1-0.001=0.999$
- $P(+|\mbox{no enfermo})=1-P(-|\mbox{no enfermo})=1-0.98=0.02$: recordemos que si $A$ y $B$ son eventos entonces $P(A|B)+P(A^c|B)=1$

$$P(+)=0.99\times0.001+0.02\times0.999=0.0209$$

Es decir, el 2% de las veces la prueba dará positiva sin importar si el paciente está o no enfermo

**Teorema de Bayes**

Si $E_1,E_2,...,E_k$ forman una partición de $\Omega$, y $A$ es un evento entonces $$P(E_i|A)=\frac{P(E_i)P(A|E_i)}{P(A)}$$



En particular, $$\textcolor{blue}{P(B|A)}=\frac{\textcolor{red}{P(A|B)}\textcolor{purple}{P(B)}}{P(A)}$$

- $\textcolor{blue}{P(B|A)}$: probabilidad de $B$ a posteriori
- $\textcolor{purple}{P(B)}$: probabilidad de $B$ a priori
- $\textcolor{red}{P(A|B)}$: verosimilitud

**Principios de Estadística Bayesiana**

$$P(B|A)\propto P(A|B)P(B)$$

**Ejemplo 1: SciData - Helado y pastel**

En SciData, la probabilidad de que a un alumno seleccionado al azar le guste el helado es del 60%, mientras que la probabilidad de que a un alumno le guste el pastel es del 36%. Además, se sabe que la probabilidad de que a un alumno le guste el pastel dado que le gusta el helado es del 40%. Calcular la probabilidad de que a un alumno le guste el helado, dado que le gusta el pastel.

- $H$: "le gusta el helado": $P(H)=0.6$
- $A$: "le gusta el pastel": $P(A)=0.36$
- $P(A|H)=0.4$

Me preguntan $P(H|A)$

$$P(H|A)=\frac{P(A|H)P(H)}{P(A)}=\frac{0.4\times 0.6}{0.36}=0.666$$

**Ejemplo 2: Urnas invertido**

Usando el ejemplo de las urnas con canicas (urna 1: 3R,7A; urna 2: 6R,4A; urna 3: 8R,2A). Supongamos que se selecciona una urna al azar, se extrae una canica y resulta ser roja. ¿Cuál es la probabilidad de que provenga de la urna 3?



- $R$: "la pelota es roja"
- $E_i$: "se eligió la caja $i$". Entonces $P(E_i)=\frac{1}{3}$

Me están preguntando $P(E_3|R)$

$$P(E_3|R)=\frac{P(R|E_3)P(E_3)}{P(R)}$$

- $P(E_3)=\frac{1}{3}$
- $P(R|E_3)=\frac{8}{10}$
- $P(R)=\frac{17}{30}$

$$P(E_3|R)=\frac{8/10\times 1/3}{17/30}=0.470$$

**Si yo elijo una urna y una pelota 100 veces, en aproximadamente 56 de esas veces voy a obtener una roja. Y de esas 56, en aproximadamente 27 o 26 esa bola roja proviene de la urna 3**

$$P(E_1|R)=\frac{P(R|E_1)P(E_1)}{P(R)}=\frac{3/10\times 1/3}{17/30}=0.1764$$

$$P(E_2|R)=\frac{P(R|E_2)P(E_2)}{P(R)}=\frac{6/10\times 1/3}{17/30}=0.3529$$

**El ejemplo médico completo (Bayes)**

Retomemos el ejemplo de la prueba médica. Una prueba tiene 99% de sensibilidad y 98% de especificidad. La prevalencia es 0.1%. Una persona da positivo. ¿Cuál es la probabilidad de que realmente tenga la enfermedad?

$$P(E|+)=\frac{P(+|E)P(E)}{P(+)}=\frac{0.99\times0.001}{0.02}=0.0495$$




**Ejemplo 1: Dos pruebas médicas (prueba del 99%/98% con prevalencia 0.1%)**

La misma persona del ejemplo anterior da positivo en una segunda prueba independiente (con las mismas características). ¿Cuál es ahora su probabilidad de estar enferma?

$$P(E|+_2)=\frac{P(+_2|E)\textcolor{red}{P(E)}}{P(+_2)}$$

$$P(+_2)=P(+_2|E)P(E)+P(+_2|E^c)P(E^c)$$ donde $P(E)$ es la probabilidad a posteriori del paso anterior. Por lo tanto:

- $P(+_2|E)=0.99$
- $P(E)=0.0495$
- $P(+_2|E^c)=1-P(-_2|E^c)=1-especificidad=1-0.98=0.02$
- $P(E^c)=1-P(E)=1-0.0495$

$$P(+_2)=\frac{0.99\times 0.0495}{0.99\times0.0495+0.02\times0.95}=0.720$$

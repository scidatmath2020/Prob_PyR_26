**Esperanza matemática**

Imaginemos que lanzamos un dado equilibrado.

Si sale 4, ganas \\$5. Si sale 1 ganas \\$2. En cualquier otro caso, pierdes \\$1.5


¿Cuánto espero ganar (o perder)?

Sea $X$ la variable aleatoria que me dice cuánto gano en el turno.

¿Qué valores puede tomar $X$?

$X$ puede ser 5, 2 o puede ser -1.5

¿Con qué probabilidades?

$P(X=5)=1/6=5/30$

$P(X=2)=1/6=5/30$

$P(X=-1.5)=4/6=20/30$

$5\times\frac{5}{30}+2\times\frac{5}{30}-1.5\times\frac{20}{30}=0.1666$

Si yo soy el organizador, cuánto me conviene cobrar por turno?

Si cobrara \\$0.5 por turno, ya gané

Básicamente lo que hice fue $5\times P(X=5)+2\times P(X=2)-1.5\times P(X=-1.5)$

**Fútbol**

El mundial de Italia 1990. Este lo ganó Alemania en la final contra Argentina.

Este es un mundial muy mal recordado: casi no hubo goles. De hecho en la final únicamente hubo dos disparos a portería (los dos de Alemania)

En esos años habían tres fenómenos:

1. El fuera de lugar.

2. El juego del portero

3. Puntaje en etapa de grupos

**Puntaje actualmente**

- Si ganas el partido, obtienes 3 puntos. 
- Si empatas el partido, obtienes 1 punto.
- Si pierdes el partido, te dan 0 puntos.

**Puntaje en Italia 90 y antes**

- Si ganas el partido, obtienes 2 puntos. 
- Si empatas el partido, obtienes 1 punto.
- Si pierdes el partido, te dan 0 puntos.

¿Qué sucedía antes?

Digamos que estás jugando y vas empatado contra el otro equipo en el partido.

Vamos a suponer que eres muy malo atacando pero muy bueno defendiendo.

Para fijar ideas, digamos que:
- Estrategia A: si buscas el empate tienes el 90% de probabilidad de éxito. Y por lo tanto, el 10% pierdes.

- Estrategia B: Si busco atacar, tengo el 40% de probabilidad de éxito (y por lo tanto tengo el 60% de probabilidad de que me ganen)

¿Cuántos puntos espero ganar en cada una de las estrategias?

En Italia 90 y antes:

- Siguiendo estrategia A: sé que en 9 de cada 10 partidos, si me dedico a defender, terminaremos empatados. Por lo tanto, bajo esta estrategia, $$\mbox{espero tener }=1\times0.9+0\times 0.1=0.9$$




- Siguiendo estrategia B: sé que en 4 de cada 10 partidos, si me dedico a atacar, terminaremos ganando. Por lo tanto, bajo esta estrategia, $$\mbox{espero tener }=2\times 0.4+0\times0.6=0.8$$

Bajo este esquema, me conviene la estrategia A: jugar al empate.

## Esperanza de una variable aleatoria $E[X]$

- Si $X$ es discreta, entonces $$E[X]=\sum x\times P(X=x)=\sum x\times f(x)$$

En el caso del lanzamiento del dado:
- $X$ puede ser 5, 2 o -1.5
- y $P(X=5)=1/6$, $P(X=2)=1/6$ y $P(X=-1.5)=4/6$ lo cual sucede cuando el dado dice $2,3,5,6$


Y por lo tanto $E[X]=5\times P(X=5)+2\times P(X=2)-1.5\times P(X=-1.5)=0.1666$

- Si $X$ es continua, entonces $$E[X]=\int_{-\infty}^\infty xf(x)\,\mathrm{d}x $$

**Importante.** La esperanza es un número que representa lo que se espera que valga $X$ si realizo el experimento muchas veces

**Esperanza matemática**

Imaginemos que lanzamos un dado equilibrado.

Si sale 4, ganas \\$5. Si sale 1 ganas \\$2. En cualquier otro caso, pierdes \\$1.5


¿Cuánto espero ganar (o perder)?

Sea $X$ la variable aleatoria que me dice cuánto gano en el turno.

¿Qué valores puede tomar $X$?

$X$ puede ser 5, 2 o puede ser -1.5

¿Con qué probabilidades?

$P(X=5)=1/6=5/30$

$P(X=2)=1/6=5/30$

$P(X=-1.5)=4/6=20/30$

$5\times\frac{5}{30}+2\times\frac{5}{30}-1.5\times\frac{20}{30}=0.1666$

Si yo soy el organizador, cuánto me conviene cobrar por turno?

Si cobrara \\$0.5 por turno, ya gané

Básicamente lo que hice fue $5\times P(X=5)+2\times P(X=2)-1.5\times P(X=-1.5)$

Bajo este esquema, me conviene la estrategia A: jugar al empate.

## Esperanza de una variable aleatoria $E[X]$

- Si $X$ es discreta, entonces $$E[X]=\sum x\times P(X=x)=\sum x\times f(x)$$

En el caso del lanzamiento del dado:
- $X$ puede ser 5, 2 o -1.5
- y $P(X=5)=1/6$, $P(X=2)=1/6$ y $P(X=-1.5)=4/6$ lo cual sucede cuando el dado dice $2,3,5,6$


Y por lo tanto $E[X]=5\times P(X=5)+2\times P(X=2)-1.5\times P(X=-1.5)=0.1666$

- Si $X$ es continua, entonces $$E[X]=\int_{-\infty}^\infty xf(x)\,\mathrm{d}x $$

**Importante.** La esperanza es un número que representa lo que se espera que valga $X$ si realizo el experimento muchas veces

## Esperanza como centro de masa



Imaginemos que $X$ es continua y tiene función de densidad $$f(x)=\frac{1}{\sqrt{2\pi}}e^{-x^2/2}$$

**Ejercicio**

Sea $f(x)=\left\{\begin{array}{l}\textcolor{red}{C}x(6-x)\mbox{ si }2\le x\le 5\\0\mbox{ en otro caso}\end{array}\right.$


1. Encontrar el valor de $\textcolor{red}{C}$ para que $f$ sea una función de densidad

**R.** Quiero que $$\int_{-\infty}^\infty f(x)\,\mathrm{d}x=1$$

Si $x<2$, entonces $f(x)=0$. También para $x>5$ tenemos $f(x)=0$

Así, $$\int_{-\infty}^\infty f(x)\,\mathrm{d}x=\int_{-\infty}^2 f(x)\,\mathrm{d}x+\int_{2}^5 f(x)\,\mathrm{d}x+\int_{5}^\infty f(x)\,\mathrm{d}x$$

Pero $$\int_{2}^5 f(x)\,\mathrm{d}x=\int_{2}^5 \textcolor{red}{C}x(6-x)\,\mathrm{d}x=C\int_{2}^5 x(6-x)\,\mathrm{d}x=24C=1$$

Por lo tanto $C=\frac{1}{24}$

2. ¿Cuál es la esperanza?

$$E[X]=\int_{-\infty}^\infty xf(x)\,\mathrm{d}x=\int_{2}^5xf(x)\,\mathrm{d}x=\int_{2}^5x\times \frac{x(6-x)}{24}\,\mathrm{d}x=\frac{109}{32}$$

**Ejemplo del dado**

Se lanzan dos dados y $X$ es el número de caras. Calcula $E[X]$

$P(X=0)=1/4$, $P(X=1)=1/2$ y $P(X=2)=1/4$

$$E[X]=0\times P(X=0)+1\times P(X=1)+2\times P(X=2)=\frac{1}{2}+\frac{2}{4}=1$$

**Dado con apuesta** 

Supongamos que jugamos el siguiente juego: lanzamos un dado. Si sale 6, ganamos 
5. Si sale cualquier otro número, perdemos $2.

¿Cuál es nuestra ganancia esperada?

Si $X$ es la ganancia, entonces $P(X=5)=1/6$ y $P(X=-2)=5/6$

$$E[X]=5\times P(X=5)-2\times P(X=-2)=\frac{5}{6}-\frac{10}{6}=-\frac{5}{6}=-0.83$$

**Me pregunto cuánto es $E[X^3]$?**

$$E[X^3]=5^3\times P(X=5)+(-2)^3P(X=-2)=125\times \frac{1}{6}-8\times\frac{5}{6}$$

## Cálculo puntual de parámetros en Estadística

Calcular $E[X^n]$ donde $n \ge 0$: $$E[X^n]=\int_{-\infty}^\infty x^nf(x)\mathrm{d}x$$

## Propiedades importantes:

- $E[aX+b]=aE[X]+b$. En particular, $E[c]=c$ y $E[aX]=aE[X]$
- $E[X+Y]=E[X]+E[Y]$. Por lo tanto $E[X-Y]=E[X]-E[Y]$ ($E[X-Y]=E[X+(-Y)]=E[X]+E[-Y]$)

## Varianza

Sea $X$ una variable aleatoria. Entonces $E[X]$ es el valor esperado.

$$Var(X)=E[(X-E[X])^2]$$

**En el ejemplo del dado:**

Aquí, $E[X]=-0.83$ y me estaría fijando en $(X-(-0.3))^2$


$E[X-\textcolor{red}{E[X]}]=E[X]\textcolor{red}{-E[X]}=0$

**Dato para matemáticos**

$$E[X]=\int_{\Omega} X\mathrm{d}P$$



En $L^2(\Omega,\mathcal{F},P)$ tienes el producto interno $\langle X,Y\rangle =\int_\Omega XY dP=E[XY]$

$\|X\|^2=\int X^2\mathrm{d}P$

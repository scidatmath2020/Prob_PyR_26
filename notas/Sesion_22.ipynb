## Bernoulli

Si $X\sim Ber(p)$, entonces
- $E[X]=\sum xP(X=x)=0P(X=0)+1P(X=1)=p$: 

**Ejemplo**. Si tengo una moneda donde $P(\mbox{águila})=0.8$, y defino *éxito* como "obtener águila", el hecho de que $E[X]=p$ me está diciendo que si lanzo MUCHAS veces esa moneda, entonces al rededor del 80% de las veces obtendré águila.

- $Var(X)=E[X^2]-E[X]^2=\textcolor{red}{E[X^2]}-p^2$

$E[X^2]=\sum x^2P(X=x)=0^2P(X=0)+1^2P(X=1)=p$

Por lo tanto $Var(X)=p-p^2=p(1-p)=pq$

- $sd(X)=\sqrt{p(1-p)}$

Notemos que $sd(X)=\sqrt{p(1-p)}\le \frac{p+1-p}{2}=\frac{1}{2}$

## Concepto de semilla en simulaciones

El concepto de semilla es un concepto propio de la Informática. La idea es la siguiente:

Todas las computadoras tienen un cierto rango de números enteros para contar, digamos desde 0 hasta N, donde N es un número gigantezco.




Ahora, digamos que quieres generar $n$ números aleatorios. Vas a tener que existe una primer manera de generarlos, una segunda, una tercera, etc...

**Ejercicio 2 (Bernoulli)**

Un votante seleccionado al azar tiene probabilidad p = 0.65 de votar por el candidato A. Calcula E[X] y Var(X). Interpreta: ¿cuántos votos esperaría obtener el candidato A si se encuestan 200 personas similares?

Primero, al ser Bernoulli, $E[X]=p=0.65$ y $Var(X)=pq=0.65\times 0.35=0.2275$ (y entonces $sd(X)=0.47)$. Finalmente, en un grupo de 200 personas esperamos que el 65% voten por A. $200*0.65=130$

# Uniforme discreta

- Si $X$ es uniforme discreta y puede tomar los valores 5,6,7,8, entonces lo denotamos por $X\sim U\{5,6,7,8\}$
- En el dado usual. Ahí $X$ puede valer 1,2,3,4,5,6 y por lo tanto $X\sim U\{1,2,3,4,5,6\}$. Aquí tenemos $a=1$ y $b=6$

Y por lo tanto, para cualquier valor $x=1,2,3,4,5,6$  es $$P(X=x)=\frac{1}{b-a+1}=\frac{1}{6-1+1}=\frac{1}{6}$$

**Esperanza**

Si $X\sim U\{a,a+1,...b\}$ entonces $$E[X]=\sum xP(X=x)=aP(X=a)+(a+1)P(X=a+1)+(a+2)P(X=a+2)+...+bP(X=b)=\frac{1}{n}(a+a+1+a+2+...+b)=\frac{1}{n}\times \frac{(a+b)n}{2}$$

**Ruleta** En el caso de la ruleta, $a=0$ y $b=36$ y $n=37$ y aquí $E[X]=\frac{0+36}{2}=18$ y $Var(X)=\frac{n^2-1}{12}=\frac{37^2-1}{12}=114$ y $sd(X)=10.67$

- Se lanza un dado justo de 6 caras numeradas del 1 al 6. Entonces calcular $P(X\le 4)=\frac{4}{6}$

**Ejercicio 4 (Aplicación práctica)** En una línea de producción, la probabilidad de que un artículo sea defectuoso es 0.03. Se inspeccionan 50 artículos.

- ¿Cuál es la probabilidad de que haya exactamente 2 artículos defectuosos?
- ¿Cuál es la probabilidad de que haya menos de 3 artículos defectuosos?
- ¿Cuántos artículos defectuosos se esperan en promedio?

Sea $X$ el número de artículos defectuosos en el lote de 50. Por lo tanto $n=50$ y $p=0.03$. Además $X\sim Bin(n=50,p=0.03)$

- Cuál es la probabilidad de que haya exactamente 2 artículos defectuosos?

**R.** Me piden $P(X=2)=\binom{50}{2}\times 0.03^2(1-0.03)^{50-2}=0.255518202062405$
  

- ¿Cuál es la probabilidad de que haya menos de 3 artículos defectuosos?

**R.** Me piden $P(X\le 2)=0.810798075369721$

- ¿Cuántos artículos defectuosos se esperan en promedio?

**R.** Me piden $E[X]=np=50\times0.03=1.5$

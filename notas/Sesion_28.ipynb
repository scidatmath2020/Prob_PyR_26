# Cuartiles de una v.a. continua

Primero, observemos la relación entre las gráficas de la función de densidad $f(x)$ y la función de distribución $F(x)$. Gráficamente, $F(x)$ representa el área desde $-\infty$ hasta $x$ que hay debajo de $f(x)$

![imagenes](im010.png)

De lo anterior, surge la siguiente pregunta: **dado un valor $0\le c\le 1$, quién es $x$ para que se cumpla $P(X\le x)=c$**. Es decir, resolver la ecuación $F(x)=c$. A este problema se le llama **cálculo de los cuantiles**

Notemos que usualmente, desde el inicio del tema de v.a., siempre nos preguntamos por $P(X\le x)$

**Ejemplo**

Sea $X\sim U(1.25,2)$. Si $c=0.732$, quiero encontrar $x$ tal que $P(X\le x)=0.732$.

**Solución**

Como $f(x)=\frac{1}{2-1.25}=\frac{1}{0.75}$ para $x\in[1.25,2]$, entonces $$0.732 = P(X\le x) = F(x) =(x-1.25)\cdot \frac{1}{0.75}$$

Entonces $x=1.799$

Esto significa que $P(X\le 1.799)=0.732$







Este problema se resuelve con las funciones `qalgo` (de R) y `algo.ppf` (de Python).

```r
################# En R
qunif(0.732, min=1.25, max=2)
```

```python
################# En Python
uniform.ppf(0.732, 1.25, 2-1.25)
```

Si $X$ es una variable aleatoria y $x$ es un número real, puedo preguntarme por $P(X\le x)$

- Si $X$ es el número de caras cuando lanzo una moneda tres veces, puedo preguntarme $P(X\le 1.3)$

- Si $T$ es el tiempo de espera para que la mujer del Muelle de San Blas tiene que esperar a que su amado llegue, puedo preguntarme por el valor de $P(T\le 5\mbox{ años})$

Ahora, notemos que $P(X\le x)$ es un número. De hecho ese número está en el intervalo $[0,1]$

Así, dada la $X$, el número $P(X\le x)$ se denota por $F_X(x)$

Por ejemplo:
- $P(X\le 1.3)=F_X(1.3)$
- $P(T\le 5)=F_T(5)$

Por lo tanto $F_X:\mathbb{R}\to[0,1]$ dada por $$F_X(x)=P(X\le x)$$

**Ejemplo.** Volvamos al lanzamiento de dos monedas. Sea $X$ el número de caras. Sabemos que:

- $P(X=0)=\frac{1}{4}$
- $P(X=1)=\frac{1}{2}$
- $P(X=2)=\frac{1}{4}$

$\Omega=\{CC,CS,SC,SS\}$

Calculemos $F_X(x)$ para cualquier valor de $x$

- Si $x<0$, entonces $F_X(x)=P(X\le x)=0$

- Si $0\le x<1$, entonces $F_X(x)=P(X\le x)=P(X=0)=\frac{1}{4}$

- Si $1\le x< 2$, entonces $F_X(x)=P(X\le x)=P(X=0)+P(X=1)=\frac{1}{4}+\frac{1}{2}=\frac{3}{4}=0.75$

- Si $2\le x$, entonces $F_X(x)=P(X\le x)=P(X=0)+P(X=1)+P(X=2)=1$

Así, al final tenemos $$F_X(x)=\left\{\begin{array}{l}0\mbox{ si }x<0\\\frac{1}{4}\mbox{ si }0\le x<1\\\frac{3}{4}\mbox{ si }1\le x<2\\1\mbox{ si }2\le x\end{array}\right.$$

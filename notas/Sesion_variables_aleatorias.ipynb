Si $X$ es una variable aleatoria y $x$ es un número real, puedo preguntarme por $P(X\le x)$

- Si $X$ es el número de caras cuando lanzo una moneda tres veces, puedo preguntarme $P(X\le 1.3)$

- Si $T$ es el tiempo de espera para que la mujer del Muelle de San Blas tiene que esperar a que su amado llegue, puedo preguntarme por el valor de $P(T\le 5\mbox{ años})$

Ahora, notemos que $P(X\le x)$ es un número. De hecho ese número está en el intervalo $[0,1]$

Así, dada la $X$, el número $P(X\le x)$ se denota por $F_X(x)$

Por ejemplo:
- $P(X\le 1.3)=F_X(1.3)$
- $P(T\le 5)=F_T(5)$

Por lo tanto $F_X:\mathbb{R}\to[0,1]$ dada por $$F_X(x)=P(X\le x)$$

**Ejemplo.** Volvamos al lanzamiento de dos monedas. Sea $X$ el número de caras. Sabemos que:

- $P(X=0)=\frac{1}{4}$
- $P(X=1)=\frac{1}{2}$
- $P(X=2)=\frac{1}{4}$

$\Omega=\{CC,CS,SC,SS\}$

Calculemos $F_X(x)$ para cualquier valor de $x$

- Si $x<0$, entonces $F_X(x)=P(X\le x)=0$

- Si $0\le x<1$, entonces $F_X(x)=P(X\le x)=P(X=0)=\frac{1}{4}$

- Si $1\le x< 2$, entonces $F_X(x)=P(X\le x)=P(X=0)+P(X=1)=\frac{1}{4}+\frac{1}{2}=\frac{3}{4}=0.75$

- Si $2\le x$, entonces $F_X(x)=P(X\le x)=P(X=0)+P(X=1)+P(X=2)=1$

Así, al final tenemos $$F_X(x)=\left\{\begin{array}{l}0\mbox{ si }x<0\\\frac{1}{4}\mbox{ si }0\le x<1\\\frac{3}{4}\mbox{ si }1\le x<2\\1\mbox{ si }2\le x\end{array}\right.$$

## Relación de $f$ con $F$ para v.a. discretas

\begin{eqnarray*}
F(x)&=&P(X\le x)\\
&=&P(X=x)+P(X=x-1)+P(X=x-2)+P(X=x-3)+...\\
&=&f(x)+f(x-1)+f(x-2)+f(x-3)+...\\
&=&f(x)+\textcolor{red}{f(x-1)+f(x-2)+f(x-3)+...}\\
&=&f(x)+\textcolor{red}{F(x-1)}
\end{eqnarray*}

$$F(x)-F(x-1)=f(x)=P(X=x)$$

**Ejemplo.** Se lanzan dos monedas. Para $X =$ número de caras:

$$f(0) = 1/4, \quad f(1) = 1/2, \quad f(2) = 1/4$$

Verificamos que $\sum f(x) = 1/4 + 1/2 + 1/4 = 1$ 

La función de distribución se puede reconstruir a partir de la función de probabilidad:

$F(0)=P(X\le 0)=P(X=0)=f(0)=1/4$

$F(1)=P(X\le 1)=P(X=1)+P(X=0)=f(1)+f(0)=1/2+1/4=3/4$

$F(2)=P(X\le 2)=P(X=2)+P(X=1)+P(X=0)=f(2)+f(1)+f(0)=1$

**Ejemplo**

Si $X$ es una v.a. con distribución $F(x)=\begin{cases}
0, & x < 0 \\
1 - e^{-0.5x}, & x \geq 0
\end{cases}$

Entonces $f(x)=\frac{dF(x)}{dx}$

Esto es: $$f(x)=\begin{cases}
0, & x < 0 \\
0.5e^{-0.5x}, & x \geq 0
\end{cases}$$

Me puedo preguntar cuánto es $$P(-2\le X<5)=\int_{-2}^5f(x)\,\mathrm{d}x=\textcolor{purple}{\int_{-2}^0f(x)\mathrm{d}x}+\int_0^5f(x)\mathrm{d}x=\int_0^5f(x)\mathrm{d}x=0.9179$$

Qué pasaría si me piden $P(-2<X<5)$

En general, si $X$ es v.a. **continua** entonces no importa si ponemos $\le$ o $<$

El resultado es el mismo porque $P(X=x)=0$

Es decir, es lo mismo $$P(-2\le X\le 5)=P(-2<X\le 5)=P(-2\le X<5)=P(-2<X<5)$$

**¿Por qué?**

Porque $P(X\le a)=P(X<a)+P(X=a)=P(X<a)+0=P(X<a)$

**Diferencia entre función de probabilidad y función de densidad**

- Si $X$ es discreta, entonces $f(x)=P(X=x)$

. Si $X$ es continua, entonces $$f(x)=\frac{dF(x)}{dx}$$
